مقدمه
        </font>
		<p></p>
		<hr>
        اگر بدانیم که ۱۷ درصد جمعیت کشور بالای ۵۰ سال سن دارند و آمارها نشان داده اند که ۲ درصد جمعیت کل کشور افراد بالای ۵۰ سالی هستند که سکته کرده اند، احتمال رخداد سکته قلبی در افراد بالای ۵۰ سال چقدر است؟
        <br>
        <br/>
        حال فرض کنید به طور معجزه آسایی پژوهش ها نشان دهد که سکته قلبی و سن بالا از یکدیگر مستقل هستند. با این فرض، احتمال رخداد سکته قلبی را در میان تمامی افراد جمعیت محاسبه کنید.

<p></p>
<br />
<div id="sec_cp" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
احتمال شرطی
        </font>
		<p></p>
		<hr>
احتمال شرطی یکی از مباحث مهم و اصلی در حوزه ی احتمال است. احتمال رخداد واقعه $A$ به شرط رخداد واقعه $B$ را به صورت زیر تعریف می‌کنیم:
        $$P(A|B) = \frac{P(A \cap B)}{P(B)}$$
        <br>
         <figure>
  <img src="figs/conditional_b.png" style="float:center; width:40%"/>
  <figcaption>
    <div align="center">
                فضای احتمال $A|B$
    برابر است با فضایی از $A$ که در 
   $B$
    دیده می‌شود.
    </div>
    </figcaption>
</figure> 
        <br/>
        برای مثال،اگر واقعه $A$ احتمالی برابر با ۰.۲، واقعه $B$ احتمالی برابر با ۰.۳ و واقعه $A \cap B$ احتمالی برابر با احتمالی برابر با ۰.۰۱ داشته باشد، احتمال $A|B$ برابر است با:
        $$P(A|B) = \frac{P(A \cap B)}{P(B)} = \frac{0.01}{0.2} = \frac{1}{20}$$
        <br>
        <br/>
        حال به حل مثالی دیگر می پردازیم. فرض کنید تاسی را پرتاب کرده‌ایم. با این فرض که می دانیم عدد به دست آمده عددی زوج است، احتمال این که عدد به دست آمده از ۴ کمتر باشد چقدر است؟
        <br>
        <br/>
        پاسخ: فرض کنید واقعه $A$ را برابر با رخ‌داد عدد کمتر از ۴ در نظر بگیریم و واقعه $B$ را برابر با رخ‌داد عددی زوج بگیریم. حال خواهیم داشت:
        $$P(A|B) = \frac{P(A \cap B)}{P(B)}$$
        $$P(A \cap B) = \frac{1}{6}$$
        $$P(B) = \frac{1}{2}$$
        $$ \Longrightarrow P(A|B) = \frac{\frac{1}{6}}{\frac{1}{2}} = \frac{1}{3}$$
        <br>
        <br/>
	</font>
</div>

<div style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=blue size=6>
کنجکاوی:
        </font><br/>
به نظر شما، آیا رابطه $P(A|B) \leq P(A)$  همواره برقرار است؟ در صورت مثبت بودن جواب، می‌توانید آن را ثابت کنید؟ در صورت منفی بودن جواب، می‌توانید مثال نقضی برای آن بیابید؟
        <br/>
آیا می توان احتمالی را در دنیای واقعی مثال زد که شرطی نباشد؟
	</font>
</div>

<p></p>
<br />
<div id="sec_ind" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
استقلال
        </font>
		<p></p>
		<hr>
همانطور که پیش‌تر دیدیم دو رویداد $A$ و $B$ مستقل‌اند هرگاه:
        $$P(A \cap B) = P(A)P(B)$$
         حال که با احتمال شرطی آشنا شدیم بد نیست استقلال دو متغیر را از این دیدگاه نیز بررسی کنیم. 
        <br/>
        <br/>
         اگر دو رویداد $A$ و $B$  مستقل باشند داریم:
         $$P(A \cap B) = P(A)P(B) \\
         \Rightarrow P(A) = \frac{(A \cap B)}{P(B)}=P(A|B) $$
        <br/>
        این بدان معنا است که چه رویداد $B$ رخ دهد و چه اطلاعی از رویداد $B$ نداشته باشیم، احتمال رویداد $A$ تغییری نمی کند.
	</font>
</div>

<div style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=darkpink size=6>
تمرین:
        </font>
        نشان دهید دو رویداد $A$ و $B$ مستقل‌اند اگر و فقط اگر 
        $$.P(A|B) = P(A|\overline{B})$$
        تعبیر این عبارت ریاضی به زبان طبیعی چیست؟
</font>
</div>

<div style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=darkpink size=6>
تمرین:
        </font>
گزاره های زیر را ثابت کنید:
        <br>
        <br/>
        ۱) اگر وقایع $A$ و $B$ مستقل باشند، نقیض $A$ و $B$ نیز مستقلند.
        <br>
        <br/>
        ۲)اگر $A$ و $B$ مستقل باشند، نقیض $A$ و نقیض $B$ نیز مستقلند.
</font>
</div>

<div style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=blue size=6>
کنجکاوی:
        </font>
        اگر واقعه $A$ از واقعه $C$ و واقعه $B$ مستقل باشد، آیا می توان گفت که واقعه $A$ از واقعه $B \cap C$ مستقل است؟ به مفهوم این گزاره خوب دقت کنید. آیا می توان تعریفی از استقلال سه واقعه ارائه داد؟
	</font>
</div>

		<font color=#FF7500 size=6>
بازگشت به مسأله سکته
        </font>
		<p></p>
		<hr>
داریم:
        $$P(Heart \ attack \ | \ age \ greater \ than \ 50) = \frac{\frac{2}{100}}{\frac{17}{100}} = \frac{2}{17}$$
        و پس از اعلام استقلال فرض‌ها داریم:
        $$P(Heart \ attack \ | \ age \ greater \ than \ 50) = P(Heart \ attack) = \frac{2}{17}$$
	</font>
</div>

<br />
<img src="figs/game_show_1.png" style="float:center; width:50%"/>

شما یکی از درها را به صورت تصادفی انتخاب می کنید. پس از انتخاب شما، مجری در اقدامی ناگهانی، اعلام می کند که قصد دارد به شما در انتخاب کمکی کند. او یکی از درهایی را که شما انتخاب نکرده بودید باز می‌کند و بز پشت آن نمایان می‌شود. سپس مجری از شما می‌پرسد که آیا مایل هستید که انتخاب خود را تغییر دهید؟

<div id="sec_tot" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
قانون احتمال کل
        </font>
		<p></p>
		<hr>
قانون احتمال کل ارتباط میان یک پیشامد و افرازهای فضای نمونه به بخش های مختلف را بیان می کند. فرض کنید $B_1,B_2,...,B_m$ افراز فضای نمونه $\Omega$ به $m$ زیربخش باشد. حال اگر $A$ را پیشامد رخ داده در $\Omega$ درنظر بگیریم، داریم: 
        $$P(A) = P(A|B_1)P(B_1) + P(A|B_2)P(B_2) + ... + P(A|B_m)P(B_m)$$
        <br>
        <br/>
        <font color=#FF7500 size=5>
مثال
        </font>
: فرض کنید دو جعبه لامپ از کارخانه رسیده است. جعبه اول شامل ۲۱ لامپ است که ۱۲ تای آن سالم و ۹ تای آن خراب است. جعبه دوم شامل ۲۱ لامپ است که ۱۰ تای آن سالم و ۱۱ تای آن خراب است. لامپی را از جعبه بیرون می‌آوریم. احتمال این که آن لامپ خراب باشد چقدر است؟
        <br>
        <br/>
        پاسخ: داریم:
        $$P(Bad \ light \ bulb) = P(Bad \ light \ bulb \ | \ Box \ 1)P(Box \ 1) + P(Bad \ light \ bulb \ | \ Box \ 2)P(Box \ 2)$$
        $$ \Longrightarrow P(Bad \ light \ bulb) = \frac{1}{2} \cdot \frac{9}{21} + \frac{1}{2} \cdot \frac{11}{21} = \frac{10}{21}$$
	</font>
</div>

<p></p>
<br />
<div id="sec_bayes" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
قاعده بیز
        </font>
		<p></p>
		<hr>
در تمامی بحث‌های پیشین، هدف ما یافتن احتمال رخ‌داد یک واقعه $A$ به شرط وقوع قاعده $B$ بود. اما گاهی این موضوع پیش می‌آید که ما احتمال رخ‌داد واقعه $B$ به شرط وقوع واقعه $A$ را نیاز داریم. برای محاسبه این احتمال، باید از قاعدهٔ بیز کمک بگیریم. قاعده بیز به صورت زیر بیان می‌شود:
        $$P(B|A) = \frac{P(A|B)P(B)}{P(A)}$$
        <br>
        <br/>
        <font color=#FF7500 size=5>
مثال
        </font>
: در مثال قسمت قبل، احتمال این که لامپ خراب متعلق به جعبه دوم باشد چقدر است؟
        <br>
        <br/>
        پاسخ: داریم:
        $$P(Box \ 2 \ | \ Bad \ light \ bulb) = \frac{P(Bad \ light \ bulb \ | \ Box \ 2)P(Box \ 2)}{P(Bad \ light \ bulb)}$$
        $$ \Longrightarrow P(Box \ 2 \ | \ Bad \ light \ bulb) = \frac{\frac{1}{2} \cdot \frac{11}{21}}{\frac{10}{21}} = \frac{11}{20}$$
	</font>
</div>

<div style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=blue size=6>
کنجکاوی:
        </font><br/>
به نظر شما، قاعده بیز در چه زمینه‌هایی کاربرد دارد؟ نمونه‌هایی از کاربردهای آن را بیابید.
	</font>
</div>

<div id="sec_problem" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
بازگشت به مسأله مسابقه تلویزیونی
        </font>
		<p></p>
		<hr>
واقعه های $A$، $B$ و $C$ را به ترتیب وجود داشتن ماشین در درهای ۱، ۲ و ۳ بگیرید. به وضوح می دانیم:
        $$P(A) = P(B) = P(C) = \frac{1}{3}$$
        حال بدون از دست دادن کلیت فرض کنید ما در شماره ۱ را انتخاب کرده باشیم و مجری در ۲ را برایمان باز کند.
         احتمال رخداد $B$ را توسط مجری با $P(HB)$ نشان می دهیم. حال روابط زیر به وضوح برقرارند:
        $$P(HB|A) = \frac{1}{2}, \ P(HB|B) = 0, \ P(HB|C) = 1$$
        برای بررسی تعویض در، کافیست احتمال های $P(A|HB)$ و $P(C|HB)$ را محاسبه کنیم. طبق قاعده بیز داریم:
        $$P(A|HB) = \frac{P(HB|A)P(A)}{P(HB)} = \frac{P(HB|A)P(A)}{P(HB|A)P(A) + P(HB|B)P(B) + P(HB|C)P(C)} = \frac{\frac{1}{2} \cdot \frac{1}{3}}{\frac{1}{2} \cdot \frac{1}{3} + 0 \cdot \frac{1}{3} + 1 \cdot \frac{1}{3}} = \frac{1}{3}$$
        $$P(C|HB) = \frac{P(HB|C)P(C)}{P(HB)} = \frac{P(HB|C)P(C)}{P(HB|A)P(A) + P(HB|B)P(B) + P(HB|C)P(C)} = \frac{1 \cdot \frac{1}{3}}{\frac{1}{2} \cdot \frac{1}{3} + 0 \cdot \frac{1}{3} + 1 \cdot \frac{1}{3}} = \frac{2}{3}$$
        بنابراین، تعویض در انتخاب کار معقولانه‌ای خواهد بود.
	</font>
</div>

	<font face="XB Zar" size=5>

خلاصه و نتیجه‌گیری
        </font>
		<p></p>
		<hr>
        در این بخش با موارد زیر آشنا شدیم:
        <ul>
            <li>
            احتمال شرطی
            </li>
            $$P(A|B) = \frac{P(A \cap B)}{P(B)}$$
            <li>
            استقلال
            </li>
            $$P(A|B) = P(A) (چرا؟)$$ 
            <li>
            به کار گیری احتمال شرطی و استقلال برای حل مسائل احتمال
            </li>
            به وابسته‌ها و وابستگی‌ها اهمیت بده!
            <li>
            قانون احتمال کل
            </li>
            $$P(A) = P(A|B_1)P(B_1) + P(A|B_2)P(B_2) + ... + P(A|B_m)P(B_m)$$
            <li>
            قاعده بیز
            </li>
            $$P(B|A) = \frac{P(A|B)P(B)}{P(A)}$$
            <li>
            مسأله مسابقه تلویزیونی
            </li>
            گاهی هم به مجری مسابقه اعتماد کن!
        </ul>
