<div id="sec_intro_pmf" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مقدمه
        </font>
		<p></p>
		<hr>
در بخش قبل با تعریف متغیر تصادفی و محاسبهٔ احتمال‌های مرتبط با آن آشنا شدیم. 
        <br />
همان طور که احتمالا متوجه شدید، استفادهٔ دائم از عبارات $P(x(\omega)=a)$ و یا $P(x(\omega)\le a)$ سخت و خسته‌کننده است. به منظور حل این مشکل، از دو مفهوم تابع جرمی احتمال (PMF) و تابع توزیع انباشته (CDF) استفاده می‌کنیم.
<br />
(توجه کنید که در این جا متغیر تصادفی حقیقی مد نظر ماست؛ متغیر تصادفی مختلط به هر یک از نقاط دامنه عددی مختلط را نسبت می‌دهد)
	</font>
</div>

<p></p>
<br />
<div id="sec_pmf" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
تابع جرمی احتمال
        </font>
		<p></p>
		<hr>
        مهم‌ترین راه برای شناخت یک متغیر تصادفی، احتمال وقوع مقادیر ممکن برای آن است. برای متغیرهای تصادفی گسسته، این مقادیر از طریق تابع جرمی احتمال (Probability Mass Function) مشخص می‌شوند.
        <br />
        فرض کنید $X$ یک متغیر تصادفی و $x$ یکی از مقادیر ممکن برای آن باشد. چگالی احتمال $x$ که با $p_X(x)$ نمایش داده می‌شود برابر با احتمال وقوع رویداد $\{X=x\}$ است. یعنی:
        $$p_X(x)=P(\{X=x\})$$
        اگر تنها با یک متغیر تصادفی سر و کار داشته باشیم، می‌توانیم اندیس $X$ را حذف کرده و $p_X(x)$ را به صورت $p(x)$ بنویسیم.
        <br />
        در یکی از مثال‌های بخش قبل، آزمایش پرتاب ۳ سکهٔ سالم را در نظر گرفته و متغیر تصادفی $X$ را برابر با تعداد شیرهای موجود در این سه پرتاب تعریف کردیم.
        <br />
        حال برای $i=0, 1, 2, 3$ داریم:
        <br />
        $$P_X(i)=P(\{X=x\})={3\choose{i}}(0.5)^i(0.5)^{3-i}=\frac{{3\choose{i}}}{8}=\Bigg\{
	\begin{array}{ll}
		\frac{1}{8}  & x = 0, 3 \\
		\frac{3}{8}  & x = 1, 2
	\end{array}$$
    برای سایر $i$ها هم داریم: $P_X(i)=0$.
    <br />
    تصویر زیر تابع $P_X(x)$ را نمایش می‌دهد که تنها در یک سری نقاط محدود یا قابل شمارش مقدار غیرصفر دارد.
    <img src="figs/pmf.png" style="float:center; width:40%"/>
    <br />
    <br />
    با توجه به این که هر یک از $P(x_i)$هااحتمال وقوع یک پیشامد را نشان می‌دهند، مشخص است که مقدار این تابع همواره باید مقداری بین صفر تا یک باشد. همچنین داریم:
    $$\sum_{i=0}^\infty P(x_i)=1$$
    با استفاده از مقادیر تابع جرمی احتمال، به راحتی می‌توان احتمال هر واقعه را محاسبه کرد. مثلا در همین مثال داریم:
    $$P(\{X\le1.1\})=P(\{X=0\})+P(\{X=1\})=P_X(0)+P_X(1)=\frac{1}{8}+\frac{3}{8}=\frac{4}{8}$$
    دقت کنید که در این جا متغیر تصادفی گسسته بود، یعنی مقادیری که می‌توانست اختیار کند شمارا بود. در حالتی که توزیع مد نظر پیوسته باشد از مفهومی به نام تابع توزیع احتمال (PDF) استفاده می‌کنیم که در آینده با آن بیشتر آشنا خواهید شد.
	</font>
</div>

<p></p>
<br />
<div id="sec_cdf" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
تابع توزیع انباشته
        </font>
		<p></p>
		<hr>
     مفهومی دیگری که برای بررسی یک متغیر تصادفی به کار می‌رود، تابع توزیع انباشته (Cumulative Distribution Function) است که بدین صورت تعریف می‌شود:
        $$F_X(x)=P(\{X\le x\})$$
        مشابه تابع جرمی احتمال، اگر تنها یک متغیر تصادفی در حال بررسی باشد اندیس $X$ را می‌توانیم حذف کنیم.
        <br />
        نکتهٔ قابل توجه این است که اگر احتمال‌های $\{X\le x\}$ را به ازای تمام $x$ها داشته باشیم، احتمال تمام واقعه‌ها مشخص خواهد شد. یعنی $F(x)$ می‌تواند یک متغیر تصادفی را به طور کامل توصیف کند.
        <br />
        در مثالی که در بخش قبل بررسی شد داریم:
        $$F_X(-0.001)=P(\{X\le0.001\})=0$$
        $$F_X(0)=P(\{X\le0\})=\frac{1}{8}$$
        $$F_X(1.001)=P(\{X\le1.001\})=\frac{4}{8}$$
        $$F_X(10)=P(\{X\le=10\})=1$$
        نمودار این تابع نیز به صورت زیر است:
        <img src="figs/cdf.png" style="float:center; width:40%"/>
       <br />
       برای متغیرهای تصادفی گسسته، تابع $F_X(x)$ به صورت پلکانی است که در نمودار بالا نیز مشاهده می‌شود.
	</font>
</div>

<p></p>
<br />
<div id="sec_attr" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
خواص تابع توزیع انباشته
        </font>
		<p></p>
		<hr>
     در ادامه به تعدادی از خواص تابع توزیع انباشته اشاره می‌شود:
<ol>
    <li>
        $$F(-\infty)=0$$
        زیرا:
        $$F(-\infty)=P(\{X\leq-\infty\})=0$$
    </li>
    <br />
    <li>
        $$F(+\infty)=1$$
        زیرا:
        $$F(+\infty)=P(\{X\leq+\infty\})=1$$
    </li>
    <br />
    <li>
        $$x_1<x_2\Rightarrow F(x_1)\le F(x_2)$$
        این ویژگی به این معناست که $F(x)$ تابعی غیرنزولی است. زیرا:
        $$x_1<x_2\Rightarrow\{\omega:X(\omega)\le x_1\}\subset\{\omega:X(\omega)\le x_2\}\Rightarrow P(\{X\le x_1\})\le P(\{X\le x_2\})\Rightarrow F(x_1)\le F(x_2)$$
        همچنین از این جا نتیجه می‌شود که: $0\le F(x)\le1$، چون دیدیم که در $-\infty$ و $+\infty$ مقدار تابع $F$ به ترتیب صفر و یک است.
    </li>
    <br />
    <li>
    $$P(\{X>x\})=1-F(x)$$
    زیرا:
    $$\{\omega:X(\omega)>x\}=\{\omega:X(\omega)\le x\}^c\Rightarrow P(\{X>x\})=1-P(\{X\le x\})=1-F(x)$$
    </li>
    <br />
    <li>
    $$P(\{x_1<x\le x_2\})=F(x_2)-F(x_1)$$
    زیرا (توجه کنید که دو مجموعهٔ حاصل از تجزیه از یک‌دیگر جدا هستند):
    $$\{X\le x_2\}=\{X\le x_1\}\cup\{x_1<X\le x_2\}\Rightarrow P(\{X\le x_2\})=P(\{X\le x_1\})+P(\{x_1<X\le x_2\})\Rightarrow P(\{x_1<X\le x_2\})=F(x_2)-F(x_1)$$
    </li>
</ol>
	</font>
</div>

<p></p>
<br />
<div id="sec_intro_expected" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مقدمه
        </font>
		<p></p>
		<hr>
در قسمت‌های قبلی درس با احتمال و متغیرهای تصادفی آشنا شدیم  و در مسائل، احتمال وقوع رخدادهای مختلف را به صورت جداگانه حساب کردیم. حال سوالی که برای ذهن ایجاد می‌شود این است که
چگونه با دانستن احتمال وقوع رویدادهای مختلف و با ترکیب این اطلاعات با یک‌دیگر، می‌توان شهودی نسبت به نتیجه یک آزمایش تصادفی به دست آورد؟
        <br/>
        یکی از ساده‌ترین جواب‌ها به سوال بالا استفاده از مفهوم امید ریاضی است. به عبارت ساده و نادقیق، امید ریاضی میانگین وزن‌دار بر روی متغیر تصادفی است.
        <br/><br />
به طور مثال، متغیر تصادفی
$X$
مقدار
۱
را در صورتی که نتیجه پرتاب سکه شیر شود، خروجی می‌دهد و در غیر این‌صورت مقدار آن
صفر
خواهد بود.     
حال اگر سکه متقارن باشد برای محاسبه میانگین وزن‌دار یا همان امید ریاضی در این حالت خواهیم داشت:
$$
\frac{1}{2}\times 1 + \frac{1}{2}\times 0 = \frac{1}{2}
$$
عدد
$\frac{1}{2}$
دقیقاً در وسط خروجی‌های متغیر تصادفی قرار گرفته و از ۰ و ۱ به یک فاصله است. بنابراین این عدد با اطلاع ما که سکه
((متقارن))
است، همخوانی دارد.
<br/>
به طور شهودی یعنی اگر این سکه به تعداد زیادی بار پرتاب شود،
میانگین خروجی متغیر تصادفی
$X$
در این پرتاب‌ها
برابر با
$\frac{1}{2}$
خواهد بود. این جمله بیانی از قانون اعداد بزرگ است که به صورت دقیق‌تر در جلسه‌های آتی مطالعه خواهیم کرد.
        </font>
</div>

<p></p>
<br />
<div id="sec_definition" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
تعریف امید ریاضی
        </font>
		<p></p>
		<hr>
برای متغیر تصادفی گسسته
        $X$
امید ریاضی به صورت زیر تعریف می‌شود:
$$\mathop{\mathbb{E}}[X] = \sum_{x:p(x)>0}{xp(x)}$$
که در آن
p
تابع جرم احتمال متغیر تصادفی
$X$
است.
به عبارت دیگر تعریف بالا معادل این است که بر روی تمام حالت‌های ممکن متغیر
$X$
میانگین وزن‌دار بر حسب میزان محتمل بوده‌شان گرفته شود.
<br/>
	</font>
</div>

<p></p>
<br />
<div id="sec_sum_example" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مثال حالت گسسته
        </font>
		<p></p>
		<hr>
فرض کنید که یک تاس سالم با ۶ وجه داریم. اگر
متغیر تصادفی
$X$
را برابر عدد خروجی پرتاب یک بار تاس تعریف کنیم،
امید ریاضی آن چند است؟

$$\mathop{\mathbb{E}}[X] = \sum_{x:p(x)>0}{xp(x)}$$
$$
\mathop{\mathbb{E}}[X] = 1(\frac{1}{6}) + 2(\frac{1}{6}) + 3(\frac{1}{6}) + 4(\frac{1}{6}) + 5(\frac{1}{6}) + 6(\frac{1}{6}) = \frac{7}{2}
$$
</font>
</div>

<p></p>
<br />
<div id="sec_function" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
امید ریاضی تابعی از متغیر تصادفی
        </font>
		<p></p>
		<hr>
        تابعی از یک متغیر تصادفی خود یک متغیر تصادفی است.
 در برخی از کاربردها نیاز به محاسبه امید ریاضی تابعی از یک متغیر تصادفی است. به طور مثال محاسبه‌ی امید ریاضی مربع یک متغیر تصادفی
از مواردی است که در درس به زودی با آن مواجه خواهید شد.
<br/>
نکته‌ای که باید به آن توجه کرد این است که امید ریاضی تابعی از یک متغیر لزوماً برابر با اعمال آن تابع بر روی امید ریاضی نیست.
<br/>
به طور مثال فرض کنید متغیر تصادفی
$X$
دو مقدار
۰
و
۲
را با احتمال برابر بتواند اختیار کند. داریم:
$$
\mathop{\mathbb{E}}[X] = 0 \times \frac{1}{2} + 2 \times \frac{1}{2} = 1
$$
این در حالی است که امید ریاضی متغیر تصادفی
$Y = g(X) = X^2$
برابر است با:
$$
\mathop{\mathbb{E}}[Y] = 0^2 \times \frac{1}{2} + 2^2 \times \frac{1}{2} = 2
$$
در نتیجه
$\mathop{\mathbb{E}}[X^2] = 2 \neq (\mathop{\mathbb{E}}[X])^2 = 1^2$
و در حالت کلی نتیجه می‌شود که:
$$
g(\mathop{\mathbb{E}}[X]) \neq \mathop{\mathbb{E}}[g(X)]
$$
<br/>
در حالت کلی امید ریاضی تابع
$g$
از متغیر تصادفی
X
برابر است با
$$
\mathop{\mathbb{E}}[g(X)] = \sum_{i}{g(x_i)p(x_i)}
$$
که در آن
به ازای
$i \geq 1$،
$x_i$
ها
همه مقادیری هستند که
$X$
می‌تواند اختیار کند.
<br/>
برای اثبات رابطه بالا فرض می‌کنیم که به ازای
$i \geq 1$،
های مختلف،
مقادیر
$y_j$
ها که
$j \geq 1$
برابر با مقادیر متفاوت
$g(x_i)$
ها باشند. سپس از طریق دسته‌بندی همه
$i$
هایی که
$g(x_i)$
یکسان است،
رابطه ثابت می‌گردد.
        $$
        \begin{aligned}
        \sum_{i}{g(x_i)p(x_i)} &= \sum_{j}\sum_{i:g(x_i)=y_j}{g(x_i)p(x_i)}\\
        &= \sum_{j}\sum_{i:g(x_i)=y_j}{y_jp(x_i)}\\
        &= \sum_{j}y_j\sum_{i:g(x_i)=y_j}{p(x_i)}\\
        &= \sum_{j}y_jp\{g(X)=y_j\}\\
        &= \mathop{\mathbb{E}}[g(X)]
        \end{aligned}
        $$
<br/>
	</font>
</div>

<p></p>
<br />
<div id="sec_definition_example_discrete" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مثال
        </font>
		<p></p>
		<hr>
تکه چوبی به طول ۱ واحد داریم.
فرض کنید یک سر این چوب منطبق بر عدد
0
محور افقی و سر دیگر آن منطبق بر عدد
1
محور می‌باشد.
        این تکه چوب را از نقطه
$\frac{1}{2^U}$
می‌شکنیم که
U
متغیر تصادفی‌ای با تابع جرم احتمال به صورت زیر است
:
$$
P_U(u) = \begin{cases}
       \frac{1}{2^u} &\quad u\in \mathop{\mathbb{N}}\\
       0 &\quad\text{otherwise} \\ 
     \end{cases}
$$
عدد ثابت
$p$
که
$p \in \mathop{\mathbb{N}}$
است،
به ما داده شده است.
امید ریاضی طول قطعه چوبی را که شامل نقطه
$\frac{1}{2^p}$
بر روی محور است،
به دست آورید.
به شکل زیر توجه کنید.
<br/>
<img src="figs/wood.png" style="float:center; width:35%"/>
<br />
در حالت تساوی دو عدد
$p$
و
$u$
قطعه چوبی را که یک انتهای آن عدد صفر محور است،
در نظر بگیرید.
<br/>
<br/>
<br/>
تابع
$L_p(U)$
از متغیر تصادفی
U
را تعریف می‌کنیم. این تابع خود یک متغیر تصادفی خواهد بود و امید ریاضی آن پاسخ به سوال است
$$
L_p(U) = \begin{cases}
       \frac{1}{2^U} &\quad U \leq p\\
       1 - \frac{1}{2^U} &\quad U > p \\ 
     \end{cases}
$$
حال برای به دست آوردن امید ریاضی این تابع داریم:
$$
\begin{aligned}
\mathop{\mathbb{E}}[L_p(U)] &= \sum_{u=1}^{\infty}{L_p(u)P_U(U=u)} \\
&= \sum_{u=1}^{p}{(\frac{1}{2^u}).P_U(U=u)} + \sum_{u=p+1}^{\infty}{(1 - \frac{1}{2^u}).P_U(U=u)}\\
&= \sum_{u=1}^{p}{(\frac{1}{2^u})(\frac{1}{2^u})} + \sum_{u=p+1}^{\infty}{(1 - \frac{1}{2^u})(\frac{1}{2^u})}\\
&= \frac{1}{3}(1 - \frac{1}{4^{p+1}}) + \frac{1}{2^p} - \frac{1}{3}(\frac{1}{4^p})
\end{aligned}
$$
	</font>
</div>

<p></p>
<br />
<div id="sec_sum" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
خطی بودن امید ریاضی و جمع متغیرهای تصادفی
        </font>
		<p></p>
		<hr>
.یکی از ویژگی‌های مهم امید ریاضی خطی بودن آن است
$$
\mathop{\mathbb{E}}[aX + b] = a\mathop{\mathbb{E}}[X] + b
$$
اثبات رابطه بالا به کمک قسمت قبل به دست می‌آید.
برای این کار تابع زیر را تعریف می‌کنیم:
$$
g(X) = aX + b
$$
سپس با توجه به قسمت قبل داریم:
$$
\begin{aligned}
\mathop{\mathbb{E}}[aX + b] &= \mathop{\mathbb{E}}[g(X)] \\
&= \sum_{i}{g(x_i)p(x_i)} \\
&= \sum_{i}{(ax_i + b)p(x_i)} \\
&= \sum_{i}{axp(x_i)} + b\sum_{i}{p(x_i)} \\
&= a\sum_{i}{xp(x_i)} + b.1 \\
&= a\mathop{\mathbb{E}}[X] + b
\end{aligned}
$$
 به طور خاص استفاده از این ویژگی و توابع نشانگر (که در ادامه این جلسه معرفی می‌شوند)
می‌تواند حل برخی از مسائل را بسیار ساده‌تر کند.
<br/>
<br/>
دیدیم که ضرب عدد ثابت در متغیر تصادفی یا جمع آن با عدد ثابت،
مقدار امید ریاضی را به همان میزان جابجا کرد.
حال سوال بعدی این است که اگر به جای عدد ثابت،
متغیر تصادفی را یک متغیر تصادفی دیگر جمع کنیم،
چه اتفاقی برای امید ریاضی آن‌ها می‌افتد؟
به عبارت دقیق تر اگر
$Z = X + Y$
باشد،
آیا
گزاره
$$
\mathop{\mathbb{E}}[Z] = \mathop{\mathbb{E}}[X + Y] = \mathop{\mathbb{E}}[X] + \mathop{\mathbb{E}}[Y]
$$
برقرار است؟
<br/>
<br/>
پاسخ مثبت است. اثبات به کمک استفاده از توزیع مشترک دو متغیر
$X$
و
$Y$
(در جلسات آتی توزیع مشترک را یاد می‌گیریم):
$$
\begin{aligned}
\mathop{\mathbb{E}}[Z] &= \mathop{\mathbb{E}}[(X + Y)]\\
&= \int_x\int_y(x+y)f_{XY}(x,y)dydx \\
&= \int_x\int_yxf_{XY}(x,y)dydx + \int_y\int_xyf_{XY}(x,y)dxdy \\
&= \int_xxf_{X}(x)dx + \int_yyf_{Y}(y)dy\\
&= \mathop{\mathbb{E}}[X] + \mathop{\mathbb{E}}[Y]
\end{aligned}
$$
همان طور که مشاهده می‌کنید در اثبات بالا در هیچ جا از فرض مستقل بودن یا نبودن دو متغیر تصادفی نسبت به یکدیگر استفاده نشده است.
یعنی این گزاره برای
<strong>
هر دو متغیر تصادفی‌ای، چه مستقل چه وابسته،
</strong>
برقرار است.
<br/>
<br/>
اثبات گفته شده در بالا کمی نقص دارد. در آن به نکته‌ای اشاره نشده است. آیا می‌توانید آن را پیدا کنید؟ 
	</font>
</div>

<div id="sec_indicator_example" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مثال
        </font>
		<p></p>
		<hr>
$n$
نفر
در یک اتاق کلاه‌هایشان را به هوا پرتاب می‌کنند. کلاه‌ها با یکدیگر جابه‌جا شده و سپس هر نفر به صورت تصادفی یک کلاه انتخاب می‌کند. امید ریاضی تعداد افرادی که همان کلاه اولیه خودشان را انتخاب می‌کنند،
چه مقدار است؟
<br/>
<br/>
مسئله را به کمک توابع نشانگر می‌توان ساده کرد.
توابع نشانگر
$
X_i ( 1 \leq i \leq n)
$
را به نحو زیر تعریف می‌کنیم:
$$
X_i = \begin{cases}
       1 &\quad \text{if the i-th person selects his own hat}\\
       0 &\quad \text{otherwise}\\ 
     \end{cases}
$$
امید ریاضی
متغیرهای نشانگر برابر با احتمال وقوع آن‌هاست. بنابراین:
$$
\mathop{\mathbb{E}}[X_i] = \frac{1}{n}
$$
دلیل آن این است که هر فرد طبق صورت سوال به صورت تصادفی و با احتمال برابر کلاه خود را پس از پرتاب به هوا انتخاب می‌کند.
<br/>
حال اگر متغیر مورد سوال را
$X$
بنامیم،
داریم:
$$
X = X_1 + \dots + X_n
$$
در نتیجه:
$$
\begin{aligned}
\mathop{\mathbb{E}}[X] &= \mathop{\mathbb{E}}[X_1 + \dots + X_n] \\
&=\mathop{\mathbb{E}}[X_1] + \dots + \mathop{\mathbb{E}}[X_n] \\
&= \frac{1}{n} + \dots + \frac{1}{n} \\
&= n(\frac{1}{n}) = 1
\end{aligned}
$$
	</font>
</div>

مثال
        </font>
		<p></p>
		<hr>
تعداد
$h$
شکارچی

$b$
پرنده داریم.

$p$
به پرنده اصابت می‌کند.
<br/>
امید ریاضی تعداد پرندگانی را که بدون اصابت تیر به آن‌ها جان سالم به در می‌برند،
محاسبه کنید.
<br/>
<br/>
متغیر تصادفی مد نظر را
$X$
می‌نامیم.
حال این متغیر را می‌توان به تعدادی تابع نشانگر تجزیه کرد.
$$
X_i = \begin{cases}
       0 &\quad \text{if the i-th bird gets shot}\\
       1 &\quad \text{otherwise}\\ 
     \end{cases}
$$
و مشاهده می‌کنیم که رابطه زیر برقرار است:
$$
X = X_1 + \dots + X_n
$$
از طرفی برای هر کدام از
$X_i$
ها داریم:
$$
\mathop{\mathbb{E}}[X_i] = p\{X_i = 1\}
$$
که برابر احتمال تیر نخوردن پرنده

است.
<br/>
حال احتمال این‌که شکارچی
j-ام
پرنده

را بزند
برابر است با:

\frac{1}{b} \times p = \frac{p}{b}

در نتیجه

تیر نخوردن پرنده

در مجموع

در نتیجه امید ریاضی

که جمع این متغیرهاست برابر است با:

دقت کنید

نیازمند فرض استقلال نیست، می‌توان امید ریاضی جواب را بر حسب جمع امید ریاضی تعدادی متغیر نشانگر نامستقل محاسبه کرد.

در بخش قبل با امید ریاضی یا مقدار میانگین آشنا شدیم و دیدیم که به کمک این کمیت مهم می‌توان شهود نسبت به توزیع

<div id="sec_def_variance" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
تعریف واریانس
        </font>
		<p></p>
		<hr>
       واریانس یک متغیر تصادفی گسسته
را که معیاری از پراکندگی داده خروجی‌های ممکن آن است،
به شرح زیر تعریف می‌کنیم:
$$
Var(X) = \mathop{\mathbb{E}}[(X - \mathop{\mathbb{E}[X]})^2]
$$
یا در برخی نمادگذاری‌های دیگر به صورت زیر تعریف می‌شود:
$$
Var(X) = \mathop{\mathbb{E}}[(X - \mu)^2]
$$
که در آن
$\mathop{\mathbb{E}}[X] = \mu$
است.
<br/>
با ساده کردن عبارت بالا داریم:
$$
\begin{aligned}
Var(X) &= \mathop{\mathbb{E}}[(X - \mu)^2] \\
&= \sum_{x}(x-\mu)^2p(x) \\
&= \sum_{x}(x^2+\mu^2-2\mu{x})p(x) \\
&= \sum_{x}x^2p(x) + \sum_{x}\mu^2p(x) -2\mu\sum_{x}xp(x)\\
&= \mathop{\mathbb{E}}[X^2] + \mu^2.1 -2\mu.\mu \\
&= \mathop{\mathbb{E}}[X^2] - \mu^2 \\
Var(X) &= \mathop{\mathbb{E}}[X^2] -  (\mathop{\mathbb{E}}[X])^2\\
\end{aligned}
$$
<br/>
همچنین به جذر واریانس، انحراف معیار می‌گویند.
$$
SD(X) = \sqrt{Var(X)}
$$
    </font>
</div>

<p></p>
<br />
<div id="sec_prop_variance" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
خواص واریانس
        </font>
		<p></p>
		<hr>
واریانس معیاری از پراکندگی خروجی‌های مختلف یک متغیر تصادفی است.
بنابراین انتقال یک متغیر تصادفی به اندازه یک عدد ثابت نباید اثری در آن داشته باشد چون پراکندگی نسبی خروجی‌های متغیر تصادفی انتقال‌یافته، نسبت به یکدیگر عوض نمی‌شود.
<br/>
همچنین واریانس با توان دوم مرتبط است. بنابراین دور از  
انتظار نیست که ضرب یک متغیر تصادفی در مقدار
ثابت رابطه توان دوم با واریانس متغیر اولیه داشته باشد.
<br/>
دو خاصیت بالا را در رابطه زیر می‌یابیم:
$$
\begin{aligned}
Var(aX + b) &= \mathop{\mathbb{E}}[(aX+b-a\mu-b)^2] \\
&= \mathop{\mathbb{E}}[a^2(X-\mu)^2] \\
&= a^2\mathop{\mathbb{E}}[(X-\mu)^2] \\
&= a^2Var(X)
\end{aligned}
$$
    </font>
</div>

<div id="sec_example_variance" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مثال
        </font>
		<p></p>
		<hr>
در جعبه‌ای
تعداد ۵ گوی آبی و ۵ گوی قرمز است.
دو گوی را به تصادف بدون جایگذاری از جعبه در می‌آوریم.
اگر دو گوی هم‌رنگ باشند
۱.۱
دلار
به دست می‌آوریم و اگر ناهم‌رنگ باشند ۱ دلار می‌بازیم.
اگر متغیر تصادفی مقدار پولی که
به دست می‌آوریم
،
$X$
باشد،
$Var(X)$
چند است؟
<br/>

$$
\mathop{\mathbb{E}}[X] = +1.1 \times (\frac{{5\choose2}+{5\choose2}}{{10\choose2}}) + (-1) \times (1-\frac{{5\choose2}+{5\choose2}}{{10\choose2}}) = -\frac{1}{15}
$$

$$
\mathop{\mathbb{E}}[X^2] = +1.1^2 \times (\frac{{5\choose2}+{5\choose2}}{{10\choose2}}) + (-1)^2 \times (1-\frac{{5\choose2}+{5\choose2}}{{10\choose2}}) = \frac{82}{75}
$$

$$
\begin{aligned}

&= \frac{82}{75} - (\frac{-1}{15})^2\\
&= 1.0889 \\
\end{aligned}
$$
