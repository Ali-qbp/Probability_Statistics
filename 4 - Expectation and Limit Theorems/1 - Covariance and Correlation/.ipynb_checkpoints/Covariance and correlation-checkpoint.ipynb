<div id="sec_reminder" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
یادآوری
        </font>
		<p></p>
        <hr>
        واریانس در یک متغیر تصادفی را به یاد آورید. قصد داشتیم میزان تغییر یک متغیر تصادفی نسبت به میانگینش را به دست آوریم. برای این کار ابتدا اختلاف هر مشاهده با میانگین را حساب کردیم. با جمع زدن این اختلافات به مقدار صفر رسیدیم. برای رفع این مشکل مجذور اختلافات را با هم جمع کرده و میانگین گرفتیم.
        <br> <br>
        	فرمول واریانس به صورت زیر بود: 
            <br> 
            $$Var(X) = E[(X - \mu_X)^2] = E[X^2] - \mu_X^2$$
	</font>
</div>

<p></p>
<br />
<div id="sec_cov_meaning" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
معنای کوواریانس
        </font>
		<p></p>
        <hr>
        با توجه به اینکه واژه‌ی co به معنای «هم» است، کوواریانس را با استفاده از معنی واریانس معنی می‌کنیم. کوواریانس میزان تغییر دو متغیر با همدیگر نسبت به میانگینشان است.
   <br> <br> 
        	فرمول کوواریانس به صورت زیر است:
            <br> 
            $$Cov(X,Y) = E[(X - \mu_X)(Y - \mu_Y)]$$
            <br> <br> 
            <font color=blue size=5>
            تمرین.
            </font> سعی کنید اثبات کنید$$Cov(X, Y) = E[XY] - \mu_x.\mu_y$$
            <br> <br> 
            هم‌چنین با کمی دقت متوجه می‌شوید که $$Cov(X, X) = Var(X)$$
	</font>
</div>

<div id="sec_cov_disc_cont" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
 محاسبه‌ی کوواریانس در متغیرهای تصادفی گسسته و پیوسته
        </font>
		<p></p>
        <hr>
        اگر دو متغیر تصادفی گسسته‌ی $X$ و $Y$ دارای توزیع توأم $p(x,y)$ باشند آنگاه کوواریانس آن دو مطابق زیر محاسبه می‌شود:
        <br>
        $$Cov(X,Y) = \sum_{i=1}^{n} \sum_{j=1}^{m} p(x_i, y_j)(x_i -\mu_X)(y_j - \mu_Y) = \Big(\sum_{i=1}^{n} \sum_{j=1}^{m} p(x_i, y_j)x_iy_j \Big)~-~ \mu_X\mu_Y$$
        <br> <br> 
        اگر دو متغیر تصادفی پیوسته‌ی $X$ و $Y$ بر روی بازه‌ی $[a, b]   \times [c,d]$ دارای توزیع توأم $f(x, y)$ باشند آنگاه کوواریانس آن دو مطابق زیر محاسبه می‌شود: 
        <br> 
        $$Cov(X, Y) = \int_c^d \int_a^b (x - \mu_x)\,(y - \mu_y)\,f(x,y) \,\mathrm{d}x\mathrm{d}y = \Big(\int_c^d \int_a^b x\,y\,f(x,y) \,\mathrm{d}x\mathrm{d}y\Big) - \mu_x\mu_y$$
        <br> 
        <font color=blue size=6>
مثال. 
        </font>
        <br/>
        دو متغیر تصادفی X و Y به صورت زیر تعریف شده‌اند:
        $$X \sim U(-1,\,1)$$
        $$Y = X^2$$
        <br>
        می‌خواهیم کوواریانس این دو را به‌دست آوریم. در واقع باید عبارت زیر را محاسبه کنیم:
        <br>
        $$Cov(X, Y) = E[XY] - E[X]E[Y] = E[X.X^2] - E[X]E[X^2] = E[X^3] - E[X]E[X^2]$$
        <br>
         اگر $X$ یک متغیر تصادفی با توزیع یونیفورم در محدوده‌ی $[a, b]$ باشد داریم:
         $$E[X] = \int_a^bx.f(x)\,\mathrm{d}x = \int_a^bx.\frac{1}{b - a}\,\mathrm{d}x = \frac{x^2}{2\,(b - a)}\big|_{x = a}^{x = b} = \frac{b + a}{2}$$
         <br>
         $$E[X^2] = \int_a^bx^2.f(x)\,\mathrm{d}x = \int_a^bx^2.\frac{1}{b - a}\,\mathrm{d}x = \frac{x^3}{3\,(b - a)}\big|_{x = a}^{x = b} = \frac{b^3 - a^3}{2\,(b - a)}$$
         <br>
         $$E[X^3] = \int_a^bx^3.f(x)\,\mathrm{d}x = \int_a^bx^3.\frac{1}{b - a}\,\mathrm{d}x = \frac{x^4}{4\,(b - a)}\big|_{x = a}^{x = b} = \frac{b^4 - a^4}{2\,(b - a)}$$
         <br> 
         با توجه به روابط بالا می‌توانیم عبارات زیر را بدست آوریم:
         <br>
         $$E[X] = 0 \quad \quad \quad  E[X^2] = \frac{1}{2} \quad  \quad \quad E[X^3] = 0$$
         <br>
         با جایگذاری در رابطه‌ی اول می‌رسیم به اینکه $Cov(X,Y) = 0$
         <br><br> 
         این را در R نیز تست می‌کنیم. با استفاده از دستور runif مشاهده‌هایی از متغیر تصادفی $X$ را تولید می‌کنیم و با استفاده از آنها مشاهده‌هایی از $Y$ را تولید می‌کنیم:
	</font>
</div>

```r
X <- runif(n = 10000, min = -1, max = 1)
Y <- X ^ 2
```

<div style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
    حال با استفاده از دستور cov کوواریانس آن دو را بدست می‌آوریم:
    </font>
</div>

```r
cov(X, Y)
```

        <br/><br>
        $
        1. \quad Cov(X_1 + X_2\,,Y) = Cov(X_1\,,Y) + Cov(X_2\,,Y) \\
        2. \quad Cov(aX + b\,,cY + d) = acCov(X\,,Y) \\
        3. \quad Var(X + Y) = Var(X) + Var(Y) + 2Cov(X\,,Y) \\
        \quad 3.1. \quad Var\Big(\sum_{i = 1}^nX_i\Big) = \sum_{i = 1}^n\sum_{j = 1}^nCov(X_i\,,X_j) =   
\sum_{i = 1}^nVar\big(X_i\big) + 2\sum_{(i, j): i < j}Cov(X_i\,, X_j) \\
        4. \quad if~X~and~Y~are~idependent~then~Cov(X\,,Y) = 0
        $
        <br> <br>
		<font color=blue size=6>
تذکر ۱. 
        </font> دقت کنید که برعکس مورد ۴ صادق نیست. یعنی صرفا صفر بودن کوواریانس دو متغیر تصادفی بیانگر مستقل از هم بودن آن دو نیست.
        <br>
        		<font color=blue size=6>
تذکر ۲. 
        </font> با استفاده از مورد ۴ و ۳ میتوان دید که در صورتی‌که دو متغیر تصادفی مستقل از هم باشند، واریانس جمع‌شان برابر با جمع واریانس‌هایشان می‌باشد. 
        <br> <br>
        مورد ۲ را اثبات می‌کنیم:  
        <br>
        $
        Cov(aX + b\,, cY + d) = E[(aX + b)(cY + d)] - E[aX + b]E[cY + d] = E[acXY + adX + bcY + bd] - (aE[X] + b)(cE[Y] + d) = acE[XY] + adE[X] + bcE[Y] + bd - acE[X]E[Y] - adE[X] - bcE[Y] - bd = acE[XY] - acE[X]E[Y] = ac(E[XY] - E[X]E[Y] = acCov(X\,,Y)
        $
        <br><br>
        حال به صورت عملی در R نیز نشان می‌دهیم.
        <br><br>
        	داده‌ی فروش بستنی و دمای هوا را در نظر بگیرید.	
            <br>
            فرض کنید قیمت یک بستنی قیفی آماده ۵ هزار تومان باشد. هم‌چنین فروشنده برای خرید مواد اولیه و … باید ماهانه مبلغ ۱۰۰ هزار تومان بدهند.  اگر متغیر تصادفی تعداد بستنی‌های فروش رفته در هر ماه را $X$ بنامیم سود ماهانه‌ی فروشنده برابر با $5X - 100$ خواهد بود. 
        <br>
        ستون دما در داده‌های موردنظر برحسب سلسیوس است. آن را بر حسب فارنهایت می‌نویسیم: $T_{(^{\circ}F)} = T_{(^{\circ}C)} \times 1.8\, + \,32$
	</font>
</div>

```r
icecream$income = icecream$sales * 5 - 100
icecream$temp_fahrenheit = icecream$temp * 1.8 + 32
```

<div style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
    	کوواریانس داده‌های قبلی را مجدد بدست می‌آوریم.
    </font>
</div>

```r
cov(icecream$temp, icecream$sales)
```

<div style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>

<div style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
    		<font color=blue size=5>
تمرین ۱.
        </font>سعی کنید سایر ویژگی‌ها را نیز ابتدا به صورت تئوری اثبات کنید. سپس در R شبیه‌سازی کنید. 
        <br>
        <font color=blue size=5>
تمرین ۲.
        </font>ثابت کنید که در صورتی‌ که $X$ و $Y$ دو متغیر تصادفی و $n$ عددی ثابت باشند اگر داشته باشیم $X + Y = n$ آن‌گاه $Cov(X\,,Y) = -Var(X)$
        <br>
        راهنمایی.  از ویژگی ۲ استفاده کنید.

            		<font color=blue size=5>
تمرین ۳.
        </font>اگر $X$ و $Y$ دو متغیر تصادفی با واریانس برابر باشند، ثابت کنید $X + Y$ و $X - Y$ دارای کوواریانس صفر هستند. 
    </font>

<div id="sec_corr" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
    <hr>
		<font color=red size=6>
			<div align=center>
هم‌بستگی
            </div>
		</font>
		<p></p>
		<hr>
        <br>
        	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6.5>
یاد آوری
        </font>
    <p></p>
		<hr>
بالاتر تفسیر عدد کوواریانس را گفتیم. به یاد آورید که برای تفسیر صرفا به علامت آن توجه کردیم و نه به مقدار آن. وقتی مستطیل‌های رنگی میکشیدیم نیز صرفا گفتیم مثلا اگر رنگ نمودار بیشتر مایل به سبز باشد تا قرمز، دو متغییر در یک جهت با هم تغییر می‌کنند. ولی نگفتیم اگر دو نمودار مایل به سبز داشته باشیم و یکی <i> سبزتر </i> از دیگری باشد، چه تفسیری می‌کنیم. 
	</font>
</div>

<div style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
واقعیت این است که مقدار کوواریانس تقریبا هیچ شهودی در رابطه با <i> میزان تغییر </i> دو متغیر نسبت به هم نمی‌دهد. صرفا <i> جهت </i> آن را به ما اطلاع می‌دهد. چرا؟
    </font>
</div>

<div style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
فرمول کوواریانس را به یاد آورید. به نظرتان واحد کوواریانس چیست؟ 
<br> 
<br>
$$ Cov(X, Y) = E[XY] - \mu_X\mu_Y $$
<br>
واحد کوواریانس، واحد متغیر $X$ در واحد متغیر $Y$ است. به همین دلیل اگر ما صرفا واحد‌های متغیر‌ها را تغییر بدهیم، عدد کوواریانس نیز تغییر خواهد کرد. 
<br>
مثلا نسبت قد و وزن انسان‌ها را در نظر بگیرید. فرض کنید یک بار قد را بر حسب سانتی‌متر و یک بار بر حسب متر نوشته‌ایم. در حالت اول کوواریانس صد برابر حالت دوم خواهد شد، در حالی که تغییر قد و وزن نسبت به همدیگر تفاوتی نکرده‌است. 
<br>
یا در مثال بستنی و دمای هوا. اگر دمای هوا را به سلسیوس بگیریم، میزان کوواریانس با زمانی که دما را به فارنهایت می‌گیریم فرق میکند، در حالی که میزان تغییر این دو متغیر نسبت به هم ثابت است.
<br>
</font>
</div>

<div style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
    به نظرتان برای رفع این مشکل چه کار باید بکنیم؟
</font>
</div>

<p></p>
<br/>
<div id="sec_corr_meaning" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
معنای هم‌بستگی
        </font>
		<p></p>
        <hr>
        متغیر $X$ و $Y$ را استاندارد میکنیم:
    $$X \rightarrow \frac{X - \mu_X}{\sigma_X} \quad \quad Y \rightarrow \frac{Y - \mu_Y}{\sigma_Y} $$
    <br>
    حال کوواریانس این دو متغیر استاندارد‌شده را در نظر بگیرید. (دقت کنید که میانگین صفر شده است):
    <br>
    <br> 
$$ Cov(\frac{X - \mu_X}{\sigma_X}, \frac{Y - \mu_Y}{\sigma_Y}) = 
    E[\frac{X - \mu_X}{\sigma_X} \frac{Y - \mu_Y}{\sigma_Y}] = 
\frac{1}{\sigma_X\sigma_Y}E[(X - \mu_X)(Y - \mu_Y)] = 
\frac{Cov(X, Y)}{\sigma_X\sigma_Y}$$
<br> <br>
واضح است که در کسر نهایی، واحد صورت و مخرج یکسان است. پس مشکل مقیاس که بالاتر مطرح کردیم را نداریم. 
<br>
مقدار $\frac{Cov(X,~Y)}{\sigma_X\sigma_Y}$ را هم‌بستگی یا $\rho$ می‌نامیم. 
</font>
</div>

<p></p>
<br/>
<div id="sec_corr_attr" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
خواص هم‌بستگی
        </font>
        <hr>
        <br>
    بعضی از خواص هم‌بستگی عبارتند از:
    <ul>
    <li> همانطور که قبلا گفتیم، هم‌بستگی مقیاس ندارد. به این معنی که اگر مقداری را در $X$ یا $Y$ ضرب کنیم، مقدار هم‌بستگی تغییری نخواهد کرد.</li>
    <li> هم‌بستگی دو متغیر تصادفی مستقل از هم برابر با صفر است. </li>
    <li> $ -1 \leq \rho \leq 1$</li>
    </ul>
    خاصیت آخر مهم‌ترین خاصیت هم‌بستگی است. هرچقدر مقدار هم‌بستگی به $1$ یا $-1$ نزدیک‌تر باشد، میزان رابطه‌ی خطی $X$ و $Y$ نیز (مثبت یا منفی) افزایش می‌یابد.
    همچنین داریم:
    <br> <br>
    $$ 
    \rho = +1  \iff  Y = aX + b, ~ a > 0 \\
    \rho = -1 \iff Y = ax + b, ~ a < 0
    $$
    <br> <br> 
   	</font> 
</div>

<div style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
<font color=blue size=5>
            تمرین.
            </font> 
            سعی کنید دو رابطه‌ی بالا را اثبات کنید.
            <br> <br>
            راهنمایی. عبارات $Var(\frac{X}{\sigma_X} \pm \frac{Y}{\sigma_Y})$ 
            را در نظر بگیرید.
            <br>
            <font color=blue size=5>
            <br>
تمرین عملی.
            </font> 
دیتاست `human.csv` واقع در پوشه‌ی res را بخوانید و هم‌بستگی ستون‌های مختلف آن را بدست آورید. هم‌بستگی در `R` با استفاده از دستور `cor()` محاسبه می‌شود. 
<br>
بخش اختیاری.
تلاش کنید correlation map برای ستون‌های داده رسم کنید. 
</font>
</div>

<div id="sec_corr_cause" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
هم‌بستگی و علیت
        </font>
		<p></p>
	</font>
</div>

<div style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
حال که معنای هم‌بستگی را یاد گرفتید شاید به ذهنتان برسد که اگر دو متغیر $X$ و $Y$ هم‌بستگی مثبت داشته باشند، یعنی افزایش یکی <b>موجب</b> افزایش دیگری می‌شود. 
ولی این تفکر درستی نیست.
<br> <br>
<center>
<b>هم‌بستگی به معنای علیت نیست</b>
</center>
<br>
مثلا فرض کنید $X$ متغیر تصادفی تعداد بستنی‌های فروخته شده و $Y$ تعداد حمله‌های کوسه‌ها در هر ماه باشد. اگر هم‌بستگی این دو مثبت باشد، دلیل بر اینکه نمی‌شود که فروش زیاد بستنی‌ها موجب حمله‌ی کوسه می‌شود. 
احتمالا فروش زیاد بستنی به علت گرمی هوا بوده، و در هوای گرم تعداد افراد کنار ساحل و داخل دریا بیشتر می‌شود و بدین ترتیب حمله‌ي کوسه‌ها نیز افزایش می‌یابد
<br> 
</font>
</div>
