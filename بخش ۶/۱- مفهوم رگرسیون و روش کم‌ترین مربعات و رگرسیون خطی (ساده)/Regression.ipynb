مقدمه
        </font>
		<p></p>
		<hr>
        جان که سال دیگر قصد شرکت در انتخابات ریاست جمهوری آمریکا را دارد، می‌خواهد به درون تشکیلات رقیب اصلی خود یعنی دانلد نفوذ کند. به همین منظور جان قصد خرید آپارتمانی در برج ترامپ می‌گیرد. دانلد که اصلا از جان به دلیل توافق افتضاح برجام خوشش نمی‌آید برای فروش یک آپارتمان ۶۳ متری در این برج از او قیمت ۶۳ میلیارد دلار طلب می‌کند. جان به دلیل اینکه چندسال وزیر دولت بود از قیمت تعداد زیادی آپارتمان دیگر خبر دارد به همین دلیل قیمت این آپارتمان برایش به طرز عجیبی غیرمعقول به‌نظر می‌رسد. به همین جهت از دانشگاه شریف کمک می‌خواهد تا به ریاست جمهوری برسد و به برجام بازگردد. او سوالی که از شما می‌پرسد این است که آیا می‌توان تخمین خوبی از قیمت این آپارتمان بدست آورد؟

<div id="regression_why" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
رگرسیون چرا؟
        </font>
		<p></p>
		<hr>
به مساله مطرح شده خوب فکر کنید. زمانی که با داده های آماری سر و کار داریم بسیار پیش می آید که بخواهیم رابطه یک متغیر را با متغیر های دیگر بسنجیم. در این مساله از ما خواسته شده است که قیمت آپارتمان ها را تخمین بزنیم.
        واضح است که قیمت هر آپارتمان به موارد زیادی بستگی دارد. مثلن توقع نداریم که آپارتمانی با متراژ کمتر در مقایسه با آپارتمانی با شرایط یکسان اما متراژ بیشتر، قیمت بالا تری داشته باشد. این یعنی به طور طبیعی انتظار داریم با بالا رفتن متراژ آپارتمان ها، قیمت آن ها نیز افزایش یابد. اما سوال های اصلی که در اینجا مطرح میشوند این است که آیا با 2 برابر شدن(یا
        n
        برابر شدن)
        متراژ آپارتمان ها لزوما، قیمت آن ها نیز دو برابر میشود؟
        همچنین میتوان به این سوال فکر کرد که آپارتمانی که متراژ بالا اما سن بیشتری دارد، گران تر است یا آپارتمانی با متراژ پایین اما جدید تر؟
        <br/>
تمام سوال های مطرح شده با دانستن رابطه بین قیمت و عوامل تاثیرگذار بر آن 
        پاسخ داده میشوند.
        پس کار ما برای پاسخ به سوال آپارتمان ها، پیدا کردن این روابط است.
        <br/>
        رگرسیون ابزاری است که به ما در کشف این روابط کمک میکند.
        <br/>
       سوال: آیا میتوانید مساله های دیگری مطرح کنید که برای پاسخ دهی به آن ها نیازمند رگرسیون باشیم؟
</font>
</div>

<div id="linear_regression" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
رگرسیون خطی
        </font>
		<p></p>
           <hr>
فرض کنید ما تنها متراژ هر آپارتمان را میدانیم و از روی آن میخواهیم قیمت آپارتمان را تخمین بزنیم
    .
    <br/>
        یک مدل ساده این است که فرض کنیم قیمت آپارتمان در رابطه با متراژ آن به صورت خطی تغییر میکند. 
        در این صورت میتوانیم عبارت زیر را برای این رابطه بنویسیم:
        <div dir = "ltr">
          $y_i = B_0 + B_1\;x_i$
            
  </div>
  در اینجا
  y
  بیانگر قیمت و 
  x
  بیانگر متراژ هر آپارتمان هستند.
  <br/>
  البته باید توجه داشت که این جواب ها بدون خطا نیستند و 
 عبارت بالا تنها تخمینی است از قیمت واقعی
    .
    پس به طور دقیق تر میتوان نوشت که
            <div dir = "ltr">
          $y_i = B_0 + B_1\;x_i + \epsilon _i$
            
  </div>
  که اپسیلون یک متغیر تصادفی است که خطای ما را در تخمین قیمت آپارتمان نشان میدهد. در رگرسیون خطی هدف این است که دقیق ترین 
  $B_0$
  و
  $B_1$
  را پیدا کنیم که این خطا ها را به کمترین مقدار برسانند.
  <br/>
  تصویر زیر این مطلب را به روشنی بیان میکند:
  <br/>
  <img src="regression1.png" style="float:center; width:50%"/>
  <br/>
    </font>
</div>

<div id="solve_regresion" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
حل مساله رگرسیون خطی ساده
        </font>
		<p></p>
    	<hr>
        پیش تر گفتیم که 
$\epsilon_i$
        ها متغیر های تصادفی ای از خطای ما هستند که قصد کم کردن مقدار آن ها را داریم.
        این متغیر های تصادفی را نرمال با میانگین صفر در نظر میگیریم:
        <div dir = "ltr">
            $ϵ_i∼N(0,σ^2)$
</div>
             دقت کنید که ما مقادیر
            $σ^2$
            و
            $B_0$
            و 
            $B_1$
             را نمیدانیم و سعی داریم آن ها را به گونه ای پیدا کنیم که بهترین تخمین را از 
             y
             داشته باشیم.
             <br/>
             یکی از روش های پیدا کردن این پارامتر ها روش کمترین مربعات میباشد.
        </font>
</div>

<div id="square" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
روش کمترین مربعات
        </font>
		<p></p>
<hr>
        تا اینجا تخمینمان از قیمت آپارتمان ها 
        (y)
        را به صورت زیر مدل کردیم:
        <div dir = "ltr">
            $\hat{y_i} = B_0 + B_1 \; x_i$
</div>
            همچنین میدانیم خطای این تخمین برابر است با تفاضل مقدار واقعی و مقدار به دست آمده:
            <div dir = "ltr">
            $e_i = y_i−\hat{y_i} =y_i − B_0− B_1x$
</div>
    توجه کنید که در عبارت بالا
$y_i$
    بیانگر مقدار واقعی و 
$\hat{y_i}$
    بیانگر مقدار تخمینی ما هستند.
</br>
    درنتیجه جمع مربعات خطا ها برابر است با:
    <div dir = "ltr">
       $E(B_0, B_1) = \Sigma{e^2_i} = \Sigma({y_i − B_0− B_1x})^2$
</div>
که تابعی است بر حسب 
$B_0$
و 
$B_1$
که میخواهیم آن را کمینه کنیم. برای کمینه کردن این تابع میتوانیم مشتق آن را بر حسب
$B_0$
و 
$B_1$
برابر با صفر قرار داده و 
پارامتر های مورد نظر را بیابیم:
<div dir = "ltr">
    $\frac{\partial{E}}{\partial{B_0}} = \Sigma{(-2)(y_i − B_0− B_1 x_i)} = 0$
</div>

<div dir = "ltr">
     $\frac{\partial{E}}{\partial{B_1}} = \Sigma{(-2x_i)(y_i − B_0− B_1 x_i)} = 0$
</div>
با حل معادلات بالا 
$B_0$
و
$B_1$
بهینه که آن ها را
$\hat{B_0}$
و
$\hat{B_1}$
می نامیم به دست می آیند:
<div dir = "ltr">
     $\hat{B_1} = \frac{\Sigma{(x_i - \bar{x_i})(y_i - \bar{y_i})}}{\Sigma{(x_i - \bar{x_i}) ^2}}$
</div>
<div dir = "ltr">
     $\hat{B_1} = \bar{y} - \hat{B_1}\bar{x}$
</div>
</font>
</div>
