<div id="sec_intro" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مقدمه
        </font>
		<p></p>
		<hr>
        جان که سال دیگر قصد شرکت در انتخابات ریاست جمهوری آمریکا را دارد، می‌خواهد به درون تشکیلات رقیب اصلی خود یعنی دانلد نفوذ کند. به همین منظور جان قصد خرید آپارتمانی در برج ترامپ می‌گیرد. دانلد که اصلا از جان به دلیل توافق افتضاح برجام خوشش نمی‌آید برای فروش یک آپارتمان ۶۳ متری در این برج از او قیمت ۶۳ میلیارد دلار طلب می‌کند. جان به دلیل اینکه چندسال وزیر دولت بود از قیمت تعداد زیادی آپارتمان دیگر خبر دارد به همین دلیل قیمت این آپارتمان برایش به طرز عجیبی غیرمعقول به‌نظر می‌رسد. به همین جهت از دانشگاه شریف کمک می‌خواهد تا به ریاست جمهوری برسد و به برجام بازگردد. او سوالی که از آقای دکتر شریفی می‌پرسد این است که آیا می‌توان تخمین خوبی از قیمت این آپارتمان بدست آورد؟
        <br/>
<img src="figs/trump.jpg" style="float:center; width:50%"/>
<br />
</font>

<div id="sec_problem" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
صورت مسئله
        </font>
		<p></p>
		<hr>
        تابعی به نام f داریم که ورودی‌هایی دارد مانند متغیرهای زیر:
        <div>
			<ul style="margin-right: 0;">
                <li>
                    متراژ
				</li>
                <li>
					سن ساختمان
				</li>
                <li>
                    محله
				</li>
                <li>
                    تعداد پارکینگ
				</li>
			</ul>
		</div>
        <br/>
        خروجی این تابع یک عدد است که قیمت این ساختمان را مشخص می‌کند.
        <br/>
        آیا می‌توان به کمک آمارگیری و فهمیدن قیمت تعدادی ساختمان از سازوکار f سردرآورد؟
        <br/>
        اگر بتوان تابع f را بفهمیم می‌توانیم به راحتی قیمت هر آپارتمان دیگری را با قرار دادن در این فرمول فهمید. پس این تابع همان تابعی است که وزیر اسبق امور خارجه‌ی آمریکا به دنبال أن است
        <br/>
        به تلاش برای تخمین مناسبی از تابع f تخمین رگرسیون می‌گویند. پرکاربرد ترین تحلیل رگرسیون روش کمترین مربعات است.
        <img src="figs/R.png" style="float:center; width:50%"/>
    </font>
</div>

<br />
<div id="sec_t_test" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
روش کمترین مربعات
        </font>
		<p></p>
		<hr>
        فرض کنید تابعی که حدس میزنیم بصورت $g(x_1,x_2,x_3,....,x_n)$ باشد که هرکدام از $x_i$ ها یکی از ویژگی ها گفته شده هستند. می‌خواهیم تابعی تعریف کنیم که با مینیموم کردن آن به تخمین خوبی در این زمینه برسیم
<br/>
        طبیعتا می‌توان این تابع را به روش های مختلفی تعیین کرد برای مثال یک g پیدا کنیم که مقدار $\max_{}^{} (f(x_1,x_2,....x_n) - g(x_1,x_2,....x_n))$ کمترین باشد.
<br />
        اما روش کمترین مربعات پیشنهاد می‌دهد که تابع g را طوری تعیین کنیم که مقدار $\sum_{m}^{} \frac{1}{m}(f(x_1,x_2,....x_n) - g(x_1,x_2,....x_n))^2$ کمترین مقدار شود. مزیت این روش نسبت به روش قبلی به طور خاص این است که اگر یک مقدار پرت در داده ها داشته باشیم و g غیر از این داده بقیه‌ی داده‌ها را به خوبی تخمین می‌زند روش اول به هیچ عنوان خوب کار نمی‌کند.
        <br/>
        مثالی برای خروجی این روش را در زیر می‌بینید.
        <img src="figs/Linear_least_squares2.png" style="float:center; width:30%"/>
<br />
