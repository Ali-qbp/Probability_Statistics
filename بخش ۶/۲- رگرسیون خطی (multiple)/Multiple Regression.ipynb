        به طور کلی توجه داشته باشید که هدف از بررسی رگرسیون در این درس، بیشتر آشنایی با آن و مفاهیم آن می‌باشد و بیشتر سعی بر آن است تا با ارائه مثال‌هایی شما را با مفاهیم آن آشنا کنیم.
        <br>

<div id="sec_intro" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
رگرسیون چندگانه : دیدگاه تئوری
        </font>
		<p></p>
		<hr>
        در رگرسیون ساده مدل ما تنها یک predictor داشت اما در این بخش به مدل‌هایی خواهیم پرداخت که بیشتر از یکی predictor دارند. (در واقعیت نیز اغلب به همین شکل می‌باشد.)
        <br>
        همانند قبل مدل ما شامل $Y$ ، $\textbf{X}$ و $\boldsymbol\beta$ می‌باشد که $\textbf{X}$ و $\boldsymbol\beta$ بردار می‌باشند و مدل ما $Y$ را بر اساس تابعی از $\textbf{X}$ و $\boldsymbol\beta$ می‌سازد.
        <br><br>
        به طور کلی برای هر داده :
        <br>
        $$Y_i =  \beta_0 X_{i0} + \beta_1 X_{i1} + \ldots + \beta_{p} X_{ip} + \epsilon_{i}\\= \sum_{k=0}^p X_{ik} \beta_k
        +\epsilon_{i}$$و برای کل داده:
        <br>
        $$\textbf{Y} = \textbf{X}\boldsymbol\beta + \boldsymbol\epsilon$$ به صورتی که:
        <br>
        \[
        \textbf{Y} =
        \begin{bmatrix}
        y_{1} \\
        y_{2} \\
        \vdots \\
        y_{n} 
        \end{bmatrix}
        ~
        ~
        ,
        ~
        ~
        \textbf{X} =
        \begin{bmatrix}
        1 & x_{11} & x_{12} & \dots  & x_{1p} \\
        1 & x_{21} & x_{22} & \dots  & x_{2p} \\
        \vdots & \vdots & \vdots & \ddots & \vdots \\
        1 & x_{n1} & x_{n2} & \dots  & x_{np}
        \end{bmatrix}
        \]و
        <br>
        \[
        \boldsymbol\beta =
        \begin{bmatrix}
        \beta_{0} \\
        \beta_{1} \\
        \vdots \\
        \beta_{p} 
        \end{bmatrix}
        ~
        ~
        ,
        ~
        ~
        \boldsymbol\epsilon =
        \begin{bmatrix}
        \epsilon_{1} \\
        \epsilon_{2} \\
        \vdots \\
        \epsilon_{n} 
        \end{bmatrix}
        \]<br><br>
        همانند قبل می‌توان با استفاده از روش کمترین مربعات، تخمین $\boldsymbol\beta$ را پیدا کرد.
        <br><br>
        حال مدل رگرسیون ما به صورت زیر خواهد بود. 
        <br>
        $$ \hat{y} = X\hat{\beta} $$ <br> 
        توجه : منظور از رگرسیون خطی ، خطی بودن به نسبت ضرایب $\beta$ می‌باشد. به عنوان مثال :
        <br>
        $$   y = \beta_0+\beta_1 x +\beta_2 x^2+\epsilon$$<br>
        یک مدل رگرسیون خطی می‌باشد زیرا به نسبت $\beta_0, \beta_1, \beta_2$ خطی است.

		<font color=#FF7500 size=6>
رگرسیون چندگانه : دیدگاه عملی
        </font>
		<p></p>
		<hr>
        فرض کنید اطلاعات میزان فروش یک کالا را بر اساس میزان هزینه‌های متفاوتِ تبلیغات در رادیو ، روزنامه و تلویزیون داریم و ما به عنوان یک مشاور فروش می‌خواهیم برنامه‌ای برای تبلیغات سال بعد ارائه کنیم.
        <br><br>
        (در این‌جا از داده Advertising که در کنار این نوت‌بوک نیز قرار داده شده استفاده خواهیم کرد.)
        <br><br>
        با رگرسیون خطی ساده آشنا شده‌اید. اگر بخواهیم در این‌جا به بررسی رابطه میزان فروش با چند متغیر بپردازیم ساده ترین راهی که به نظر می‌رسد پیاده کردن سه مدل رگرسیون خطی ساده برای هر یک از متغیر ها می‌باشد.
        <br>
        آیا این کار درست می‌باشد ؟
        <br>
        اولین مشکلی که می‌توان به آن اشاره کرد این است که نمی‌توان مشخص کرد که یک مقدار فروش بر اساس چه مقادیری از هر یک از سه متغیر به طور همزمان می‌باشد.
        <br>
        مشکل دیگر این است که با ساختن سه مدل رگرسیونی، در هر یک، دو متغیر دیگر درنظر گرفته نمی‌شود. (و در صورت وجود همبستگی بین متغیرها آن‌ها به مشکل برخواهیم خورد.)
        <br><br>
               در اصل روش درست ساخت یک مدل رگرسیون چندگانه می‌باشد تا مدل خطی خود را برای بیش از چند متغیر گسترش دهیم.
