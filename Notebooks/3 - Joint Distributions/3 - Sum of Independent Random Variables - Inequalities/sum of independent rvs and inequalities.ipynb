        <font face="B Nazanin" size=4>
                در مطالعه آمار و احتمالات، معمولا محاسبه توزیع X + Y از روی توزیع‎های X و Y زمانی که X و Y مستقل باشند، از اهمیت بالایی برخوردار است.
        فرض کنید X و Y دو متغیر تصادفی مستقل با توزیع‎های پیوسته $ f_X $ و $ f_Y $ باشند.
            تابع توزیع  X + Y را می‎توان به شکل زیر محاسبه کرد.
            ابتدا تابع توزیع تجمعی 
            $ F_{X+Y} $
            را محاسبه می‎کنیم.
       
$$
\begin{align*}
F_{X+Y}(a) &= P(X + Y \le a) \\ &= \iint\limits_{x + y \le a} f_X(x)f_Y(y) \, \mathrm{d}x \mathrm{d}y \\
&= \int_{- \infty}^{\infty} \int_{- \infty}^{a – y} f_X(x)f_Y(y) \, \mathrm{d}x \mathrm{d}y \\
&= \int_{-\infty}^{\infty} f_Y(y) \int_{- \infty}^{\infty} f_X(x) \mathrm{d}x \, \mathrm{d}y \\
&= \int_{-\infty}^{\infty} F_X(a - y)f_Y(y) \,\mathrm{d}y
\end{align*}
$$
<p></p>
            حالا تابع توزیع
            $f_{X+Y}$
            را محاسبه می‎کنیم
<p></p>
$$
    \begin{align*}
    f_{X+Y} &= \frac{d}{da} \int_{-\infty}^{\infty} F_X(a-y)f_Y(y) dy
    \\ &= \int_{-\infty}^{\infty} \frac{d}{da} F_X(a-y)f_Y(y)\, dy
    \\ &= \int_{-\infty}^{\infty} f_X(a-y)f_Y(y)\, dy
    \\ &= f_X(a)*f_Y(a)
    \end{align*}
$$
<p></p>
            همانطور که محاسبات نشان می‎دهد، توزیع حاصل همان convolution دو توزیع 
            $f_X(x)$
            و
            $f_Y(y)$
            است.
            از خواص convolution می‎توان به جابجایی و شرکت پذیری اشاره کرد.

            به طور مثال فرض کنید که دو متغیر تصادفی مستقل با توزیع uniform(0, 1) داشته باشیم.
            حالا می‎خواهیم توزیع مجموع این دو متغیر تصادفی را حساب کنیم. روش اول استفاده از روش شهودی کانولوشن است. به این ترتیب که نمودار توزیع یک متغیر را نسبت به محور y قرینه کنیم و بعد نمودار آن را از روی نمودار توزیع متغیر دیگر عبور دهیم.
            اما اینجا می‎خواهیم از فرمول‎های بدست آمده استفاده کنیم.
            طبق توزیع uniform داریم:
            
            
$$
f_X(a) = f_Y(a) =
\left\{
	\begin{array}{ll}
		1  & 0 \lt a \lt 1 \\
		0 & \mbox{otherwise}
	\end{array}
\right.
$$
<p></p>
            با جایگذاری این توزیع‎ها در رابطه کانولوشن داریم:     
$$
f_{X+Y}(a) = \int_{0}^{1} f_X(a-y)\,\mathrm{d}y
$$
<p></p>
            و
$$
0 \le a \le 1
$$
<p></p>
            بنابراین
$$
f_{X+Y}(a) = \int_0^{a} \mathrm{d}y
$$
<p></p>
            با حالت بندی روی a به توزیع نهایی می‎رسیم:
            
$$
f_{X+Y} = \left\{
            \begin{array}{ll}
            a & 0 \le a \le 1 \\
            2 - a & 1 \lt a \le 2 \\
            0 & \mbox{otherwise}
            \end{array}
            \right.
$$
<p></p>
            با اجرای کد پایین می‎توان درستی نتیجه گیری را مشاهده کرد. به دلیل شکل خاص این توزیع به آن توزیع مثلثی می‎گویند. بنابراین در این قسمت با این موضوع آشنا شدیم که اگر دو متغیر تصادفی مستقل با توزیع یکنواخت داشته باشیم، مجموع این دو متغیر از یک توزیع مثلثی خواهد بود که شکل آن در هیستوگرام زیر مشاهده خواهد شد.

        <font face="B Nazanin" size=4>
            در این قسمت می‎خواهیم مجموع چند متغیر تصادفی را بررسی کنیم. فرض کنید به جای دو متغیر تصادفی n متغیر تصادفی با توزیع یکنواخت داشتیم. توزیع تجمعی مجموع این n متغیر تصادفی را به شکل زیر تعریف می‎کنیم:
            
$$
F_n(x) = P\{X_1 + X_2 + ... + X_n \le x \}
$$
<p></p>           
با وجود اینکه بدست آوردن یک رابطه صریح در حالت کلی برای 
$ F_n(x) $
دشوار به نظر می‎رسد، اما زمانی که 
$ x \le 1 $
رابطه‎ای بدست می‎آید که ظاهر مناسبی دارد. به طور دقیقتر
            
$$
F_n(x) = \frac{x^n}{n!}\,\,\, (0 \le x \le 1)
$$
<p></p>
            در ادامه با استقرا رابطه بالا را ثابت می‎کنیم. 
<p></p>
            پایه استقرا برای $n = 1$ واضح است. در گام استقرا داریم:

$$
F_{n-1}(x) = \frac{x^{n-1}}{(n-1)!} \,\,\, (0 \le x \le 1)
$$
<p></p>
            از طرفی،

$$
\sum\limits_{i=1}^{n} X_i = \sum\limits_{i=1}^{n-1} + X_n \\
$$
<p></p>
            حالا از رابطه‎ای که برای $F_{X+Y}$ در قسمت قبل بدست آمد استفاده می‎کنیم:

$$
\begin{align*}
F_n(x) &= \int_{0}^{1} F_{n-1}(x - y)f_{X_n}(y) \, dy \\
&= \frac{1}{(n-1)!} \int_{0}^{1} (x-y)^{n-1}\, dy \\
\end{align*}
$$
<p></p>
            با توجه به اینکه تمامی $X_i$ها مثبت هستند و همچنین
            $ 0 \le x \le 1$
            می‎توانیم بنویسیم:

$$
\begin{align*}
F_n(x) &= \int_{0}^{x} (x-y)^{n-1}\,dy \\
&= \frac{x^n}{n!}
\end{align*}
$$
<p></p>
            و به این ترتیب اثبات کامل می‎شود.
