<div id="sec_def" style="direction:rtl;line-height:300%;">

		<p></p>
		<hr>
       «دانشجو» در یک استارتاپ مسیریابی کار می‌کند. مدیرش از او خواسته‌است تا روشی پیدا کند که بتوان مقدار پرترافیک بودن هر خیابان را مشخص کرد. دانشجو با دیدن میلاد و سینا به خود می‌آید و تصمیم می‌گیرد از داده آنها برای کارش استفاده کند. او معیار پرترافیک بودن هر خیابان را 
        <b> متوسط تعداد ماشین‌های عبوری در یک مدت زمان معین</b> از آن خیابان در نظر می‌گیرد (مثلا یک دقیقه). 
        <br>
         در زیر ترافیک خیابانی که میلاد از آن فیلم گرفته بود را طبق معیار دانشجو حساب میکنیم:(هر سطر در داده‌ها نشان‌دهنده زمان عبور یک ماشین است)

<p></p>
<br />
<div id="sec_def" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<p></p>
        اما دانشجو که آدم محتاطی است، تصمیم می‌گیرد این عدد را به مدیرش گزارش نکند و به‌جای استفاده از یک عدد به عنوان ترافیک خیابان‌ها از یک مدل احتمالاتی استفاده کند. در واقع سعی می‌کند به جای اتکای خالص بر داده‌ها کمی هم  شهودش را به کار گیرد. پس تلاش می‌کند به تعداد ماشین‌های عبوری در مدت زمان معین از یک خیابان یک احتمال نسبت دهد.<br>
         برای این‌کار فرض‌های زیر را روی عبور ماشین‌ها در یک بازه زمانی معین(مثلا یک دقیقه) در نظر می‌گیرد:
        <br>
        <li>
            تعداد ماشین‌های عبوری در بازه‌های زمانی <b>ناهم‌پوش </b>از بازه معین شده، از یکدیگر <b>مستقل</b> باشند. برای مثال دانستن اینکه در یک ثانیه قبل چند ماشین از این خیابان عبور کرده اند دانش ما درباره ثانیه بعد را تحت تاثیر قرار ندهد. (البته این فرض به طور واضح درباره مثال ماشین‌ها اشتباه است. مثلا فرض کنید ماشینها پشت چراغ قرمز بودند و پس از سبز شدن چراغ اولین ماشین را مشاهده میکنیم. بنابراین پس از مشاهده پی به سبز شدن چراغ و در نتیجه ورود ماشینهای بیشتر به خیابان خواهیم برد. به هرحال گاهی چنین فرضهایی در نتیجه عملی کارساز است)         </li>
        <br>
        <li>
            احتمال گذر <b>دقیقا یک</b> ماشین در یک بازه زمانی <b> کوتاه </b> به طول $h = \frac{1}{n}$ دقیقه، به طور تقریبی <b>متناسب با طول بازه</b> باشد. برای سادگی فرض میکنیم ضریب این تناسب یک مقدار ثابت $\lambda > 0$ است. بنابراین احتمال گذر یک ماشین در بازه کوتاه $h$ تقریبا برابر $\lambda h = \frac{\lambda}{n}$ است.
            <br>
            به نظر شما این فرض منطقی است؟
        </li>
        <br>
        <li>
            احتمال عبور <b>دو یا بیش‌تر</b> ماشین در یک بازه زمانی <b>کوتاه</b> (مثلا $h$) <b>صفر</b> باشد. در واقع بتوان بازه‌ها را آنقدر کوچک کرد که در هر بازه تنها حداکثر یک رخداد روی دهد. این فرض درباره حرکت ماشین‌ها تا حدی منطقی است. در واقع این‌که دو یا چند ماشین دقیقا در یک لحظه از جلوی دوربین(به عنوان یک نقطه) عبور کنند نامحتمل است.
        </li>
	</font>
</div>


<p></p>
<br />
<div id="sec_pois" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
           فرایند پوآسون تقریبی و توزیع‌های پوآسون       </font>
		<p></p>
		<hr>
        اگر $X$ را یک متغیر تصادفی در نظر بگیریم که تعداد وقوع یک رخداد را در یک بازه پیوسته (زمانی و یا مکانی)  نشان می‌دهد، می‌گوییم $X$ از یک فرایند پوآسون تقریبی پیروی می‌کند اگر سه فرضی که در انتهای قسمت قبل گفته شد برقرار باشد. ($X$ در مثال بالا تعداد ماشین‌های عبوری از جلوی دوربین در یک بازه زمانی پیوسته است.)
        <br>
        بازه‌ی معینی که متغیر $X$ روی آن تعریف شده را در نظر بگیرید. (مثلا یک دقیقه) آن را به $n$ قسمت تقسیم می‌کنیم طوری که بازه‌های به طول $h = \frac{1}{n}$ دقیقه در فرض‌های دو و سه صدق کنند.
                <img src="figure/poisson.jpg" width="500">
        <br>
        وقوع یک رخداد در هر کدام از این بازه‌ها با توجه به فرض دو و سه از یک توزیع برنولی با پارامتر $\frac{\lambda}{n}$ می‌آید.(در هر بازه کوچک احتمال وقوع یک رخداد برابر $\frac{\lambda}{n}$ و احتمال وقوع بیش از یک رخداد برابر صفر است). بنابراین پیشامد وقوع $k$ رخداد در بازه یک دقیقه برابر پیشامد وقوع یک رخداد در $k$ بازه کوچک
         از $n$ بازه است. از طرفی طبق فرق مستقل بودن تعداد رخدادها در این بازه‌های کوچک(فرض اول) داریم:
        <br>
        <center>
        $P(X=k) = P((k) \text{ success and }(n-k) \text{ failure}) = \binom{n}{k} (\frac{\lambda}{n})^{k} (1 - \frac{\lambda}{n})^{n-k}$
        </center>
	</font>
</div>

<p></p>
<br />
<div id="sec_def" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
        <ul>
            حال اگر اندازه بازه‌ها را به طور دلخواه کوچک کنیم ($n\rightarrow \infty$) خواهیم داشت:<br>
            <center>
            $lim_{n\to\infty} \binom{n}{k} (\frac{\lambda}{n})^{k} (1 - \frac{\lambda}{n})^{n-k} = lim_{n\to\infty} \frac{n!}{k!(n-k)!} (\frac{\lambda^k}{n^k}) (1 - \frac{\lambda}{n})^{n}(1 - \frac{\lambda}{n})^{-k} = lim_{n\to\infty}\frac{\lambda^k}{k!} \frac{n(n-1)(n-2)\dots(n-k+1)}{n\cdot n\cdot n \dots n} (1 - \frac{\lambda}{n})^{n}(1 - \frac{\lambda}{n})^{-k}$
            </center>
        </ul>
	</font>
</div>

<p></p>
<br />
<div id="sec_def" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
        از ریاضیات دبیرستان می‌دانیم $\lim_{n\to\infty} (1-\frac{\lambda}{n})^n = e^{-\lambda}$ از طرفی با توجه به ثابت بودن $k$ خواهیم داشت:<br>
        <center>$\lim_{n\to\infty} (1-\frac{\lambda}{n})^{-k} = 1^{-k} = 1$</center><br>
        <center>$\lim_{n\to\infty} \frac{n(n-1)(n-2)\dots(n-k+1)}{n\cdot n\cdot n \dots n} = \lim_{n\to\infty}1(1-\frac{1}{n})(1-\frac{2}{n})\dots(1-\frac{k-1}{n}) = 1$</center><br>
	</font>
</div>

<p></p>
<br />
<div id="sec_def" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
        با توجه به روابط بالا دانشجو سرانجام به فرمول زیر برای محاسبه احتمال تعداد ماشین‌های عبوری از یک خیابان در مدت یک دقیقه رسید:<br>
        <center>
            $P(X=k) = e^{-\lambda}\frac{\lambda^k}{k!}\quad k=0,1,2,\dots$
            </center>
        <br>
        تمرین: ثابت کنید فرمول به دست‌آمده برای توزیع پوآسون ویژگی‌های یک p.m.f را دارد( مثبت بودن و جمع برابر یک).
        <br>
                اما هنوز یک مشکل وجود دارد. پارامتر $\lambda$ که در فرض دوم به عنوان یک ضریب ثابت احتمال بازه‌های کوچک بود هنوز مجهول است. در واقع این $\lambda$ است که توزیع تعداد ماشین‌های عبوری برای خیابان‌های مختلف را متمایز می‌کند. دانشجو تصمیم می‌گیرد مقدار مناسب را برای خیابان میلاد حساب کند. او قبلا متوسط تعداد ماشین‌های عبوری در یک دقیقه را حساب کرده بود. تلاش می‌کند تا این متوسط را از روی توزیع احتمال به‌دست بیاورد.<br>
        تمرین: ثابت کنید امیدریاضی (متوسط) متغیر تصادفی $X$ برابر $\lambda$است. سعی کنید واریانس این توزیع را هم محاسبه کنید.<br>
 پس معنی پارامتر $\lambda$ در توزیع پوآسون، متوسط تعداد رخدادها در مدت زمان معین‌شده (مثلا یک دقیقه در مثال خیابان میلاد) است.<br>
        در زیر نمودار توزیع پوآسون به‌ازای $\lambda$ های مختلف رسم شده‌است. محور افقی نشان‌دهنده مقادیر مختلف متغیر تصادفی $X$ است و محور عمودی احتمال هر کدام از این مقادیر را نشان می‌دهد. هر کدام از رنگ‌ها یک توزیع پوآسون خاص($\lambda$ خاص) را نمایش می‌دهد. <br>
        سوال: به نظر شما چرا با افزایش مقدار $\lambda$ توزیع متقارن‌تر نمایش داده می‌شود؟ به طور کلی آیا توزیع پوآسون می‌تواند یک توزیع متقارن باشد؟
	</font>
</div>

```r
color_set <- c('steelblue', 'firebrick', 'forestgreen')
lambdas <- c(1,5,10)
x <- seq(1, 20)
plot(dpois(x=x, lambda = lambdas[1]), col = color_set[1], xlab = "k", ylab = "P(X=k)")
points(dpois(x=x, lambda = lambdas[2]), col = color_set[2])
points(dpois(x=x, lambda = lambdas[3]), col = color_set[3])
legend('topright', legend=lambdas, title = "Lambda", col = color_set, pch = c(1, 1, 1), lwd = c(NA, NA, NA))
```

	<font face="XB Zar" size=5>
 در مثال ماشین‌های میلاد متوسط تعداد ماشینهای عبوری در یک دقیقه برابر 35.1 محاسبه شد.<br> در زیر توزیع پوآسون با مقدار $\lambda=35.1$ رسم شده است (نقاط قرمز). علاوه بر این با استفاده از داده‌های میلاد، تعداد ماشین‌های عبوری در دقایق مختلف (ده دقیقه) به دست آمده و نمودار مستطیلی احتمال این داده‌ها رسم شده است. (مستطیلهای آبی)

```r
# Function for plotting histogram of data and Poisson density
plot_hist_pois <- function(Ks, lambda) {
bins <- (max(Ks) - min(Ks)) / 2
rng <- c((min(Ks) - 1): (max(Ks) + 1))
probs <- dpois(x=rng, lambda = lambda)
hist(Ks, bins, probability = T, ylab = "P(X=k)", xlab = "k", col = color_set[1], main = "")
points(rng, probs, col = color_set[2], pch=19)
legend("topright", legend=c("Observed data", "Poisson dist"), lty = c(1, NA), 
       pch = c(NA, 19), col = color_set)
}
# Number of passing cars per minute
Ks <- sapply(c(0:9), function(x) sum(cars$minute==x))
lambda <- 35.1
plot_hist_pois(Ks, lambda)
```

<div id="simple_sample_question" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
        <ul style="margin-right: 0;">
               دانشجو به این فکر می‌کند که تا به حال تمام مدلش را در زمان یک دقیقه در نظر گرفته بود. در واقع در  فرایند رسیدن به توزیع پوآسون واحد زمان یک دقیقه بود. دانشجو اما حوصله دوباره حساب کردن مقدار $\lambda$ را برای مقادیر مختلف زمان (مثلا در 30 ثانیه) ندارد و می‌خواهد از همین عدد به دست آمده در یک دقیقه ($\lambda=35.1$) استفاده کند. با توجه به اینکه $\lambda$ میانگین تعداد ماشین‌ها در این زمان است به نظرش میرسد که برای 30 ثانیه احتمالا این میانگین نصف شود. (سعی کنید این نتیجه‌گیری را صرفا با استفاده از فرایند اثبات توزیع پوآسون هم ثابت کنید)
            <br>
            در زیر با استفاده از داده‌های میلاد تعداد ماشین‌های گذرنده در بازه‌های 30 ثانیه محاسبه و نمودار بالا  یکبار برای مقدار $\lambda=35.1$ و یکبار برای $\lambda=\frac{35.1}{2}$ رسم شده است. 
        </ul>
	</font>
</div>

```r
# Number of passing cars per 30 seconds
new_Ks <- unlist(sapply(c(0:9), function(x) list(sum(cars$minute==x & cars$second < 30), 
                                sum(cars$minute==x & cars$second >= 30))))
plot_hist_pois(new_Ks, lambda)
plot_hist_pois(new_Ks, lambda / 2)
```

<div id="sec_pois_eg" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
           یک مثال       </font>
		<p></p>
		<hr>
        فرض کنید $X$ نشان‌دهنده تعداد غلطهای املایی در یک صفحه چاپی با میانگین 5 غلط باشد. احتمال این‌که در سه صفحه چاپی تصادفی مجموعا بیش از 10 غلط املایی وجود داشته باشد چقدر است؟
        <br>
        پاسخ: اولا در نظر بگیرید که $X$ می‌تواند از توزیع پوآسون پیروی کند(تعداد رخدادها(غلط املایی) در یک بازه مکانی(صفحه چاپی). سعی کنید سه فرض فرایند تقریبی پوآسون را با این مساله چک کنید)
        <br>
        اما $X$ نشان‌دهنده تعداد غلط‌های املایی در یک صفحه است و نه سه صفحه. متغیر تصادفی $Y$ را تعداد غلطهای املایی در سه صفحه تصادفی تعریف می‌کنیم. با توجه به قسمت قبل می‌دانیم $Y$ از یک توزیع پوآسون می‌آید طوری که:<br>
        <center>
            $\lambda_{Y} = 3\lambda{x} = 3 \times 5 = 15$
        </center> 
        <br>
         <center>
            $P(Y>10) = 1 - P(Y\leq10) = 1 - \sum_{k=0}^{10} P(Y=k) =  1 - \sum_{k=0}^{10} e^{-15}\frac{15^k}{k!}$
        </center> 
        <br>
        در زبان R با تکه کد زیر می‌توان حاصل این احتمال را حساب کرد.
	</font>
</div>

```r
# using poisson dist function
1 - ppois(10, lambda=15, lower.tail = T)
# using formula
pois_prob <- function(k, lambda) {return(exp(-lambda) * (lambda**k) / factorial(k) )}
1 - sum(sapply(c(0:10), function(x) pois_prob(x, lambda = 15)))
```

		<font color=#FF7500 size=6>
           توزیع‌های نمایی       </font>
		<p></p>
		<hr>
        یک معیار دیگر برای در نظر گرفتن ترافیک یک خیابان، فاصله زمانی عبور دو ماشین متوالی در آن است (فاصله زمانی کمتر به معنی ترافیک بیشتر). این معیار برخلاف معیار قبلی که تعداد ماشین‌ها ها را حساب می‌کرد یک معیار پیوسته است. سعی می‌کنیم توزیع احتمال این فاصله زمانی را به دست بیاوریم. <br>
         فرایند تقریبی پوآسون که در دو قسمت قبل معرفی شد را در نظر بگیرید. ($X$ متغیر تصادفی تعداد ماشین‌های عبوری در بازه زمانی این بار ده ثانیه). فرض کنید متوسط تعداد ماشین‌های عبوری در این زمان برابر 6 باشد. شکل زیر یک نمونه از ماشینهای عبوری در یک بازه زمانی دلخواه ده‌ثانیه (به عنوان واحد زمان) است. اختلاف زمانی دو ماشین سوم و چهارم در این بازه به عنوان مثال مشخص شده است. حال اگر بازه جدیدی که بلافاصله پس از دیدن ماشین سوم شروع میشود و تا آخر ادامه میابد را در نظر بگیریم، اختلاف زمانی ماشین سوم و چهارم روی بازه زمانی اولیه برابر زمان انتظار تا مشاهده اولین ماشین در این بازه جدید خواهد بود. (هر دو متغیر تصادفی را با $T$ نشان می‌دهیم.)
        <br>
                            <img src="figure/exponential.jpg" width="500">
       <br>
       احتمال مشاهده اولین ماشین در کمتر از زمان $t$ (احتمال کمتر از $t$ بودن اختلاف زمانی دو ماشین متوالی) برابر:
        <br>
        <center>
            $P(T \leq t) = 1 - P(T > t)$
        </center> 
        <br>
        اما $P(T > t)$ یعنی در بازه زمانی $t$ واحد زمان هیچ ماشینی دیده نشود. ار طرفی $X$ نشان‌دهنده تعداد ماشین ها در بازه ده ثانیه(واحد زمان) و از توزیع پوآسون است. متغیر $Y$ را تعداد ماشین‌های عبوری در بازه زمانی $t$ واحد تعریف میکنیم. مطابق آنچه در بخش قبل (یک مثال) دیدیم، اگر $\lambda$ پارامتر مربوط به توزیع پوآسون اولیه باشد، $\lambda t$ پارامتر مربوط به $Y$ خواهد بود.
                <center>
            $F_{T}(t) = P(T \leq t) = 1 - P(T > t) = 1 -  P(Y = 0) = 1 - e^{-\lambda t} \frac{(\lambda t)^0}{0!} = 1 - e^{-\lambda t}$
        </center> 
        <br>
                <center>
            $f_T(t) = F^{'}_{T}(t) = (1 - e^{-\lambda t})^{'} = \lambda e^{-\lambda t} \quad 0\leq t<\infty$
        </center> 
        <br>
        $\lambda$ میانگین تعداد ماشین‌ها در واحد زمان(ده ثانیه) است. می‌توان گفت میانگین زمان انتظار برای دیدن اولین ماشین (زمان بین دو ماشین متوالی) برابر $\theta = \frac{1}{\lambda}$ واحد زمان است. (مثلا در مثال بالا که میانگین تعداد ماشین‌ها در ده ثانیه برابر 6 بود، میانگین زمان انتظار برابر $\frac{1}{6}$ واحد زمان و در نتیجه $\frac{10}{6}\text{ sec}$  خواهد بود.) <br>تابع چگالی احتمال را می‌توان به صورت $f_T(t) = \frac{1}{\theta} e^{\frac{-t}{\theta}} \quad \theta > 0$ نوشت.
        <br>
        تمرین: ثابت کنید $f_T(t)$ معرفی شده در بالا، ویژگی های یک p.d.f را دارد. (بزرگتر از صفر و جمع برابر یک)
        <br>
        تمرین: ثابت کنید میانگین توزیع نمایی برابر $\theta$ است. واریانس این توزیع را هم به دست بیاورید.
        <br>
        در زیر نمودار توزیع نمایی به‌ازای $\lambda = \frac{1}{\theta}$ های مختلف رسم شده‌است. محور افقی نشان‌دهنده مقادیر مختلف متغیر تصادفی $T$ است و محور عمودی مقدار چگالی احتمال را در هر کدام از این مقادیر نشان می‌دهد. هر کدام از رنگ‌ها یک توزیع نمایی خاص($\lambda$ خاص) را نمایش می‌دهد. <br>
	</font>
</div>

```r
lambdas <- c(2, 1, 0.5)
x <- seq(0, 5, 0.01)
plot(x, dexp(x=x, rate = lambdas[1]), col = color_set[1], xlab = "t", ylab = "pdf", type="l")
lines(x, dexp(x=x, rate = lambdas[2]), col = color_set[2])
lines(x, dexp(x=x, rate = lambdas[3]), col = color_set[3])
legend("topright", legend=lambdas, title = "Lambdas", col = color_set, pch = c(NA, NA, NA), lwd = c(1,1,1))
```

           یک مثال       </font>
		<p></p>
		<hr>
         مشتریان بانکی که تنها یک صندوقدار دارد، طبق یک فرایند تقریبی پوآسون با میانگین 30 مشتری در ساعت وارد بانک می‌شوند. با فرض اینکه صندوقدار به سرعت کار مشتری‌ها را راه می‌اندازد، احتمال اینکه بیش از سه دقیقه منتظر بماند تا مشتری جدید وارد شود چقدر است؟
        <br>
        $X$ تعداد مشتری‌ها در یک ساعت، از توزیع پوآسون با $\lambda = 30$ می‌آید. بنابراین زمان انتظار برای ورود مشتری جدید به طور متوسط $\frac{1}{30}\text{ hour} = 2\text{ min}$ خواهد بود. حال اگر $W$ نشاندهنده زمان انتظار برای دیدن مشتری جدید باشد، (می‌توان گفت) از توزیع نمایی می‌آید با پارامتر $\theta = 2\text{ min or }\frac{1}{30}\text{ hour}$.
        <br>
        <center>
            $P(W > 3) = \int_{3}^{\infty} f_W(t) dt = \int_{3}^{\infty} \frac{1}{2} e^{\frac{-t}{2}} dt = \frac{1}{2} \lim_{a\to \infty} \int_{3}^{a}  e^{\frac{-t}{2}} dt = \frac{1}{2} \lim_{a\to \infty} \bigg[{\frac{e^{\frac{-t}{2}}}{\frac{-1}{2}}}\bigg]_{3}^{a} = \\
            e^{\frac{-3}{2}} = 0.223$
        </center> 
        <br>
        تمرین: ثابت کنید اگر $X$ متغیر تصادفی از توزیع نمایی (با پارامتر $\theta$) باشد، $P(X > k) = e^{\frac{-k}{\theta}}$
        <br>
        در زبان R با تکه کد زیر می‌توان حاصل این احتمال را حساب کرد. (در این دستور به جای پارامتر $\theta$ از پارامتر $\lambda$ استفاده می‌شود و چون این پارامتر روی بازه یک ساعت تعریف شده است، به جای سه دقیقه از $\frac{3}{60}$ ساعت استفاده می‌کنیم.)
	</font>

           بی‌حافظه بودن      </font>
		<p></p>
		<hr>
        مقدار مسافتی (برحسب کیلومتر) که یک ماشین خاص می‌تواند طی کند تا باتری‌اش تمام شود از یک توزیع نمایی با میانگین 10000 کیلومتر می‌آید. (در اینجا بازه مکانی است). دارنده ماشین باید یک سفر 5000 کیلومتری داشته باشد، احتمال اینکه در طی سفر باتری‌اش تمام نشود چقدر خواهد بود؟
        <br>
پاسخ: به نظر می‌رسد باید اطلاعات بیشتری درباره مساله وجود داشته باشد، در واقع باید بدانیم این ماشین با باتری‌اش تا به حال چند کیلومتر سفر داشته است. با توجه به اینکه توزیع نمایی است، بیایید دقیقا مساله را با فرایند تقریبی پوآسون مدل کنیم. در اینجا تمام شدن باتری به عنوان یک رخداد مطرح است (فرض کنید بلافاصله بعد از تمام شدن باتری فورا تعمیر می‌شود و یا باتری دقیقا مشابه جایگزین می‌شود). $\theta = 10000 \Rightarrow \lambda = \frac{1}{10000}$ به این معنی که در یک کیلومتر میانگین تمام شدن باتری برابر $\frac{1}{10000}$ و به عبارتی در هر بازه مکانی 10000 کیلومتر میانگین تعداد این رخداد برابر 1 است. (توزیع پوآسون) حال سوال مانند این است که در یک بازه 5000 کیلومتر احتمال وقوع 0 رخداد(تمام شدن باتری) چقدر خواهد بود؟ <br>
         $X$ متغیر تصادفی تعداد رخداد ها در بازه 5000 کیلومتر است. بنابراین $X$ از توزیع پوآسون با میانگین $\lambda = 1\times \frac{5000}{10000} = \frac{1}{2}$ می‌آید، داریم:<br>
        <center>
            $P(X=0) = e^{\frac{-1}{2}} \frac{(\frac{-1}{2})^0}{0!} = e^{\frac{-1}{2}} = .604$
        </center>
        <br>
        پس احتمال اینکه دارنده ماشین در طی سفرش باتری تمام نکند تقریبا برابر 0.604 است.
        <br>
        پاسخ بالا یک ویژگی مهم از توزیع‌های نمایی را نشان می‌دهد. در طی پاسخ برای ما زمانی که باتری تا به حال عمر کرده بود مهم نبود. اگر $T$ متغیر تصادفی عمر باتری باشد، و $x$ نشان‌دهنده‌ی عمر باتری تا به حال باشد. $T > x$ یعنی می‌دانیم باتری تا الان به اندازه حداقل $x$ عمر کرده است. و ما به دنبال احتمال این هستیم که این باتری به اندازه $y$   کیلومتر بیشتر عمر کند یعنی احتمال اینکه $T > x + y$. اما در حل مثال نشان دادیم که نیازی به دانستن عمر باتری تا به حال نداشتیم یعنی این احتمال برابر احتمال $T > y$ است. به این ویژگی بی‌حافظه بودن (memoryless) توزیع نمایی میگوییم.
        <br>
       تمرین:<br>
       <li>
       نشان دهید توزیع نمایی بی‌حافظه است یعنی اگر $X$ از توزیع نمایی باشد داریم: $P(X > x + y | X >x) = P(X > y)$
       </li>
       <li>
       یک توزیع بی‌حافظه دیگر توزیع هندسی (گسسته) است. ویژگی بی‌حافظه بودن آن را هم ثابت کنید.
       </li>
       <li>
       به‌طور کلی تنها همین دو توزیع بی‌حافظه وجود دارند (یکی در توزیع‌های پیوسته و دیگری در گسسته). آیا میتوانید یکتایی این توزیع‌ها را نشان دهید.
       </li>
	</font>
