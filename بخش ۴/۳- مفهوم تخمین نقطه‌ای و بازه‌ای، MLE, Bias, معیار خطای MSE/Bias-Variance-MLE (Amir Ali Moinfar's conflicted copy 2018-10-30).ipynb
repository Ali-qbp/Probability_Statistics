		<p></p>
		<hr>
بسیاری از اوقات، ما قصد داریم که با مشاهده داده های تولید شده توسط یک فرایند تصادفی، آن را بهتر بشناسیم. در حالت کلی، هدف ما این است که با استفاده از نمونه های $X_1, X_2, … X_n$ به توزیع متغیر تصادفی $X$ پی ببریم.
        مثلا با استفاده از داده های یک فرایند پوآسون، پارامتر $\lambda$ آن فرایند را بیابیم.
        <br>
       تخمین را می توان در حالت کلی، به دو دسته پارامتری و غیر پارامتری تقسیم کرد. در تخمین 
       پارامتری، ما فرض می کنیم که مدلی که متغیر تصادفی ما از آن تبعیت می کند با یک بردار با بعد متناهی قابل توصیف است. مثلا مدل توزیع نرمال که در آن تمامی متغیر های تصادفی را می توان با استفاده از زوج $(\mu, \sigma)$ 
   توصیف کرد. هر حالتی جز این غیر پارامتری حساب می شود.
        <img src="figs/parametric-est.png">
        <br>
خود تخمین پارامتری را نیز می توان به دو نوع نقطه ای و بازه ای تقسیم کرد.
            در تخمین نقطه ای، یک مقدار به عنوان تخمین ما از پارامتر مساله معرفی می شود. مثلا با مشاهده ۱۰۰ پرتاب سکه که ۳۰ تا شیر و ۷۰ تا خط بوده اند، تخمین می زنیم که احتمال شیر آمدن این سکه هفتاد درصد است.
        در تخمین بازه ای اما، یک بازه را که احتمال می دهیم پارامتر مورد تخمین در آن باشد معرفی می کنیم.
        مثلا در مثال قبل، می گوییم که احتمال شیر آمدن سکه مقداری بین ۶۰ درصد و ۸۰ درصد است.
           <br>
        در بخش بعد بیشتر به تخمین بازه ای می پردازیم اما در این بخش تمرکز ما روی تخمین گر های پارامتری نقطه ای است.
<br>
از این جا به بعد منظور ما از یک تخمین گر تابعی است از فضای نمونه که پارامتر $\theta$ را تخمین می زند. برای نشان دادن مقدار تخمین زده شده، از نماد کلاهک روی پارامتر تخمین ($\hat{\theta}$) استفاده می کنیم. 
دقت کنید که از آن جایی که $\hat{\theta}$ تابعی از فضای نمونه است، خود یک متغیر تصادفی است که توزیع آن به مقدار واقعی $\theta$ بستگی دارد.
<br>
با یک مثال شروع می کنیم

        فرض کنید که متغیر تصادفی $X$، یک متغیر تصادفی پوآسون است و ما از این متغیر تصادفی، داده های $x_1, x_2, ... x_{15}$ را جمع آوری کرده ایم و حالا قصد داریم با استفاده از این داده ها، پارامتر $\lambda$ این متغیر را تخمین بزنیم.
        <br>
        در قدم اول به نظر می رسد که راحت ترین کار این باشد که با توجه به این که امید ریاضی یک متغیر تصادفی پوآسون برابر با پارامتر آن است، برای تخمین پارامتر، ما میانگین این داده ها را محاسبه کنیم.
        <br>
        یک راه دیگر شاید این باشد که برای کوتاهی محاسبه، فقط میانگین ۵ عدد اول را حساب کنیم و دیگر اعداد را دور بیندازیم.
        <br>
روش دیگری که منطقی است این است که با توجه به این که واریانس این متغیر تصادفی نیز برابر با همان $\lambda$ است، ما مقدار واریانس نمونه (<a href = "https://en.wikipedia.org/wiki/Variance#Sample_variance" style="text-decoration:none;">sample variance</a>) که برابر است با 
        $\frac{\sum_{i = 1}^{n}(x_n - \bar{x})^2}{n}$ را، که در این جا $n = 15$ ، به 
        عنوان تخمین خود برای این پارامتر معرفی کنیم.
        <br>
نیازمندیم که برای انتخاب تخمین گر مناسب، به معیاری برای مقایسه آن ها دست پیدا کنیم. ابتدا برای بررسی رفتار این تخمین گر ها، این آزمایش را انجام داده و به تعداد زیاد تکرار می کنیم تا اطمینان حاصل کنیم که نتایج ما تصادفی نیست. یعنی ۱۵ داده از فرایند پوآسون با یک پارامتر ثابت (مثلا ۹) گرفته و تخمینی که هر کدام از روش های بالا برای این پارامتر ارایه می دهند را می یابیم

ابتدا کتاب خانه های لازم را اضافه کرده، و ثابت های مساله را تعیین می کنیم.
            <br>
            پارامتر، همان لاندا مساله است که برابر با ۹ قرار داده ایم.
            <br>
            منظور از num_trial تعداد دفعاتی است که آزمایش بالا را تکرار می کنیم تا .
            <br>
            منظور از num_sample هم تعداد نمونه هایی است که در هر آزمایش از توزیع گرفته و تخمین گر 
            را روی آن ها اعمال می کنیم.
        </font>

```r
library(ggplot2)
library(reshape2)
#constants
num_trial <- 10000
num_sample <- 15
paramater <- 9
```

<p></p>
<br />
<div id="ex1" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
        در این مرحله، برای هر کدام از تخمین گر ها یک تابع، تعریف می کنیم
    </font>
</div>

```r
#estimators
est_1 <- mean

est_2 <- function(data) mean(data[1:5])

est_3 <- function(data) mean((data-mean(data))^2)

estimators <- list(est_1, est_2, est_3)
```

<p></p>
<br />
<div id="ex1" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
        تابع simulator_function را تعریف می کنیم با ورودی گرفتن یک تخمین گر و پارامتر لاندا،
        num_trial مرتبه از توزیع پوآسون با پارامتر لاندا، num_sample نمونه می گیرد و نتیجه تخمین گر روی این num_sample نمونه را به دست می آورد.
خروجی این تابع یک بردار به طول num_trial می باشد.
    </font>
</div>

```r
simulator_function <- function(estimator, lambda) {
  return(sapply(1:num_trial, function(i) estimator(rpois(num_sample, lambda))))
}
```

<p></p>
<br />
<div id="ex1" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
برای این که داده ها را به صورت قابل نمایش در بیاوریم، یک دیتافریم تشکیل می دهیم که ستون اول آن 
        شماره تخمین گر و ستون دومش، نتیجه یکی از آزمایش ها باشد 
        (یعنی در کل 3 * num_trial سطر)
        .
        برای این کار، ابتدا برای هر کدام از تخمین گر ها یک دیتا فریم به شکل بالا به دست می آوریم و سپس این 
        دیتافریم ها را ترکیب می کنیم.
</font>
</div>

```r
#لیستی از دیتافریم ها. این لیست ۳ عضو دارد که هر عضوش دیتافریم مربوط به یک تخمین گر است.
ldf <- lapply(1:length(estimators), function(i) {
  simulation <- simulator_function(estimators[[i]], paramater)
  df <- data.frame(rep(paste("estimator", i), length(simulation)), simulation)
  colnames(df) <- c("estimator number", "estimated value")
  return(df)
}) 
#ترکیب دیتافریم ها
df <- do.call(rbind,ldf)
```

```r
ggplot(df, aes(`estimated value`, fill=`estimator number`)) + geom_density(alpha=0.8) + 
  scale_fill_manual(values = c('orange', 'blue', 'red')) +
  geom_vline(xintercept = paramater, color='dark green')
```

		<p></p>
		<hr>
        در نمودار بالا، خط سبز، نشان دهنده مقدار صحیح $\lambda$ است و نمودار های زرد و آبی و قرمز، به ترتیب نشان دهنده ی رفتار تخمین گر های اول و دوم و سوم هستند.
        با توجه به نمودار بالا، می توان مقایسه ای بین این تخمین گر ها انجام داد.
        اولین نکته ای که واضح است این است که نمودار زرد، متمرکز تر از نمودار آبی و آن نمودار نیز متمرکز تر از نمودار قرمز است. این در واقع به این معنی است که تخمین گر اول، با احتمال نسبتا خوبی حول یک بازه کوچکی تخمین می زند و وقتی که از این تخمین گر استفاده می کنیم می دانیم که مقدار به دست آمده نزدیک به مقدار میانگین خود است و اگر امکان تکرار آزمایش برای تعداد دفعات بالا وجود داشت، به عددی مشابه می رسیدیم. برای این که بتوانیم این خاصیت را به طور کمی تعریف کنیم، داریم:
        <h6>تعریف</h6><i>
        فرض کنید که $\hat{\theta}$  تخمین گری برای پارامتر $\theta$ باشد.
        به مقدار $E[(\hat{\theta} - E[\hat{\theta}])^2]$، <i>$Var(\hat{\theta})$</i> می گوییم.
        </i>
        <br><br>
        دقت کنید که از آن جایی که توزیع $\hat{\theta}$ وابسته به $\theta$ است، در حالت کلی برای محاسبه امید ریاضی نیاز 
        است که مقدار $\theta$ مشخص باشد و در نتیجه، واریانس یک تخمین گر تابعی از $\theta$ است.
        در کل، هر چه واریانس یک تخمین گر کمتر باشد، بهتر است اما این قاعده همیشه درست نیست. برای مثال، تخمین گر ثابت $\hat{\theta} = 19$، واریانس صفر دارد ولی به وضوح تخمین گر خوبی نیست.
بنابراین به بررسی  ویژگی های دیگر نمودار های بالا می پردازیم 
        <br>
        از نمودار بالا مشخص است که تخمین گر اول و دوم، خاصیتی دارند که تخمین گر سوم ندارد. 'مرکز' این دو تخمین گر، بسیار نزدیک به 
        خط سبز است و در نتیجه این تخمینگر ها، به طور میانگین برابر با مقدار واقعی پارامتر $\lambda$ هستند در حالی که تخمین گر سوم این خاصیت را ندارد. به طور دقیق تر داریم:
        <br>
        <h6>تعریف</h6><i>
        فرض کنید که $\hat{\theta}$ تخمین گری برای پارامتر $\theta$ باشد. به مقدار $E [\hat{\theta}] - \theta$
        ، <b>($\hat{\theta}$)bias</b> می گوییم.
        به تخمین گری که bias آن برای تمامی مقادیر $\theta$ برابر با صفر باشد، <b>unbiased</b> می گوییم 
        </i><br><br>
        <div align=center>
            <img src="figs/darts.jpeg">
            <font size=4>
                واریانس در واقع متمرکز بودن تخمین گر و
                $bias$ نزدیکی میانگین تخمین گر به هدف را می سنجد
            </font>
        </div>
        <br>
                   همان طور که خواهیم دید، می توان این دو ویژگی را در ویژگی زیر ترکیب کرد
        <h6>تعریف</h6><i>
        مانند قبل، $\hat{\theta}$ تخمین گری برای $\theta$ است.
        به کمیت زیر، MSE یا mean squared error می گوییم.
            <div align=center>            $MSE(\hat{\theta}) = E[(\hat{\theta} - \theta)^2]$
            </div>
        </i>
        در واقع MSE، معیاری از فاصله تخمین ما با مقدار درست است.
        <h6>قضیه</h6><i>
            <div align=center>            $MSE(\hat{\theta}) = Var(\hat{\theta}) + bias^2(\hat{\theta})$
            </div>
        <b>اثبات:</b>
        دقت کنید که از آن جایی که مقدار $\theta$ ثابت است،  $Var(\hat{\theta}) = Var(\hat{\theta} - \theta)$.
       مانند از طرفی می دانیم که برای هر متغیر تصادفی $Z$، $Var(Z) = E[Z^2] - (E[Z])^2$.
        در نتیجه برای $Z = \hat{\theta} - \theta$ داریم
        <div align=center>            $Var(\hat{\theta}) = Var(\hat{\theta} - \theta) = E[(\hat{\theta} - \theta)^2] 
        - (E[\hat{\theta} - \theta])^ 2 = MSE(\hat{\theta}) - bias(\hat{\theta})^2$
            </div>
        <br>
</font>
</div>

        <br>
                فرض کنید که از یک متغیر تصادفی برنولی، $n$ داده داریم و می خواهیم
            با استفاده از این داده ها، پارامتر $p$ این متغیر را 
            تخمین بزنیم.
            با فرض این که داده ها برابرند با $x_1, ... x_n$
            تعریف کنید، $s = \sum_{i = 1}^{n} x_i$
            در این صورت تخمین گر اول $T_a = \frac{s}{n}$ را برابر با 
            و تخمین گر دوم را برابر با $T_b = \frac{s + 1}{n + 2}$ 
            معرفی می کنیم.
            داریم:
        <br><br>
            <div align=center> $bias(T_a) = \frac{n\theta}{n} - \theta = 0,\;\; bias(T_b) = \frac{n\theta + 1}{n + 2} - \theta = \frac{1 - 2\theta}{n}$
                <br>$Var(T_a) = \frac{\theta(1 - \theta)}{n}, \;\; Var(T_b) = \frac{n\theta(1 - \theta)}{(n + 2)^2}$
                <br>$MSE(T_a) = \frac{\theta(1 - \theta)}{n}, \;\; MSE(T_b) = \frac{n\theta(1 - \theta) \;+ \;(1 -2\theta)^2}{(n + 2)^2}$
            </div>
        <br>
		<hr>
	</font>
</div>

<div id="consistency" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
                    <a href = "#table" style="text-decoration:none;">
سازگاری‌
        </a>
            <br>
        </font>
		<p></p>
        آخرین معیاری که با آن می پردازیم که البته از نمودارهای بالا مشخص نیست، مفهوم سازگاری یا consistency است.
        در عمل، تخمین گر هایی که ما می سازیم، تابعی از $n$ که تعداد داده ها است می باشد و ما انتظار داریم که با افزایش تعداد داده ها، شناختمان نسبت به توزیع بیشتر شود و تخمین گری که این خاصیت را نداشته باشد مطلوب نیست. به طور دقیق تر، به تخمین گر$T$ سازگار یا consistent می گوییم هر گاه
        $\lim_{n \rightarrow \infty} T(x^n) \xrightarrow{P} \theta$.
        <br>
        در مثال پوآسون، تخمین گر $T(x_1, ... x_n) = \frac{\sum_{i = 1}^{n} x_i}{n}$ سازگار است ولی تخمین گر
        $T(x_1, ... x_n) = x_1$ سازگار نیست.
        شاید ابتدا به نظر برسد که این مفهموم، به علت  قضیه حد مرکزی، مشابه bias باشد اما این دو مفهوم متفاوت اند.
        همان طور که دیدیم یک تخمین گر می تواند unbiased ولی ناسازگار باشد ($T(x_1, ... x_n) = x_1$) و نیز می تواند
        سازگار ولی biased باشد. به عنوان مثال برای مثال پیدا کردن $\mu$ برای یک متغیر تصادفی نرمال، تخمین گر $\frac{\sum x_i}{n} + \frac{1}{n}$، biased است ولی 
        با میل دادن n به بی نهایت، می توان دید که این تخمین گر، سازگار است.
        <br>

	</font>

        </a>

		<p></p>
      در این بخش قصد داریم که یک روش کلی برای ساخت تخمین گر ارایه دهیم. فرض کنید که توزیع ما از خانواده $\{f(;\theta)| \theta \in \Theta\}$ باشد، یعنی با دانستن $\theta$ توزیع به طور یکتا مشخص شود. هدف ما تخمین پارامتر $\theta$ است.
        برای این کار، برای مقادیر مختلف $\theta$، احتمال داده ها را حساب می کنیم و به آن به آن تابع درست نمایی یا likelihood می گوییم.
         به طور دقیق تر در حالت گسسته داریم:
        <div align=center>   $\mathcal{L} (\theta | x) = p_{\theta} (x)$
        </div>
        و در حالت پیوسته داریم
         <div align=center>   $\mathcal{L} (\theta | x) = f_{\theta} (x)$
        </div>
        تخمین گر درست نمایی بیشینه، تخمین گری است که با مشاهده داده، مقدار $\theta$ ای که برای آن تابع
        $\mathcal{L} (\theta | x)$ بیشینه است را به عنوان تخمین خود از $\theta$ معرفی می کنیم.
        در واقع به طور شهودی از بین $\theta$ های مختلف، آن مقداری که داده ما را بهتر 
        'توجیه' می کند انتخاب می کنیم.

        در زیر نمودار likelihood را برای پارامتر های مختلف آورده ایم.
        همان طور که از نمودار واضح است، مقدار بیشینه، در 0.7 حاصل می شود

```r
library(ggplot2)
#داده ها
outcome <- c(rep(1, 70), rep(0, 30))
#مقادیر مختلف پارامتر
possible_ps <- seq(0, 1, by = 0.001)
#تابع درست نمایی
likelihood_function <- function(ps) sapply(ps, function(p) 
  prod(sapply(outcome, function(x_i) p*x_i + (1 - p) * (1 - x_i))))
#پارامتری که بیشترین درست نمایی را دارد
p_ML <- possible_ps[which.max(likelihood_function(possible_ps))] 
print(p_ML)
ggplot(data = data.frame(possible_ps), mapping = aes(x = possible_ps)) + ylab("likelihood") +
  stat_function(fun = likelihood_function) + geom_vline(xintercept = p_ML, color="dark green")
```

        بسیاری از مواقع، ما از یک توزیع تعداد زیادی داده مستقل داریم. برای یافتن تخمین گر بیشینه باید 
        عبارت $\mathcal{L(\theta | D)} = \prod P(X_i | \theta)$
        را بیشینه کنیم. اکثر مواقع برای یافتن $\theta$ ای که عبارت فوق را بیشینه می کند، باید از آن مشتق بگیریم که اگر چه با توجه به قواعد مشتق گیری، قابل انجام است، از نظر جبری وقت گیر است. تکنیکی رایج در این مواقع این است که 
        به جای این که از خود تابع درست نمایی مشتق بگیریم، از لگاریتم آن مشتق گرفته و مقدار آن مشتق را برابر صفر قرار دهیم. دقت کنید که از آن جایی که تابع لگاریتم صعودی است، $\theta$ ای که $\mathcal{L}(\theta | D) $
        را بیشینه کند همان $\theta$ ای است که $ln(\mathcal{L}(\theta | D))$ را بیشینه می کند.
        از طرفی با این کار محاسبات ما راحت تر می شود زیرا 
        <div align=center>               $\frac{d}{d \theta}ln(\mathcal{L}(\theta | D)) = \frac{d}{d \theta}ln(\prod P(X_i | \theta))
             = \frac{d}{d \theta} \sum ln P(X_i | \theta) = \sum \frac{d}{d \theta} ln P(X_i | \theta)$
        </div>
        که محاسبه عبارت آخر معمولا ساده تر از محاسبه عبارت $\frac{d}{d \theta} \mathcal{L}(\theta | D)$ می باشد.
        <br>
        مثال بالا را بار دیگر با استفاده از لگاریتم درست نمایی بدست می‌آوریم.
        <br>
        همان طور که گفتیم، باید مقدار 
        $\sum \frac{d}{d \theta} ln P(x_i | \theta)$ را برابر با صفر قرار دهیم.
        <div align=center>     $0 = \sum \frac{d}{d \theta} ln P(x_i | \theta) = \sum \frac{d}{d \theta} ln (\theta^{x_i}(1 - \theta)^{1 - x_i})
    = \sum \frac{d}{d \theta} (x_i ln(\theta) + (1 - x_i) ln (1 - \theta)) = \sum (\frac{x_i}{\theta} - \frac{1 - x_i}{1 - \theta})$    <br> $\rightarrow \theta_{ML} = \frac{\sum x_i}{n} = 0.7$
        </div>
</font>
