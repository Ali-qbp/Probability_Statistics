مقدمه
        </font>
        <p></p>
        <hr>
        اکثر مسائلی که تا اینجای کار بررسی کردیم به این شکل بودند که ما ویژگی های یک جامعه بزرگ (به عنوان مثال توزیع احتمال یک متغیر تصادفی در آن جامعه) را می‌دانستیم و می‌خواستیم براساس آن یک رویداد را پیش‌بینی کنیم. اما اگر بخواهیم ویژگی های جامعه را به دست آوریم باید چه کنیم؟ برای مثال اگر بخواهیم پیش‌بینی کنیم در انتخابات آتی چه نامزدی رای خواهد آورد باید بدانیم اکثریت جامعه به کدام نامزد رای خواهد داد اما اینکار بسیار هزینه بر است. در مبحث احتمال دیدیم که می‌توان ویژگی‌های یک نمونه کوچک را از روی ویژگی‌های جامعه تعیین کرد. علم آمار به ما می‌گوید عکس این هم امکان‌پذیر است. یعنی می‌توان از روی ویژگی‌های یک نمونه کوچک از جامعه به ویژگی های کل جامعه پی‌برد. برای مثال اگر بخواهیم نتیجه انتخابات را پیش‌بینی کنیم می‌توانیم صرفا از یک نمونه ۱۰۰ یا ۱۰۰۰ نفری استفاده کنیم و اکثریت آن را به جامعه نیز تعمیم دهیم. البته این‌کار به شرطی امکان پذیر است که نمونه ما به اندازه کافی تصادفی باشد که آن را در ادامه بررسی می‌کنیم.

<div id="sec_infer" style="direction:rtl;line-height:300%;">

استنتاج آماری
        </font>
        <p></p>
        <hr>
        به طور کلی استنتاج آماری به دو شیوه تقسیم بندی می‌شود. شیوه اول که به آن شیوه <b> فروان‌گرایانه </b> (frequentist) می‌گویند فرض می‌کند پارامتری از جامعه که به دنبال تخمین مقدار آن هستیم یک مجهول غیر تصادفیست که از طریق نمونه‌گیری به تعداد زیاد قابل تخمین است. به عنوان مثال با این شیوه اگر بخواهیم میانگین یک جامعه را تخمین بزنیم یک نمونه تصادفی از افراد آن جامعه انتخاب می‌کنیم. جواب‌های آن ها را یادداشت میکنیم و میانگین آن‌ها را به عنوان میانگین جامعه انتخاب می‌کنیم.
        <br>
        استفاده از بازه اطمینان برای تخمین بازه‌ای و p-value برای آزمون فرض از مثال‌های به‌کارگیری این شیوه هستند که در ادامه با آن‌ها آشنا خواهید شد.
        <br><br>
        شیوه دوم که به آن شیوه <b> بیزی </b> (Bayesian) می‌گویند فرض می‌کند پارامتری از جامعه که به دانستن آن علاقه‌مندیم خود یک متغیر تصادفیست. بنابراین در این شیوه بحث درباره تخمین مقدار پارامتر نیست بلکه درباره تخمین توزیع آن است. در این شیوه ابتدا یک حدس اولیه درباره توزیع متغیر تصادفی مورد علاقه‌مان می‌زنیم و سپس با آزمایش و مشاهده یک نمونه، توزیع جدید آن را با کمک قانون بیز به دست می‌آوریم. به عبارت دیگر اگر متغیر تصادفی مورد علاقه ما $\theta$ باشد و متغیر تصادفی مورد آزمایش $X$ به $\theta$ بستگی داشته باشد آنگاه داریم 
        <br>
        <center>$ P(\theta|X) = \frac{P(X|\theta)P(\theta)}{P(X)}$</center>
        که در رابطه فوق $P(\theta)$ حدس اولیه ما و $P(\theta|X)$ نتیحه‌گیری ما از روی آزمایش است. تخمین‌گر MAP و توزیع conjugate prior از مثال‌های استفاده از این شیوه هستند. 
    </font>
</div>

کاربردهای آمار
        </font>
        <p></p>
        <hr>
        برای انجام تحلیل‌های آماری در مرحله اول به طراحی آزمایش و جمع‌آوری داده نیاز داریم. در این مرحله باید داده‌های ما به گونه‌ای مناسب برداشته شوند که آن‌را در ادامه بررسی می‌کنیم. اما سوال اینجاست که پس از جمع آوری این اطلاعات چه کارهایی می‌توان با آن‌ها انجام داد؟ 
        <ul>
            <li>
                فرض‌کنید طبق معمول یک سکه داریم که این بار احتمال شیرآمدنش را نداریم. برای اینکه بتوانیم این احتمال را حدس بزنیم سکه را چندین بار پرتاب می‌کنیم و تعداد شیرها را ثبت می‌کنیم. پس از انجام نمونه‌گیری به این شیوه، باید یک روش مناسب برای حدس زدن مجهولمان داشته باشیم که به آن <i><b> تخمین </b></i> می‌گویند. تخمین می‌تواند به صورت نقطه‌ای یا بازه‌ای باشد یا در آن مجهول ما یک متغیر تصادفی فرض شود یا صرفا یک پارامتر نامعلوم. در ادامه درس با انواع این تخمین ها آشنا خواهیم شد.

        این موارد چند نمونه محدود از تحلیل‌های متنوعی بودند که می‌توانیم روی داده‌های آزمایش انجام دهیم و نتایج جالبی از آن به دست اوریم.

<div id="sec_sample" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
نمونه‌گیری تصادفی
        </font>
        <p></p>
        <hr>
        در قسمت قبل دیدیم که انتخاب یک نمونه تصادفی مناسب نقش مهمی در آزمایش دارد و در صورتی می‌توانیم نتایج نمونه را به کل جامعه نسبت دهیم که نمونه ما به اندازه کافی تصادفی باشد. اما معیار تصادفی بودن نمونه چیست؟ بدیهی‌است که برای اینکه نمونه ما به خوبی بیانگر جامعه باشد، هر کدام از اعضای جامعه باید شانس برابری برای حضور در نمونه داشته باشند. به عنوان مثال، زمانی که تازه اینترنت فراگیر شده بود استفاده از نظرسنجی اینترنتی به عنوان نمونه برای پیش‌بینی نتایج انتخابات اشتباه بود چراکه تنها طبقه‌ی تحصیل‌کرده و مرفه جامعه به اینترنت دسترسی داشتند.
        <br/><br/>
        حال می‌خواهیم کمی ریاضی تر نمونه‌گیری را بررسی کنیم. فرض کنید می‌خواهیم یک خاصیت مشخص مثلا قد افراد جامعه را بررسی کنیم. در اینجا می‌توان قد یک فرد را یک متغیر تصادفی مانند $X$ فرض کرد که توزیع آن از فراوانی نسبی قد افراد جامعه به‌دست می‌آید. در نمونه‌گیری باید برای فهمیدن ویژگی های قد در جامعه(برای مثال امیدریاضی یا واریانس $X$) باید $n$ تا از اعضای جامعه را انتخاب کنیم. قد هر کدام از این افراد نیز یک متغیر تصادفی است، پس متغیر تصادفی قد نفر $i$ام را با $X_i$ نمایش می‌دهیم. در اینجا دو استراتژی برای انتخاب نمونه داریم. استراتژی اول انتخاب با جایگذاری است. به این معنا که ممکن است یک نفر چند بار در نمونه‌گیری ظاهر شود. در این صورت بدیهی‌است که توزیع هرکدام از $X_i$ ها برابر با توزیع $X$ خواهد شد. حتی در صورتی که نمونه‌گیری ما بدون جایگذاری هم باشد باز توزیع تمام $X_i$ ها با توزیع $X$ برابر است (چرا؟). اما درباره استقلال $X_i$ ها باید مراقب باشیم چراکه اگر نمونه‌گیری ما بدون جایگذاری باشد و تعداد اعضای جامعه متناهی باشد، دیگر $X_i$ ها مستقل نخواهند بود. اما در عمل حتی در نمونه‌گیری بدون جایگذاری معمولا تعداد اعضای جامعه به اندازه‌ای بزرگ است که نمونه‌گیری با جایگذاری و بدون جایگذاری تفاوت چندانی با هم ندارند. 
        <br/>
        <br/>
        حالا که نمونه‌گیری را انجام دادیم زمان آن رسیده از آن اطلاعاتی را برداشت کنیم. برای مثال می‌توانیم مجموع نمونه را حساب کنیم
        <br/>
        <div align=center>
       $S$ = $X_1 + X_2 + ... + X_n$
       </div>
       <br/>
       مشخص است که $S$ نیز خود یک متغیر تصادفی است. پس می‌توان برای آن امید ریاضی و واریانس حساب کرد.
       <div align=center>
        <br/>
       $\mathbf{E}(S) = \mathbf{E}(X_1 + X_2 + ... + X_n) = \mathbf{E}(X_1) + \mathbf{E}(X_2) + ... + \mathbf{E}(X_n) = n\mu$
       </div>
       <br/>
        که در رابطه فوق $\mu = \mathbf{E}(X)$. به همین ترتیب می‌توان $Var(S)$ را اگر $X_i$ها از هم مستقل باشند به دست آورد.
        <div align=center>
        <br/>
        $Var(S) = Var(X_1 + X_2 + ... + X_n) = Var(X_1) + Var(X_2) + ... + Var(X_n) = n\sigma^2$
        </div>
        <br/>
        که در رابطه فوق $Var(X) = \sigma^2$. همچنین می‌توان میانگین نمونه را به صورت $\bar{X} = \frac{S}{n}$ تعریف کرد که خود یک متغیر تصادفی دیگر است و داریم $\mathbf{E}(\bar{X}) = \mu \: \: \: Var(\bar{X}) = \frac{\sigma^2}{n}$
        <br/>
        <br/>
        حال می‌خواهیم به صورت تجربی رفتار میانگین نمونه را به عنوان یک متغیر تصادفی بررسی کنیم.در قطعه کد زیر میانگین سایز برنامه‌های پلی استور گوگل (که داده‌های آن در فایل googleplaystore.csv آمده است) را بررسی می‌کنیم. برای تخمین زدن امید ریاضی و واریانس $\bar{X}$ هر نمونه‌گیری با اندازه مشخص $n$ را ۱۰۰ بار تکرار می‌کنیم و میانگین و واریانس آن را به عنوان میانگین و واریانس $\bar{X}$ در نظر می‌گیریم. به نمودار‌های میانگین و واریانس برحسب $n$ دقت کنید و با تغییر ثوابت، تغییر رفتار در نمودار‌ها را مشاهده کنید. 
    </font>
</div>

```r
library(stringr)
library(ggplot2)


df <- read.csv("data/googleplaystore.csv")
sizes <- as.numeric(str_replace_all(df$Size, "M", ""))
sizes <- sizes[which(!is.na(sizes))]

sample_xbar <- function(n, vec) {
  return(sapply(rep(n, 100), function(i) mean(sample(vec, i))))
}

df <- data.frame(x = 1:500, mean = sapply(1:500, function(i) mean(sample_xbar(i, sizes)))
                 , var = sapply(1:500, function(i) var(sample_xbar(i, sizes))))

ggplot(df) + geom_point(aes(x = x, y = mean, color = 'sample means')) + 
  geom_line(aes(x = x, y = mean(sizes), color = 'poplutation mean')) + labs(x = 'n', y = 'mean')
ggplot(df) + geom_line(aes(x = x, y = var)) + labs(x = 'n', y = 'var')
```

	<font face="XB Zar" size=5>

خلاصه
        </font>
        <p></p>
        <hr>
       در این بخش کمی با تفاوت آمار و احتمال و کاربردهای آمار آشنا شدیم و پس از آن نمونه‌گیری تصادفی را دیدیم و همچنین مشاهده کردیم که برای یک نمونه $X_1, X_2, ..., X_n$ که از توزیع اصلی $X$ آمده اند می‌توان متغیر‌های تصادفی $S$ و $\bar{X}$ را به شکل زیر تعریف کرد
        <div align=center>
            $S = X_1 + X_2 + ... + X_n \\ \bar{X} = \frac{S}{n} \\ \mathbf{E}(S) = n\mathbf{E}(X) \\ \mathbf{E}(\bar{X}) = \mathbf{E}(X)$
            </div>
            همچنین اگر $X_i$ ها مستقل باشند (به عبارت دیگر نمونه‌گیری تصادفی بدون جایگذاری از جامعه متناهی نباشد) روابط زیر نیز برقرار هستند.
        <div align=center>
            $Var(S) = nVar(X) \\ Var(\bar{X}) = \frac{Var(X)}{n}$
            </div>
    </font>
</div>

<div id="sec_practice" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
تمرین
        </font>
        <p></p>
        <hr>
        گفتیم اگر نمونه‌گیری ما به صورت بدون جایگذاری از مجموعه متناهی باشد دیگر متغیرهای تصادفی ما مستقل نیستند. ثابت کنید در این حالت روابط زیر برقرار هستند.
        <br/>
        <br/>
        <div align=center>
            $Var(S) = n\sigma^2\frac{N - n}{N - 1} \\ Var(\bar{X}) = \frac{\sigma^2}{n}\frac{N - n}{N - 1}$
        </div>
        <br/>
        همانطور که از این روابط مشخص است واریانس جمع و میانگین نمونه در صورتی که نمونه‌گیری به صورت بدون جایگذاری باشد با ضریب $\frac{N - n}{N - 1}$ کاهش می‌یابد و در صورتی که با $n$ ثابت $N$ به بی‌نهایت میل کند این ضریب برابر با ۱ شده و بین نمونه‌گیری بدون جایگذاری و با جایگذاری تفاوتی نخواهد بود.
        <br/>
        (راهنمایی: ابتدا نشان دهید $Cov(X_i, X_j) = -\frac{\sigma^2}{N - 1}$ )
