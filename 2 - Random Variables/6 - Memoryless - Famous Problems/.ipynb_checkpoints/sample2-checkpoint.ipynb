سرقت از بانک آهنین
        </font>
		<p></p>
		<hr>
  کلثوم که تصور میکرد دستیارهای آموزشی درس آمار پول زیادی از نوشتن جزوه کسب میکنند پس از مشاهده وضعیت مالی تصمیم گرفت راه دیگری برای کسب درآمد پیدا کند. وی که خستگی از ویژگی‌های بارزش بود و همواره به دنبال راه‌های آسان و سریع برای پول درآوردن بود تصمیم گرفت به بانک آهنین دستبرد بزند...

<div id="bank" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
در بانک
        </font>
		<p></p>
		<hr>
        کلثوم پس از ماه‌ها نقشه‌ی دقیقی کشیده و وارد خزانه بانک میشود. خزانه به صورت یک راهروی بسیار طولانی میباشد که در دو طرف آن گاوصندوق‌هایی بر دیوار قرار دارند. رمز این گاوصندوق‌ها سه‌رقمیست و تنها یک بار میتوان رمز را امتحان کرد و در صورت اشتباه بودن گاوصندوق غیرفعال میشود.
        <img src="figs/Capture.jpg" width="500">کلثوم از ابتدا شروع کرده و عدد شانسش(007) را برای همه‌ی صندوق‌ها امتحان میکند. احتمال این که کلثوم نیاز داشته باشد رمز خود را بر روی بیش از یک گاوصندوق امتحان کند تا به گاوصندوقی با رمز مورد نظر برسد چقدر است؟ (رمز هرکدام از گاوصندوق‌ها عددی تصادفی بین 000 تا 999 میباشد)
        <br />
        پاسخ: متغیر تصادفی $X$ را شماره صندوقی که با این رمز باز میشود در نظر میگیریم. این متغیر تصادفی از توزیع هندسی با پارامتر $p=\frac{1}{1000}$ پیروی میکند. بنابراین احتمال باز شدن اولین صندوق با این رمز $P(X>1) = \frac{999}{1000}$ میباشد.
        <br />
        <br />
        متاسفانه رمز کلثوم هیچ کدام از 500 گاوصندوق اول را باز نمیکند. اکنون احتمال این که کلثوم نیاز داشته باشد رمز خود را بر روی بیش از یک گاوصندوق دیگر امتحان کند تا به گاوصندوقی با رمز مورد نظر برسد چقدر است؟
        <br />
       پاسخ:
        <br />
        <div style="text-align: left">$P(X > 501 | X > 500) = \frac{P(X > 501)}{P(X > 500)} = \frac{(\frac{999}{1000})^{501}}{(\frac{999}{1000})^{500}} = \frac{999}{1000}$</div>
            همانطور که مشاهده میکنید این مسئله که رمز 500 گاوصندوق اول عدد شانس کلثوم نبود ارتباطی با احتمال این که کلثوم به بیش از یک تلاش دیگر برای پیدا کردن یک گاوصندوق با رمز مورد نظر داشته باشد نداشت, به چنین ویژگی برای یک متغیر تصادفی بی‌حافظه بودن میگویند.
             <br />
	</font>
</div>


<div style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=blue size=6>
کنجکاوی:
        </font><br/>
        امید ریاضی اولین صندوقی که با این رمز باز میشود را در دو حالت محاسبه و مقایسه کنید.
        <br/>
مسئله را بار دیگر به صورت امتحان کردن رمزهای مختلف بر روی یک گاوصندوق حل کنید. آیا باز هم جواب در دو حالت یکسان است؟
	</font>
</div>

<div id="memorylessness" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
بی‌حافظگی
        </font>
		<p></p>
		<hr>
متغیر تصادفی $X$ زمانی بی‌حافظه است که توزیع شرطی احتمال زیر در مورد آن صدق کند:
        <div style="text-align: left"> $Pr[X > s+t | X > t] = Pr[ X > s], s,t > 0 $ </div>
در صورتی که $X$ را به صورت طول عمر یک دستگاه در نظر بگیریم معادله بالا بیانگر آن است که احتمال این که این دستگاه حداقل به اندازه $s + t$ ساعت عمر کند با توجه به این که $t$ ساعت عمر کرده برابر با احتمال اولیه‌ی این است که حداقل $s$ ساعت عمر کند. به بیان دیگر اگر دستگاه در زمان $t$ هنوز سالم باشد توزیع طول عمر باقیمانده‌ی آن همان توزیع طول عمر اولیه‌ی دستگاه میباشد. (دستگاه فراموش میکند که به مدت $t$ استفاده شده است)
	</font>
</div>

<div id="escape" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
فرار از بانک
        </font>
		<p></p>
		<hr>
       کلثوم که ماشین ندارد و مسیر بانک را با اسنپ آمده بود حالا شارژ موبایلش تمام شده و مجبور است یکی از ماشین‌های موجود پارکینگ را بردارد. او میداند که طول عمر باتری موجود در ماشین‌های پارکینگ بانک آهنین از توزیع نمایی پیروی میکنند. بر روی باتری هر ماشین میانگین توزیع طول عمر آن و بر روی کیلومترشمار آن مسافت طی شده توسط ماشین نوشته شده. با توجه به اطلاعات در دسترس کلثوم چگونه باید ماشینی را انتخاب کند که به احتمال بیشتری بدون تمام شدن باتری در حین فرار بتواند به مقصد برسد؟ 
        <br />
        پاسخ: توزیع نمایی یک توزیع بی‌حافظه است. زیرا:
        <div style="text-align: left"> $Pr[X > s+t | X > t] = \frac{Pr[X > s+t, X > t]}{Pr[x>t]} = \frac{1-F(s+t)}{1-F(t)} = \frac{e^{s+t}}{e^{t}} = e^s = Pr[X > s]$ </div>
        بنابراین مسافت طی شده توسط ماشین‌ها اهمیتی ندارد و بر احتمال تمام شدن باتری در طی مسافت $s$ تاثیرگذار نیست. در نتیجه بهترین انتخاب باتری با میانگین طول عمر بیشتر است هر چند که بیشتر از همه کار کرده باشد.
        <br />
        <br />
        پ.ن:در دنیای واقعی عمر باتری ماشین از توزیع نمایی پیروی نکرده و با گذر زمان احتمال بقای آن کمتر میشود اما از مثال‌های بی‌حافظگی میتوان به نیمه عمر مواد رادیواکتیو ,زمان کشف بلاک جدید بیتکوین و یا زمان رسیدن مشتری بعدی در نظریه صف اشاره کرد.
	</font>
</div>

اعتراف
        </font>
        <p></p>
        <hr>
کلثوم که از کرده‌ی خود پشیمان شده تصمیم میگیرد به نزدیکترین پاسگاه مراجعه کرده و خود را تحویل پلیس بدهد. پاسگاه دو منشی دارد که هر دو با توزیع نمایی یکسانی پاسخگو میباشند. زمانی که کلثوم به آنجا میرسد هر دو مشغول اعتراف گرفتن از فرد دیگری هستند و به محض این که اعتراف هر یک تمام شود نوبت به اعتراف کلثوم میرسد. احتمال این که کلثوم آخرین نفری باشد که اعترافش تمام میشود چقدر است؟
        <br />
        پاسخ: از آن جا که توزیع نمایی بی حافظه است زمانی که نفر اول اعترافش تمام شود مانند این است که کلثوم و فرد دیگر با توزیع یکسانی همزمان شروع به اعتراف کرده‌اند. پس احتمال این که اعتراف کلثوم دیرتر تمام شود برابر با $\frac{1}{2}$ میباشد.
        <br />
        <br />
        تمرین: <br />
        <li>قطعه کد زیر را تکمیل کنید و با استفاده از آن پاسخ مسئله بالا را به دست آورید. (همچنین میتوانید با رسم نموار زمان پایان اعترف مجرم دیگر که همزمان با کلثوم در حال اعتراف است از زمانی که کلثوم شروع به اعتراف میکند به پیروی آن از همان توزیع نمایی اصلی پی ببرید)
        </li>
       </font>
</div>
