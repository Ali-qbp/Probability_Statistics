            نکاتی که در ادامه گفته می‌شود نیز اکثرا در فیدبک سال گذشته نیز گفته شده است.
        </li>
        <li>
قالب کلی جزوه‌ها می‌بایست بدین صورت باشند که با یک مقدمه و یک سوال شروع شوند تا کنجکاوی خواننده را بیانگیزند. لطفا این قالب رعایت شود. برای نوشتن مقدمه به عنوان پیشنهاد می‌توانید از این دو لینک استفاده کنید.
            <a href=https://www.quora.com/Why-do-we-need-to-define-the-random-variable-What-is-it-that-we-can-do-with-a-random-variable-that-we-can-not-do-with-sample-space-event-space-and-probability-measure> لینک اول </a>

بخشی که متغیرهای تصادفی گسسته و پیوسته توضیح داده شده‌اند را با یک تیتر جدا کنید تا قابل تشخیص باشند.
        </li>
        <li>
            کد 
            R
           موجود ، به علت اینکه هنوز هیچ توزیعی تا اینجای درس معرفی نشده است ممکن است گیج‌کننده باشد. به نظر من یک توضیح بسیار کوتاه شهودی راجع به توزیع 
            uniform
            بدهید و بگویید در آینده بیشتر با آن آشنا خواهید شد. همچنین مورد سوم مثال که عدد حاصل از پرتاب دوم به توان دو است به نظرم همان‌طور که در فیدبک پارسال گفته شده است، خیلی به مفهوم تابعی از متغیر تصادفی وابسته است و بهتر است حذف شود.
        </li>
        <li>
            <h1 Sytle=" color:#FF7500; "> مهم </h1>
            <br>
            این جزوه ترکیب دو جزوه مجزای سال گذشته است. اما ترکیب آن‌ها جزوه را دارای دو دستگی کرده است و انگار دو جزوه مختلف فقط به هم متصل شده‌اند و تبدیل به یک جزوه واحد نشده‌اند. مثلا لزوم وجود بخش  محاسبه احتمال با استفاده از متغیر تصادفی چیست وقتی در ادامه دو بخش
            PMF
            و 
            CDF 
           وجود دارند؟ لطفا جزوه را یکبار به صورت کامل بازبینی  و بازنویسی کنید تا جزوه یکپارچه شود. 
        </li>
        <br>
        <li>
             دو بخش 
            PMF
            و
            CDF
            را همانند قالب مطلوب جزوه‌ها دربیارید. یعنی با یک مثال که کنجکاوی را برانگیزد آغاز کنید و سپس سراغ اصل موضوع بروید. همچنین در توضیحات ابتدا سعی کنید به صورت شهودی مطلب جا بیفتد و سپس به سراغ قضایای جبری بروید. خصوصا برای CDF. 
            در بخش محاسبه احتمال با استفاده از متغیر تصادفی یک توضیح خوب از علت نیاز ما به CDF نوشته شده است.
            (
            البته یک اشتباه دارد که در زیر اصلاح شده است.همچنین قرار شد بخش محاسبه احتمال با استفاده از متغیر تصادفی را پاک کنید اما این توضیح را می‌توانید به بخش CDF اضافه کنید.
            )
            <br>
            "
حال مثلا اگر احتمال 
            $\{X\le x\}$ 
            را برای هر 
            $x$
            بدانیم، 
            احتمال همهٔ واقعه‌های مورد نظر را خواهیم داشت و دیگر لزومی به دانستن فضای احتمال 
            $\Omega$ 
            نیست
            "
        </li>
        <li>
        در بخش امید ریاضی، توضیح نوشته شده خوب است. اما به نظرم با توجه به اهمیت امید ریاضی بیشتر باید توضیح داده شود و یک مثال شهودی   گذاشته شود. مثال موجود بسیار ابتدایی است. همچنین همچنان مشکل مطابق نبودن با قالب مطلوب جزوه‌ها وجود دارد. یعنی سعی کنید با یک مثال جالب وارد موضوع شوید و سپس رو به قضایای جبری بیاورید.
