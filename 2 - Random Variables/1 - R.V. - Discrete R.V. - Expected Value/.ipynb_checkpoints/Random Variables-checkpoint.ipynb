<p></p>
<br />
<div id="sec_intro" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مقدمه
        </font>
		<p></p>
		<hr>
در قسمت‌های قبلی درس یاد گرفتیم که فضای نمونه
مانند
$\Omega$
مجموعه‌ای از همه خروجی‌های مختلف یک آزمایش تصادفی است.
        <br>
        متغیرهای تصادفی توابعی هستند که این خروجی‌ها را به اعداد حقیقی نگاشت می‌کنند.
<br/>
     این نگاشت در برخی آزمایش‌ها واضح‌تر از بقیه آزمایش‌هاست. به طور مثال در آزمایش تصادفی پرتاب تاس، خروجی‌های تاس اعداد ۱ تا ۶ هستند. بنابراین اولین گزینه‌ای که برای  تعریف  متغیر تصادفی
به ذهن می‌رسد نگاشت 
((عددی که پس از پرتاب روی تاس می‌آید))
به اعداد طبیعی بازه ۱ تا ۶ است.
        <br/>
        ولی در مورد پرتاب سکه این امر کمی متفاوت است. زیرا
        شیر یا خط آمدن سکه به طور ذاتی هیچ
        ((صفر))
        یا
        ((یک))
        ای
        در خود ندارد، ولی ما معمولاً بدون این‌که به این مورد زیاد فکر کنیم یکی از آن‌ها را به صفر و دیگری را به یک می‌نگاریم. این پل بین
        شیر یا خط و 0 یا ۱
        همان متغیر تصادفی است.
<br/>
<br/>
حال شاید این سوال ایجاد شود که چرا یک تابع را
((متغیر تصادفی))
می‌نامیم؟
        <br/>
        دلیل آن این است که
چون به خروجی یک آزمایش
تصادفی
        مربوط است،
مقدار آن
((تصادفی))
است و چون با آن‌ها مانند متغیرهای ریاضی
رفتار می‌کنیم و آن‌ها را با یکدیگر جمع و ضرب می‌کنیم، به توان می‌رسانیم و عملیات ریاضی انجام می‌دهیم، آن‌ها را متغیر نیز می‌نامیم.
<br />
	</font>
</div>

<p></p>
<br />
<div id="sec_def" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
تعریف
        </font>
		<p></p>
		<hr>
به طور دقیق‌تر فرض کنید یک آزمایش و یک مجموعه از خروجی‌های ممکن (فضای نمونه)
$\Omega$
        در اختیار داشته باشیم. یک متغیر تصادفی، به هر یک از خروجی‌های ممکن مانند
$\omega$
که
$\omega \in \Omega$
یک عدد تخصیص می‌دهد. به بیان ریاضی می‌توان گفت متغیر تصادفی تابعی از نقاط فضای نمونه مثل $X(\omega)$ است که به هر یک از نقاط فضای نمونه عددی حقیقی را نسبت می‌دهد. 
<br/>
        دامنهٔ این تابع مجموعهٔ $\Omega$ بوده و برد آن مجموعه‌ای از اعداد حقیقی است.
        <br />
<img src="figs/rv.png" style="float:center; width:50%"/>
<br />
(توجه کنید که در این جا متغیر تصادفی حقیقی مد نظر ماست؛ متغیر تصادفی مختلط به هر یک از نقاط دامنه عددی مختلط را نسبت می‌دهد)
<br/>
به طور مثال در پرتاب تاس خروجی آزمایش تصادفی،
به طور دقیق
،
((دایره‌های روی وجه بالایی))
تاس است که معمولاً به کمک یک متغیر تصادفی، خروجی را به یک عدد حقیقی می‌نگاریم و عملیات عددی انجام می‌دهیم.
<img src="figs/dice4_2.png" style="width:50%;" />
<br/>
        متغیرهای تصادفی را به صورت کلی به دو نوع پیوسته و گسسته می‌توان تقسیم کرد.
<br/>
        متغیر تصادفی گسسته خروجی‌های مختلف یک آزمایش تصادفی را به زیر مجموعه‌ای شمارا (متناهی یا نامتناهی) از اعداد حقیقی نگاشت می‌کند. مثلاً خروجی یک بار پرتاب تاس اعداد طبیعی ۱ تا ۶ است یا
تعداد دفعاتی که بایستی یک سکه انداخته شود تا برای اولین بار شیر بیاید، مجموعه همه اعداد طبیعی است که شمارا است. بنابراین در این آزمایش‌ها متغیر تصادفی گسسته است.
<br/>
متغیرهای تصادفی پیوسته هم متغیرهایی هستند که خروجی یک آزمایش تصادفی را به زیرمجموعه‌ی نامتناهی ناشمارا از اعداد حقیقی نگاشت می‌کنند. این زیرمجموعه، که برد متغیر تصادفی است، به طور مثال می‌تواند
$[4,7]$،
$[0, \infty)$
یا
$\mathbb{R}$
باشد.
متغیر تصادفی پیوسته برای نگاشت خروجی‌های آزمایشی به کار می‌رود که
فضای نمونه آن
ناشمارا باشد.
معمولاً در این درس مجموعه ناشمارای مهمی که با آن سر و کار داریم، مجموعه اعداد حقیقی است.
یعنی به طور ساده می‌توان جمله بالا را این‌گونه تعبیر کرد که اگر آزمایش تصادفی ما خروجی‌ای حقیقی و پیوسته داشته باشد، متغیر تصادفی ما پیوسته است.
<br/>
        به طور مثال همه افراد زنده روی کره زمین در یک لحظه خاص
را در نظر بگیرید. فردی را به تصادف انتخاب می‌کنیم. اگر داده مورد علاقه ما سن افراد باشد، متغیر تصادفی‌ای که تعریف می‌کنیم به عنوان ورودی یک انسان می‌گیرد و به عنوان خروجی عدد سن فرد را فرضاً به سال تحویل می‌دهد
(مثلاً فردی می‌تواند
$12.3456$
سال سن داشته باشد).
        اگر فرض کنیم که انسان حداکثر ۱۵۰ سال می‌تواند عمر کند،
برد متغیر تصادفی، اعداد حقیقی بازه
$[0, 150]$
است.
این مجموعه ناشماراست، بنابراین متغیر تصادفی ما پیوسته است.
<br/>
(به بیان ساده اگر خروجی آزمایش از جنس گسسته مثل عدد تاس و شیر یا خط بودن سکه بود، متغیر تصادفی گسسته است. اگر از جنس پیوسته مثل طول و زمان بود، متغیر پیوسته است.)
<br/>
از آن‌جایی که درک بهتر متغیر تصادفی پیوسته نیازمند تعریف  تابع چگالی احتمال است، در این بخش با متغیر تصادفی گسسته آشنا می‌شویم و در بخش‌های بعدی متغیر پیوسته را به همراه نحوه دقیق تعریف کردن آن می‌آموزیم.
	</font>
</div>

<p></p>
<br />
<div id="sec_example" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
چند مثال
        </font>
		<p></p>
		<hr>
        چند نمونه آزمایش‌های زیر را در نظر بگیرید:
<ul>
    <li>
        در یک آزمایش شامل پرتاب ۵ سکه، تعداد «شیر»ها در این دنباله یک متغیر تصادفی است. البته توجه کنید که خود دنباله یک متغیر تصادفی نیست چون یک مقدار عددی مشخص ندارد.
    </li>
    <li>
        در یک آزمایش ۲ بار یک تاس را پرتاب می‌کنیم. هر یک از موارد زیر یک متغیر تصادفی محسوب می‌شوند:
        <ol>
            <li>
                مجموع عدد حاصل از دو پرتاب
            </li>
            <li>
                تعداد ۶ها در دو پرتاب
            </li>
            <li>
                عدد حاصل از پرتاب دوم به توان ۲
            </li>
        </ol>
    </li>
    <li>
        در یک آزمایش شامل انتقال یک پیام، زمان لازم برای انتقال پیام، تعداد نشانه‌های اشتباه دریافت شده و تاخیر دریافت پیام همه متغیرهای تصادفی به حساب می‌آیند.
    </li>
</ul>
        در ادامه قطعه کدی ساده مرتبط با متغیر تصادفی آزمایش دوم آورده شده است.
        <br />
        در این کد، آزمایش گفته شده را ۱۰۰۰ بار تکرار کرده و به ازای هر بار مقدار متغیرهای تصادفی گفته شده را حساب می‌کنیم. نمودار فراوانی هر یک از متغیرهای تصادفی نیز رسم شده‌اند.
	</font>
</div>

```r
library(ggplot2)
a <- floor(runif(1000, min=0, max=6)) + 1
b <- floor(runif(1000, min=0, max=6)) + 1
X_1 <- a + b
X_2 <- (a == 6) + (b == 6)
X_3 <- b ^ 2
ggplot() + aes(X_1)+ geom_bar()
ggplot() + aes(X_2)+ geom_bar()
ggplot() + aes(X_3)+ geom_bar()
```

	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
محاسبهٔ احتمال با استفاده از متغیر تصادفی
        </font>
		<p></p>
		<hr>
        آزمایش ۳ بار پرتاب یک سکهٔ سالم را در نظر بگیرید. فضای حالات ممکن بدین گونه است:
        <br/>
        $$\Omega=\{HHH, HHT, \dots, TTH, TTT\}$$
        حال فرض کنید متغیر تصادفی
        $X$
        را، تعداد شیرهای واقعهٔ
        $\omega$ تعریف کنیم.
        <br/>
        مقدار این متغیر تصادفی به این صورت خواهد بود:
        </font>
        </div>
<table style="width:100%">
  <tr>
    <th>$\omega$</th>
    <th>HHH</th> 
    <th>HHT</th> 
    <th>HTH</th> 
    <th>HTT</th> 
    <th>THH</th> 
    <th>THT</th> 
    <th>TTH</th> 
    <th>TTT</th> 
  </tr>
  <tr>
    <td>$X(\omega)$</td>
    <td>3</td>
    <td>2</td>
    <td>2</td>
    <td>1</td>
    <td>2</td>
    <td>1</td>
    <td>1</td>
    <td>0</td>
  </tr>
</table>
<div id="sec_probability2" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
حال در این مثال احتمال این که $X(\omega)\le1$
باشد چیست؟
        <br/>
        احتمال روی همهٔ واقعه‌ها تعریف می‌شود. احتمال $X(\omega)\le1$ یعنی احتمال مجموعهٔ آن $\omega$هایی که برای آن‌ها $X(\omega)\le1$ باشد. یعنی:
$$P\{X\le1\}=P\{\omega:X(\omega)\le1\}=P\{HTT, THT, TTH, TTT\}=\frac{4}{8}$$
دقت کنید که معمولا به طور خلاصه می‌نویسیم: $\{X\le x\}$، ولی این مجموعه‌ای از اعداد نیست، بلکه یعنی: $\{\omega:X(\omega)\le x\}$.
<br/>
حال مثلا اگر احتمال $\{X\le x\}$ را برای هر $X$ بدانیم، احتمال همهٔ واقعه‌های مورد نظر را خواهیم داشت و دیگر لزومی به دانستن فضای احتمال $\Omega$ نیست.
	</font>
</div>

در قسمت‌های قبلی درس با احتمال و متغیرهای تصادفی آشنا شدیم  و در مسائل، احتمال وقوع رخدادهای مختلف را به صورت جداگانه حساب کردیم. حال سوالی که برای ذهن ایجاد می‌شود این است که
چگونه با دانستن احتمال وقوع رویدادهای مختلف و با ترکیب این اطلاعات با یک‌دیگر، می‌توان شهودی نسبت به نتیجه یک آزمایش تصادفی به دست آورد؟
        <br/>
        یکی از ساده‌ترین جواب‌ها به سوال بالا استفاده از مفهوم امید ریاضی است. به عبارت ساده و نادقیق، امید ریاضی میانگین وزن‌دار بر روی متغیر تصادفی است.
        <br/><br />
    </font>

<div id="sec_definition" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
تعریف امید ریاضی
        </font>
		<p></p>
		<hr>
برای متغیر تصادفی گسسته
        $X$
امید ریاضی به صورت زیر تعریف می‌شود:
$$\mathop{\mathbb{E}}[X] = \sum_{x:p(x)>0}{xp(x)}$$
که در آن
p
تابع جرم احتمال متغیر تصادفی
$X$
است.
به عبارت دیگر تعریف بالا معادل این است که بر روی تمام حالت‌های ممکن متغیر
$X$
میانگین وزن‌دار بر حسب میزان محتمل بوده‌شان گرفته شود.
<br/>
	</font>
</div>

<p></p>
<br />
<div id="sec_sum_example" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مثال حالت گسسته
        </font>
		<p></p>
		<hr>
فرض کنید که یک تاس سالم با ۶ وجه داریم. اگر
  متغیر تصادفی
$X$
را برابر عدد خروجی پرتاب یک بار تاس تعریف کنیم،
امید ریاضی آن چند است؟
$$\mathop{\mathbb{E}}[X] = \sum_{x:p(x)>0}{xp(x)}$$
$$
\mathop{\mathbb{E}}[X] = 1(\frac{1}{6}) + 2(\frac{1}{6}) + 3(\frac{1}{6}) + 4(\frac{1}{6}) + 5(\frac{1}{6}) + 6(\frac{1}{6}) = \frac{7}{2}
$$

</br>
برای مثالی دیگر میتوانیم به سکه اشاره کنیم.
به طور مثال، متغیر تصادفی
$X$
مقدار
۱
را در صورتی که نتیجه پرتاب سکه شیر شود، خروجی می‌دهد و در غیر این‌صورت مقدار آن
صفر
خواهد بود.     
حال اگر سکه متقارن باشد برای محاسبه میانگین وزن‌دار یا همان امید ریاضی در این حالت خواهیم داشت:
$$
\frac{1}{2}\times 1 + \frac{1}{2}\times 0 = \frac{1}{2}
$$
        
</font>
</div>

<div id="sec_function" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
امید ریاضی تابعی از متغیر تصادفی
        </font>
		<p></p>
		<hr>
        تابعی از یک متغیر تصادفی خود یک متغیر تصادفی است.
 در برخی از کاربردها نیاز به محاسبه امید ریاضی تابعی از یک متغیر تصادفی است. به طور مثال محاسبه‌ی امید ریاضی مربع یک متغیر تصادفی از مواردی است که در درس به زودی با آن مواجه خواهید شد.
<br/>
نکته‌ای که باید به آن توجه کرد این است که امید ریاضی تابعی از یک متغیر لزوماً برابر با اعمال آن تابع بر روی امید ریاضی نیست.
<br/>
به طور مثال فرض کنید متغیر تصادفی
$X$
دو مقدار
۰
و
۲
را با احتمال برابر بتواند اختیار کند. داریم:
$$
\mathop{\mathbb{E}}[X] = 0 \times \frac{1}{2} + 2 \times \frac{1}{2} = 1
$$
این در حالی است که امید ریاضی متغیر تصادفی
$Y = g(X) = X^2$
برابر است با:
$$
\mathop{\mathbb{E}}[Y] = 0^2 \times \frac{1}{2} + 2^2 \times \frac{1}{2} = 2
$$
در نتیجه
$\mathop{\mathbb{E}}[X^2] = 2 \neq (\mathop{\mathbb{E}}[X])^2 = 1^2$
و در حالت کلی نتیجه می‌شود که:
$$
g(\mathop{\mathbb{E}}[X]) \neq \mathop{\mathbb{E}}[g(X)]
$$
<br/>
در حالت کلی امید ریاضی تابع
$g$
از متغیر تصادفی
X
برابر است با
$$
\mathop{\mathbb{E}}[g(X)] = \sum_{i}{g(x_i)p(x_i)}
$$
که در آن
به ازای
$i \geq 1$،
$x_i$
ها
همه مقادیری هستند که
$X$
می‌تواند اختیار کند.
<br/>
برای اثبات رابطه بالا فرض می‌کنیم که به ازای
$i \geq 1$،
های مختلف،
مقادیر
$y_j$
ها که
$j \geq 1$
برابر با مقادیر متفاوت
$g(x_i)$
ها باشند. سپس از طریق دسته‌بندی همه
$i$
هایی که
$g(x_i)$
یکسان است،
رابطه ثابت می‌گردد.
        $$
        \begin{aligned}
        \sum_{i}{g(x_i)p(x_i)} &= \sum_{j}\sum_{i:g(x_i)=y_j}{g(x_i)p(x_i)}\\
        &= \sum_{j}\sum_{i:g(x_i)=y_j}{y_jp(x_i)}\\
        &= \sum_{j}y_j\sum_{i:g(x_i)=y_j}{p(x_i)}\\
        &= \sum_{j}y_jp\{g(X)=y_j\}\\
        &= \mathop{\mathbb{E}}[g(X)]
        \end{aligned}
        $$
<br/>
	</font>
</div>

<p></p>

<div id="sec_definition_example_discrete" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مثال
        </font>
		<p></p>
		<hr>
تکه چوبی به طول ۱ واحد داریم.
فرض کنید یک سر این چوب منطبق بر عدد
0
محور افقی و سر دیگر آن منطبق بر عدد
1
محور می‌باشد.
        این تکه چوب را از نقطه
$\frac{1}{2^U}$
می‌شکنیم که
U
متغیر تصادفی‌ای با تابع جرم احتمال به صورت زیر است
:
$$
P_U(u) = \begin{cases}
       \frac{1}{2^u} &\quad u\in \mathop{\mathbb{N}}\\
       0 &\quad\text{otherwise} \\ 
     \end{cases}
$$
عدد ثابت
$p$
که
$p \in \mathop{\mathbb{N}}$
است،
به ما داده شده است.
امید ریاضی طول قطعه چوبی را که شامل نقطه
$\frac{1}{2^p}$
بر روی محور است،
به دست آورید.
به شکل زیر توجه کنید.
<br/>
<img src="figs/wood.png" style="float:center; width:35%"/>
<br />
در حالت تساوی دو عدد
$p$
و
$u$
قطعه چوبی را که یک انتهای آن عدد صفر محور است،
در نظر بگیرید.
<br/>
<br/>
<br/>
تابع
$L_p(U)$
از متغیر تصادفی
U
را تعریف می‌کنیم. این تابع خود یک متغیر تصادفی خواهد بود و امید ریاضی آن پاسخ به سوال است
$$
L_p(U) = \begin{cases}
       \frac{1}{2^U} &\quad U \leq p\\
       1 - \frac{1}{2^U} &\quad U > p \\ 
     \end{cases}
$$
حال برای به دست آوردن امید ریاضی این تابع داریم:
$$
\begin{aligned}
\mathop{\mathbb{E}}[L_p(U)] &= \sum_{u=1}^{\infty}{L_p(u)P_U(U=u)} \\
&= \sum_{u=1}^{p}{(\frac{1}{2^u}).P_U(U=u)} + \sum_{u=p+1}^{\infty}{(1 - \frac{1}{2^u}).P_U(U=u)}\\
&= \sum_{u=1}^{p}{(\frac{1}{2^u})(\frac{1}{2^u})} + \sum_{u=p+1}^{\infty}{(1 - \frac{1}{2^u})(\frac{1}{2^u})}\\
&= \frac{1}{3}(1 - \frac{1}{4^{p+1}}) + \frac{1}{2^p} - \frac{1}{3}(\frac{1}{4^p})
\end{aligned}
$$
	</font>
</div>

<div id="sec_sum" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
خطی بودن امید ریاضی و جمع متغیرهای تصادفی
        </font>
		<p></p>
		<hr>
.یکی از ویژگی‌های مهم امید ریاضی خطی بودن آن است
$$
\mathop{\mathbb{E}}[aX + b] = a\mathop{\mathbb{E}}[X] + b
$$
اثبات رابطه بالا به کمک قسمت قبل به دست می‌آید.
برای این کار تابع زیر را تعریف می‌کنیم:
$$
g(X) = aX + b
$$
سپس با توجه به قسمت قبل داریم:
$$
\begin{aligned}
\mathop{\mathbb{E}}[aX + b] &= \mathop{\mathbb{E}}[g(X)] \\
&= \sum_{i}{g(x_i)p(x_i)} \\
&= \sum_{i}{(ax_i + b)p(x_i)} \\
&= \sum_{i}{axp(x_i)} + b\sum_{i}{p(x_i)} \\
&= a\sum_{i}{xp(x_i)} + b.1 \\
&= a\mathop{\mathbb{E}}[X] + b
\end{aligned}
$$
 به طور خاص استفاده از این ویژگی و توابع نشانگر (که در ادامه این جلسه معرفی می‌شوند)
می‌تواند حل برخی از مسائل را بسیار ساده‌تر کند.
<br/>
<br/>
دیدیم که ضرب عدد ثابت در متغیر تصادفی یا جمع آن با عدد ثابت،
مقدار امید ریاضی را به همان میزان جابجا کرد.
حال سوال بعدی این است که اگر به جای عدد ثابت،
متغیر تصادفی را یک متغیر تصادفی دیگر جمع کنیم،
چه اتفاقی برای امید ریاضی آن‌ها می‌افتد؟
به عبارت دقیق تر اگر
$Z = X + Y$
باشد،
آیا
گزاره
$$
\mathop{\mathbb{E}}[Z] = \mathop{\mathbb{E}}[X + Y] = \mathop{\mathbb{E}}[X] + \mathop{\mathbb{E}}[Y]
$$
برقرار است؟
<br/>
<br/>
پاسخ مثبت است. اثبات به کمک استفاده از توزیع مشترک دو متغیر
$X$
و
$Y$
(در جلسات آتی توزیع مشترک را یاد می‌گیریم):
$$
\begin{aligned}
\mathop{\mathbb{E}}[Z] &= \mathop{\mathbb{E}}[(X + Y)]\\
&= \int_x\int_y(x+y)f_{XY}(x,y)dydx \\
&= \int_x\int_yxf_{XY}(x,y)dydx + \int_y\int_xyf_{XY}(x,y)dxdy \\
&= \int_xxf_{X}(x)dx + \int_yyf_{Y}(y)dy\\
&= \mathop{\mathbb{E}}[X] + \mathop{\mathbb{E}}[Y]
\end{aligned}
$$
همان طور که مشاهده می‌کنید در اثبات بالا در هیچ جا از فرض مستقل بودن یا نبودن دو متغیر تصادفی نسبت به یکدیگر استفاده نشده است.
یعنی این گزاره برای
<strong>
هر دو متغیر تصادفی‌ای، چه مستقل چه وابسته،
</strong>
برقرار است.
<br/>
<br/>
اثبات گفته شده در بالا کمی نقص دارد. در آن به نکته‌ای اشاره نشده است. آیا می‌توانید آن را پیدا کنید؟ 
	</font>
</div>

<p></p>
<br />
<div id="sec_indicator_example" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مثال
        </font>
		<p></p>
		<hr>
$n$
نفر
در یک اتاق کلاه‌هایشان را به هوا پرتاب می‌کنند. کلاه‌ها با یکدیگر جابه‌جا شده و سپس هر نفر به صورت تصادفی یک کلاه انتخاب می‌کند. امید ریاضی تعداد افرادی که همان کلاه اولیه خودشان را انتخاب می‌کنند،
چه مقدار است؟
<br/>
<br/>
مسئله را به کمک توابع نشانگر می‌توان ساده کرد.
توابع نشانگر
$
X_i ( 1 \leq i \leq n)
$
را به نحو زیر تعریف می‌کنیم:
$$
X_i = \begin{cases}
       1 &\quad \text{if the i-th person selects his own hat}\\
       0 &\quad \text{otherwise}\\ 
     \end{cases}
$$
امید ریاضی
متغیرهای نشانگر برابر با احتمال وقوع آن‌هاست. بنابراین:
$$
\mathop{\mathbb{E}}[X_i] = \frac{1}{n}
$$
دلیل آن این است که هر فرد طبق صورت سوال به صورت تصادفی و با احتمال برابر کلاه خود را پس از پرتاب به هوا انتخاب می‌کند.
<br/>
حال اگر متغیر مورد سوال را
$X$
بنامیم،
داریم:
$$
X = X_1 + \dots + X_n
$$
در نتیجه:
$$
\begin{aligned}
\mathop{\mathbb{E}}[X] &= \mathop{\mathbb{E}}[X_1 + \dots + X_n] \\
&=\mathop{\mathbb{E}}[X_1] + \dots + \mathop{\mathbb{E}}[X_n] \\
&= \frac{1}{n} + \dots + \frac{1}{n} \\
&= n(\frac{1}{n}) = 1
\end{aligned}
$$
	</font>
</div>

<div id="sec_indicator_example2" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مثال
        </font>
		<p></p>
		<hr>
تعداد
$h$
شکارچی
و
$b$
پرنده داریم.
هنگامی که دسته پرندگان از بالای سر شکارچیان پرواز
می‌کنند،
هر کدام از شکارچیان، مستقل از دیگری، یکی از پرندگان را به صورت تصادفی انتخاب کرده و به سمت آن شلیک می‌کند. تیر با احتمال
$p$
به پرنده اصابت می‌کند.
<br/>
امید ریاضی تعداد پرندگانی را که بدون اصابت تیر به آن‌ها جان سالم به در می‌برند،
محاسبه کنید.
<br/>
<br/>
متغیر تصادفی مد نظر را
$X$
می‌نامیم.
حال این متغیر را می‌توان به تعدادی تابع نشانگر تجزیه کرد.
$$
X_i = \begin{cases}
       0 &\quad \text{if the i-th bird gets shot}\\
       1 &\quad \text{otherwise}\\ 
     \end{cases}
$$
و مشاهده می‌کنیم که رابطه زیر برقرار است:
$$
X = X_1 + \dots + X_n
$$
از طرفی برای هر کدام از
$X_i$
ها داریم:
$$
\mathop{\mathbb{E}}[X_i] = p\{X_i = 1\}
$$
که برابر احتمال تیر نخوردن پرنده
i-ام
است.
<br/>
حال احتمال این‌که شکارچی
j-ام
پرنده
i-ام
را بزند
برابر است با:
$$
\frac{1}{b} \times p = \frac{p}{b}
$$
در نتیجه
احتمال
تیر نخوردن پرنده
i-ام
در مجموع
برابر است با:
$$
P\{x_i=1\} = (1 - \frac{p}{b})^h
$$
در نتیجه امید ریاضی
$X$
که جمع این متغیرهاست برابر است با:
$$
\mathop{\mathbb{E}}[X] = \sum_{i=1}^{b}{\mathop{\mathbb{E}}[X_i]} = b(1 - \frac{p}{b})^h
$$
دقت کنید
در این مثال هم متغیرهای نشانگر نسبت به هم مستقل نیستند. ولی چون خطی بودن امید ریاضی،
نیازمند فرض استقلال نیست، می‌توان امید ریاضی جواب را بر حسب جمع امید ریاضی تعدادی متغیر نشانگر نامستقل محاسبه کرد. 
</font>
</div>
