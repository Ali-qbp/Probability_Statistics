### Stat98

<div dir=rtl>
<ul>
    <li>
        در بخش واریانس، قالب مطلوب جزوه رعایت نشده است. لطفا مبحث را با مثال آغاز کنید. مثالی که در ادامه آمده است مناسب است. یعنی مثلا به عنوان کنجکاوی بپرسید که میانگین هر سه دسته یک عدد است اما چه چیزی میان آن‌ها تفاوت بوجود می‌آورد.سپس اصل درس واریانس را بگویید و سپس  به عنوان بازگشت به مسئله واریانس هر سه دسته را محاسبه کنید.
    </li>
    <li>
        بعد از نوشتن عبارت ریاضی واریانس، به نظرم اگر کمی راجع به اینکه این عبارت ریاضی با شهود و تعریفی که از واریانس می‌خواهیم سازگار است توضیح دهید مناسب است.
    </li>
    <li>
        در بخش انحراف معیار ، تنها فرمول جبری آن گفته شده است، اما به نظرم بهتر است توضیح داده شود چرا هم واریانس تعریف می‌کنیم هم انحراف معیار. از لینک‌های زیر می‌توانید استفاده کنید.
        <br>
        <a href="https://www.quora.com/If-we-have-Standard-Deviation-why-do-we-need-Variance-Or-vice-versa">لینک اول </a>
        <br>
                 <a href="https://stats.stackexchange.com/questions/35123/whats-the-difference-between-variance-and-standard-deviation">لینک دوم </a>
    </li>
    <li>
       قبل از آنکه جزوه ناگهان از واریانس وارد توزیع برنولی شود، یک بخش درست کنید که اصلا به صورت شهودی توضیح دهد که یک توزیع چیست . و سپس به سراغ معرفی اولین توزیع‌ها بروید. 
    </li>
    <li>
        در توضیح توزیع برنولی ، 
        p 
        معرفی می‌شود که پارامتر توزیع است. برای همین یک توضیح در مورد اینکه پارامتر یک توزیع چیست و آنرا چگونه نمایش می‌دهند بنویسید. می‌توانید از این
        <a href"https://www.statlect.com/glossary/parameter">لینک </a>
        استفاده کنید.
    </li>
    <li>
        در محاسبه واریانس برنولی از مفهوم تابعی از متغیر تصادفی استفاده شده است که هنوز درس نداده شده است. لطفا پاک کنید.
        همچنین در محاسبه واریانس باینومیال هم از واریانس مجموع متغیرهای تصادفی استفاده شده است که هنوز درس نداده شده است. به نظرم بهتر است که یک کنجکاوی بگذارید که واریانس آن‌ها چگونه محاسبه می‌شود و همجنین اشاره کنید که در آینده ابزار محاسبه واریانس را یاد خواهند گرفت.
    </li>
    <li>
        به عنوان یک مقدمه کوتاه از جمع چند متغیر تصادفی، در جزوه نشان دهید که جمع دو متغیر تصادفی باینومیال، خود یک متغیر تصادفی باینومیال است. برای اثبات آن نیز هم نشان دهید که شهودی درست است هم اثبات جبری کنید(مثلا اثباتی که از قضیه واندرموند کمک می‌گیرد)
    </li>
    <li>
        در آخر جزوه یک کنجکاوی وجود دارد که ارتباط میان توزیع باینومیال و نرمال را نشان می‌دهد. یا این قسمت را پاک کنید یا بگویید که در آینده با یادگیری مباحث بیشتر می‌توانید دوباره برگردید و به این سوال‌ها فکر کنید.
    </li>
    <li>
        برای هر توزیع بهتر است یک شکل هم موجود باشد. می‌توانید از 
        R 
        یا از سایت
        seeing theory
        که در دراپ‌باکس و گروه درس موجوداست استفاده کنید.
    </li>
    <li>
        در بخش خلاصه و نتیجه‌گیری اگر فرمول خاص هر بخش هم اضافه شود و از این حالت تیتر وار خارج شود بهتر است. البته حجم آن باید خلاصه باقی بماند
    </li>
</ul>
