<div id="sec_intro_variance" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مقدمه
        </font>
		<p></p>
		<hr>
در بخش قبل با امید ریاضی یا مقدار میانگین آشنا شدیم و دیدیم که به کمک این کمیت مهم می‌توان شهود نسبت به توزیع
متغیر تصادفی به دست آورد.        
<br/>
در این قسمت کمیت دیگری به نام واریانس را معرفی می‌کنیم. امید ریاضی با وجود این‌که در مورد مقدار میانگین اطلاع می‌داد، در مورد
میزان
<strong>
پراکندگی
</strong>
خروجی‌های یک آزمایش تصادفی
یا متغیر تصادفی
اطلاع خاصی نمی‌داد.
<br/>
برای درک بهتر این موضوع که منظور از پراکندگی چیست، به مثال زیر توجه کنید:
<br/>
<br/>
فشار خون در سه گروه مجزا
پنج نفره از افراد را مورد بررسی قرار دادیم که به شرح جدول زیر است:

    </th>

    </th>
    <th>
        گروه Z
    </th>
</tr>
<tr>
    <td>14</td>
    <td>11</td>
    <td>10</td>
</tr>
<tr>

حال از هر گروه به تصادف یک فرد را انتخاب می‌کنیم و متغیرهای تصادفی
$X$،
$Y$
و
$Z$

<br/>
در این مثال
مقدار میانگین هر سه متغیر برابر ۱۰ است.
ولی نکته قابل توجه این است که هنگامی که از
گروه
$Z$
کسی را انتخاب می‌کنیم،
اطمینان داریم که فشار خون وی ۱۰ است.

$Y$
هم کسی انتخاب می‌شود، فشار خون وی به صورت شهودی
در 
حوالی
عدد ۱۰ یا همان
میانگین
است، چون
پراکندگی
داده ها کم است.
ولی برای گروه
$X$
دیگر نمی‌توان چنین حرفی زد. یعنی میانگین داده‌ها ۱۰ است،
ولی الزاماً
مقدار متغیر تصادفی ۱۰ یا حوالی ۱۰ نخواهد بود، چون داده‌ها
پراکنده‌تر
هستند.

<div id="sec_def_variance" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
تعریف واریانس
        </font>
		<p></p>
		<hr>
       واریانس یک متغیر تصادفی گسسته
را که معیاری از پراکندگی داده خروجی‌های ممکن آن است،
به شرح زیر تعریف می‌کنیم:
$$
Var(X) = \mathop{\mathbb{E}}[(X - \mathop{\mathbb{E}[X]})^2]
$$
یا در برخی نمادگذاری‌های دیگر به صورت زیر تعریف می‌شود:
$$
Var(X) = \mathop{\mathbb{E}}[(X - \mu)^2]
$$
که در آن
$\mathop{\mathbb{E}}[X] = \mu$
است.
<br/>
با ساده کردن عبارت بالا داریم:
$$
\begin{aligned}
Var(X) &= \mathop{\mathbb{E}}[(X - \mu)^2] \\
&= \sum_{x}(x-\mu)^2p(x) \\
&= \sum_{x}(x^2+\mu^2-2\mu{x})p(x) \\
&= \sum_{x}x^2p(x) + \sum_{x}\mu^2p(x) -2\mu\sum_{x}xp(x)\\
&= \mathop{\mathbb{E}}[X^2] + \mu^2.1 -2\mu.\mu \\
&= \mathop{\mathbb{E}}[X^2] - \mu^2 \\
Var(X) &= \mathop{\mathbb{E}}[X^2] -  (\mathop{\mathbb{E}}[X])^2\\
\end{aligned}
$$
<br/>
همچنین به جذر واریانس، انحراف معیار می‌گویند.
$$
SD(X) = \sqrt{Var(X)}
$$
    </font>
</div>

<p></p>
<br />
<div id="sec_prop_variance" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
خواص واریانس
        </font>
		<p></p>
		<hr>
واریانس معیاری از پراکندگی خروجی‌های مختلف یک متغیر تصادفی است.
بنابراین انتقال یک متغیر تصادفی به اندازه یک عدد ثابت نباید اثری در آن داشته باشد چون پراکندگی نسبی خروجی‌های متغیر تصادفی انتقال‌یافته، نسبت به یکدیگر عوض نمی‌شود.
<br/>
همچنین واریانس با توان دوم مرتبط است. بنابراین دور از  
انتظار نیست که ضرب یک متغیر تصادفی در مقدار
ثابت رابطه توان دوم با واریانس متغیر اولیه داشته باشد.
<br/>
دو خاصیت بالا را در رابطه زیر می‌یابیم:
$$
\begin{aligned}
Var(aX + b) &= \mathop{\mathbb{E}}[(aX+b-a\mu-b)^2] \\
&= \mathop{\mathbb{E}}[a^2(X-\mu)^2] \\
&= a^2\mathop{\mathbb{E}}[(X-\mu)^2] \\
&= a^2Var(X)
\end{aligned}
$$
    </font>
</div>

<p></p>
<br />
<div id="sec_example_variance" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مثال
        </font>
		<p></p>
		<hr>
در جعبه‌ای
تعداد ۵ گوی آبی و ۵ گوی قرمز است.
دو گوی را به تصادف بدون جایگذاری از جعبه در می‌آوریم.
اگر دو گوی هم‌رنگ باشند
۱.۱
دلار
به دست می‌آوریم و اگر ناهم‌رنگ باشند ۱ دلار می‌بازیم.
اگر متغیر تصادفی مقدار پولی که
به دست می‌آوریم
،
$X$
باشد،
$Var(X)$
چند است؟
<br/>
$$
\mathop{\mathbb{E}}[X] = +1.1 \times (\frac{{5\choose2}+{5\choose2}}{{10\choose2}}) + (-1) \times (1-\frac{{5\choose2}+{5\choose2}}{{10\choose2}}) = -\frac{1}{15}
$$
$$
\mathop{\mathbb{E}}[X^2] = +1.1^2 \times (\frac{{5\choose2}+{5\choose2}}{{10\choose2}}) + (-1)^2 \times (1-\frac{{5\choose2}+{5\choose2}}{{10\choose2}}) = \frac{82}{75}
$$
$$
\begin{aligned}
Var(X) &= \mathop{\mathbb{E}}[X^2] - \mathop{\mathbb{E}}[X]^2\\
&= \frac{82}{75} - (\frac{-1}{15})^2\\
&= 1.0889 \\
\end{aligned}
$$
    </font>
</div>

<div id="sec_bernoulli" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
توزیع گسسته برنولی
        </font>
		<p></p>
		<hr>
        توزیع برنولی، توزیعی گسسته است که نام آن از نام دانشمند سوئیسی ژاکوب برنولی گرفته شده‌است. توزیع برنولی یک توزیع گسسته است که مقادیر یک (در صورت موفقیت آزمایش ) و صفر را (در صورت شکست) می‌گیرد. احتمال موفقیت آزمایش برابر p است و احتمال شکست آن برابر q=1-p است. به عبارت دیگر توزیع برنولی به نوعی مربوط به یک سوال بله خیر است یعنی خروجی تنها دو حالت دارد که احتمال بله یا مثبت بودن پاسخ p و احتمال نه یا منفی بودن پاسخ $q= 1-p$ است.
        <br>
        اگر X متغیر تصادفی رندوم با توزیع برنولی باشد. احتمال آن به شکل زیر تعریف می‌شود:
        \[
        Pr(X=1)=p=1-Pr(X=0)=1-q
        \]<br>
    نوع دیگری که مسئله‌های برنولی مطرح می‌شوند ترکیبی از تکرار یک کار است که هر بار آن فقط موفقیت و یا شکست دارد. برای مثال مسئله زیر را در نظر بگیرید:
    <br>
    <b>
    مسئله: 
    </b>
    یک سکه را ۵ بار پرتاب می‌کنیم به چه احتمالی (شیر، خط، شیر، شیر، خط) به ترتیب از راست به چپ ظاهر می‌شود.
    <br>
    <b>
    راه حل: 
    </b>
    در این مسئله در پرتاب اول شیر آمدن موفقیت است و در پرتاب دوم خط آمدن و به همین ترتیب. اما در همه پرتاب‌ها امکان موفقیت $\frac{1}{2}$ و امکان شکست نیز $\frac{1}{2}$ است.
    <br>
    $Pr = \frac{1}{2} \times \frac{1}{2} \times \frac{1}{2} \times \frac{1}{2} \times \frac{1}{2} = \frac{1}{2 ^ 5}$<br>
    میانگین در توزیع گسسته برنولی برابر با امید ریاضی است و به شکل زیر محاسبه می‌گردد:
    <br>
    $E[x] = Pr(X=1) \times 1 + Pr(X=0) \times 0 = p \times 1 + q \times 0 = p$<br>
    واریانس توزیع گسسته برنولی نیز به شکل زیر محاسبه می‌شود:
    <br>
    $E[X^2]=Pr(X=1) \times 1^2 + Pr(X=0) \times 0^2 = p$<br>
	</font>
</div>

<div id="sec_problem_binomial" style="direction:rtl;line-height:300%;">

صورت مسئله
        </font>
		<p></p>
		<hr>
قرار است در هر قسمت از مسابقه علی ۱۰ نفر مسابقه بدهند که هر کدام ۲۰ سوال پیش رو دارند و در صورتی که از ۲۰ سوال ۱۸ یا بیشتر جواب درست داشته باشند ۵ میلیون برنده می شوند!
سوالات هیچ ربطی به هم ندارند و کاملا از هم مستقل اند.
سوالات چهارگزینه‌ای هستند و علی امیدوار است آنقدر سخت باشند که کسی روی زمین، جواب صحیح را نداند. برای همین محاسبات او با این پیش فرض است.
<br>
برای حل این مسئله احتمالاتی، علی نیاز دارد تا توزیع مربوطه را بداند و امید ریاضی آن را بدست آورد.

<div id="sec_binom_binomial" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
توزیع دو جمله ای
        </font>
		<p></p>
		<hr>
در یک آزمایش هنگامی که رخداد های مستقل  اما مشابه داریم، آن را
        <b>دنباله ای از آزمایشات مستقل</b> 
        می نامیم. وقتی که هر رخداد فقط دو حالت ممکن داشته باشد، آن را 
        <b>آزمایش برنولی</b>
        می نامیم. مثلا، در انداختن یک سکه فقط دو حالت "شیر" یا "خط" داریم.
<br>
آزمایشی را در نظر بگیرید که در آن $n$ بار یک سکه ناهمگن را می اندازیم. احتمال اینکه سکه رو بیاید برابر $p$ است، که عددی بین صفر و یک است. واضح است که این رویداد ها از هم مستقل اند. زیرا هر بار که این سکه را می اندازیم مستقل از نتایج قبل، با احتمال $p$ شیر می آید و  با احتمال $1-p$ خط.
<br>
حالات ممکن و احتمال هر رویداد در شرایطی که $n=3$ باشد، در تصویر زیر نشان داده شده است.  
<br>
<img src="figs/dbinom.png" style="float:center; width:50%"/>
<br>
با دانستن اینکه این رویدادها از هم مستقل اند می توانیم احتمالات را در هم ضرب کنیم و احتمال نهایی در هر کدام از حالات را بدست آوریم.  برای هر رخدادی که در آن $k$ بار شیر و $3-k$ بار خط آمده، عبارت 
$p^{k}(1-p)^{3-k}$
را میبینیم. در حالت کلی بجای 3، $n$ میگذاریم و عبارت
$p^{k}(1-p)^{n-k}$
 بدست می آید.
<br>
حال احتمال رویداد زیر را در نظر بگیرید:
<br>
$$p(k) = P(k\:heads\:in\:n\: tosses)$$
<br>
روی دادن $k$ بار شیر در $n$ پرتاب، در هر کدام از ترتیب های آمدن  شیر و خط، یکسان است و بر اساس فرمولی که در بالا بدست آمده بدست می آید و در همه ترتیب ها این احتمال ها برابر است. کافی است تعداد حالات مختلف $k$ بار شیر آمدن را در $n$ پرتاب حساب کنیم که برابر است با:
${n \choose k}$
<br>
و فرمول کلی می شود:   
$$P(k)={n \choose k}p^{k}(1-p)^{n-k}$$
<br>
اعداد
 ${n \choose k}$
 ضرایب دوجمله ای نامیده می شوند.
<br>
متغیر تصادفی $X$ که  به صورت زیر تعریف شود از توزیع دو جمله ای پیروی می کند: 
<br>
$$ X = occurrences\: of\: event\: A\: in\: n \: trials $$
<br>
$$X \sim binomial(n,p)$$
<br>
	</font>
</div>



<p></p>
<br />
<div id="sec_excpectectation_binomial" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
امیدریاضی و واریانس
        </font>
		<p></p>
		<hr>
        امیدریاضی:
        <br>
        \begin{aligned} \operatorname {E}[X]=np \end{aligned}
        <br>
 \begin{aligned}\mu &=\sum _{k=0}^{n}k{\binom {n}{k}}p^{k}(1-p)^{n-k}\\&=np\sum _{k=0}^{n}k{\frac {(n-1)!}{(n-k)!k!}}p^{k-1}(1-p)^{(n-1)-(k-1)}\\&=np\sum _{k=1}^{n}{\frac {(n-1)!}{((n-1)-(k-1))!(k-1)!}}p^{k-1}(1-p)^{(n-1)-(k-1)}\\&=np\sum _{k=1}^{n}{\binom {n-1}{k-1}}p^{k-1}(1-p)^{(n-1)-(k-1)}\\&=np\sum _{\ell =0}^{n-1}{\binom {n-1}{\ell }}p^{\ell }(1-p)^{(n-1)-\ell }&&{\text{with }}\ell :=k-1\\&=np\sum _{\ell =0}^{m}{\binom {m}{\ell }}p^{\ell }(1-p)^{m-\ell }&&{\text{with }}m:=n-1\\&=np(p+(1-p))^{m}\\&=np\end{aligned}
<br>
واریانس:
<br>
 \begin{aligned} \operatorname {Var} (X)=np(1-p)
 \end{aligned}
<br>
\begin{aligned}
\operatorname {Var} (X)=\operatorname {Var} (X_{1}+\cdots +X_{n})=\operatorname {Var} (X_{1})+\cdots +\operatorname {Var} (X_{n})=n\operatorname {Var} (X_{1})=np(1-p)
\end{aligned}
</font>
</div>

<div id="sec_jam_again_binomial" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
بازگشت به مسئله مسابقه
        </font>
		<p></p>
		<hr>
واضحا مساله علی دو توزیع دوجمله‌ای است. در هر سوال با فرض اینکه شخص هیچ دانشی نسبت به سوال ندارد، و اینکه ۴ گزینه وجود دارد، احتمال موفقیت ۰.۲۵ است.
توزیع اول برای هر فرد و سوالاتش برقرار است. یعنی ۲۰ بار برای هر فرد آزمایش یکسانی تکرار می شود که احتمال موفقیت ۰.۲۵ است. حالا در ۲۰ سوال احتمال اینکه یک نفر بیشتر مساوی ۱۸ سوال پاسخ درست بدهد برابر است  با احتمال اینکه: دقیقا ۱۸ سوال + دقیقا ۱۹ + دقیقا ۲۰ سوال پاسخ درست بدهد.  
<br>
$$p(success\: for\: an\: individual) = p(18) + p(19) + p(20) ={20 \choose 18}(0.25)^{18}(1-0.25)^{20-18} + {20 \choose 19}(0.25)^{19}(1-0.25)^{20-19} + {20 \choose 20}(0.25)^{20}(1-0.25)^{20-20} = 1.55524 \times 10^{-9} + 5.457 \times 10^{-11} + 9.1 \times 10^{-13} = 1.61072 \times 10^{-9} $$
<br>
در مرحله بعد ۱۰ نفر داریم که هر نفر احتمال موفقیتی برابر مقدار بدست آمده در رابطه بالا دارد. بطور متوسط چقدر باید پول بدهد؟  باید امید ریاضی  توزیع جدید را بیابیم.
<br>
$$X \sim binomial(n=10,p=1.61072 \times 10^{-9})$$
<br>
$$E = np = 1.61072 \times 10^{-8}$$
<br>
این در واقع امید ریاضی تعداد افراد موفق در هر قسمت از مسابقه است. علی به ازای هر نفر ۵ میلیون باید بدهد:
<br><br>
$$Ali’s\:expense = E(X) * 5000000  = 0.080536\: toman$$
<br>
علی در هر مسابقه بطور متوسط در هر قسمت 0.8 ریال  باید پول بدهد. اگرچه خیلی ارزان در می آید، اما نه کسی شرکت خواهد کرد و نه کسی آن را می بیند!
</font>
</div>

        <br>
        نمودار توزیع دو جمله‌ای چه شکلی است؟
        <br>
        در صورتی که n به بی نهایت  میل کند نمودار توزیع دو جمله‌ای شبیه کدام توزیع می شود؟
<br>
        رابطه این توزیع با توزیعات نرمال و پواسون چیست و چه زمانی این ارتباط مفید است؟
