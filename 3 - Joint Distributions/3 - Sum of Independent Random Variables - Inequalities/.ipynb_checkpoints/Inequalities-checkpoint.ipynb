<div id="sec_intro" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مقدمه
        </font>
		<p></p>
		<hr>
آندری مارکوف که در دانشگاه پترزبورگ تحصیل می‌کرد جوان تنبلی اما متفکری بود. سوالی ذهن او را مشغول کرده بود. او می‎دانست که انسان‎ها به طور میانگین 60 سال عمر می‎کنند. او می‎خواست بداند که با چه احتمالی 300 سال عمر خواهد کرد. در همین روزها بود که استادش، چبیشف، او و دوستش امینف! را به دفتر خود فراخواند تا از آن‌ها زهرچشمی بگیرد. او نامساوی زیر را جلوی آن‌ها گذاشت و از آن‌ها خواست درستی یا نادرستی آن را اثبات کنند: 
<br/><b>
اگر $X$ یک متغیر تصادفی <u>نامنفی</u> و $a$ یک عدد مثبت دلخواه باشد، داریم:
<br/></b>
<br/>
    </font>
    <font size = 6>
    <center>
$$
P(X\geq a) \leq \frac{E[X]}{a} 
$$
    </center>
    <br/>
    </font>
    <font face="XB Zar" size=5>
        مارکوف به این نامساوی نگاه کرد و با خود فکر کرد: "اگر این نامساوی درست باشد، روشی برای محاسبه احتمال اینکه او  بیشتر از 300 سال عمر کند پیدا شده است." او کاغذی پیدا کرد و روی آن نوشت:
        $$
        P\{age \ge 300\} \le \frac{60}{300} = \frac{1}{5}
        $$
        مارکوف از این نتیجه تعجب کرد اما شهودی درباره درستی این نامساوی بدست آورد. او تا به حال کسی را ندیده بود که 300 سال عمر کند اما این حقیقت درستی نامساوی را نقض نمی‎کرد. تنها نشان می‎داد کران بالایی که نامساوی ارائه می‎دهد ممکن است فاصله زیادی تا مقدار واقعی احتمال داشته باشد.
    </font>
</div>

<p></p>
<br />
<div id="sec_problem" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
بررسی مسئله
        </font>
		<p></p>
		<hr>
آن‌ها ابتدا می‌بایستی این نامساوی را روی تعدادی متغیر تصادفی دلخواه امتحان می‌کردند تا حدس‌هایی درمورد روش اثبات یا رد آن بیابند. از آن‎جا که امینف با کتاب‌خانه ggplot2 آشنا بود،‌ تصمیم گرفتند به صورت گرافیکی این‌کار را انجام دهند. 
مارکوف کار با یک تاس را پیشنهاد داد. آن‌ها تصمیم گرفتند مقدار امید ریاضی در نامساوی را ثابت نگه دارند و متغیر‌های دیگر (واریانس) را تغییر دهند. سه تاس شش وجهی با اعداد متفاوت را در نظر گرفتند :‌
<br/>
        <ul>
            <li>
                اعداد روی تاس اول 
                $ (1, 2, 3, 4, 5, 6) $
            </li>
            <li>
                اعداد روی تاس دوم
                $ (2.25, 2.75, 3.25, 3.75, 4.25, 4.75) $
            </li>
            <li>
                اعداد روی تاس سوم
                $ (2.875, 3.125, 3.375, 3.625, 3.875, 4.125) $
            </li>
        </ul>
نمودار احتمالِ 
        $ P(X \ge a) $
        و 
        $ \frac{mean = 3.5}{a} $
        بر حسب a رسم شده است:
</div>

```r
  library(ggplot2)
  x = c(seq(1,6,1) , seq(2.25,4.75,0.5) , seq(2.875, 4.125,0.25))
  #dice1 is the left one and dice3 is the right one
  #All are 6 faced dices with mean = 3
  y = rep(c(1 , 5/6 , 4/6 ,3/6 , 2/6 , 1/6) , 3)
  #y contains P(X>=a) for each face
  facet = c(rep("dice1" , 6), rep("dice2" , 6) , rep("dice3" , 6))
  m = mean(x[1:6]) # m = 3
  f1 = function(x){
    return(m/x)
  }
  d = data.frame( a = x , y = y , facet = facet) #converting to data.frame for using in ggplot
  g = ggplot(data = d , mapping = aes(x = a, y = y, color = facet)) +geom_point() + 
    facet_grid(. ~ facet)
  g1 = g  + stat_function(fun = f1, mapping = aes(color = "mean/x")) + 
    ylim(0,2) + xlim(0,6) + ylab("P(X>=a)")
  g1 
```

<div id="sec_intro" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
    مارکوف با خودش فکر کرد از آنجا که به نظر می‌رسد با کم کردن واریانس نامساوی در یکی از نقاط به تساوی نزدیک می‌شود، اگر واریانس صفر باشد (یعنی مقدار تمام نقاط برابر باشد) نامساوی تبدیل به تساوی خواهد شد. 
    <br/>
    </font>
    <font face="XB Zar" size = 6 color=#0000FF>
        تمرین
    </font>
    <br/>
    <font face="XB Zar" size = 5>
        حدس بالا را ثابت کنید. یعنی نشان دهید اگر واریانس برابر صفر باشد، نامساوی تبدیل به تساوی خواهد شد.
    </font>
</div>

<div id="sec_math" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
    <font color=#FF7500 size=6>
اثبات نابرابری مارکوف
        </font>
        <hr>
         امینف که در آن مدت با کاغذ و قلم خود ورمیرفت،‌ ایده خلاقانه‌ای به ذهنش رسید! رو به مارکوف گفت:«ما می‌دانیم که اگر تابع f در تمام نقاطی که متغیر تصادفی X در آن تعریف شده کوچکتر مساوی تابع g باشد آنگاه امید ریاضی آن نیز بر روی X کمتر از g است. به عبارتی: 
    <br/>
    <center>
    $ \forall x : f(x) \leq g(x) \Rightarrow E[f(X)] \leq E[g(X)] $
    </center>
   حال تابع مشخصه‌ای را در نظر بگیر که برای مقادیر بزرگ تر از a‌ دلخواه، یک و مقادیر کوچک تر صفر باشد. (تابع مشخصه)
    <br/>
    از آنجا که این تابع کوچکتر مساوی تابع 
    $ \frac{x}{a}$
  در تمام نقاط غیرمنفی است، پس :
    <br/>
    <center>
    $ E[I_a(X)] \leq E[\frac{X}{a}] \Rightarrow P(X \geq a) \leq \frac{E[X]}{a}$
    <br>
    (چرا؟)
    </center>
    مارکوف داد زد:‌«‌ آفرین!» و به سرعت روی کاغذ چیز‌هایی نوشت: 
    <br/>
    $$
    \begin{align*}
    E[X] &= \int^{+\infty}_{0}xf(x) dx \\
    &= \int^{a}_{0}xf(x) dx + \int^{+\infty}_{a}xf(x)dx \\
    &\geq \int^{+\infty}_{a}xf(x)dx \\
    &\geq \int^{+\infty}_{a}af(x)dx \\
    &= aP(X \geq a) \\ \\
    &\Rightarrow P(X \geq a) \leq \frac{E[X]}{a}
    \end{align*}
    $$
    </div>

<div id="sec_gen" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
        <font color=#FF7500 size=6>
تلاشی برای تعمیم
        </font>
<hr>
    مارکوف از امینف خواست تا با تست کردن چند توزیع نرمال با میانگین یکسان، ببیند شرط نامنفی بودن متغیر تصادفی X برای این نامساوی لازم است یا خیر.
امینف کد زیر را زد:
    </font>
</div>

```r
library(ggplot2)
m = 3.5 # m = E[X]
f1 = function(x){
  return(m/x)
}
g = ggplot(data = data.frame()) + stat_function(fun = pnorm , position = "identity",
                                 mapping = aes(color = "sd = 15"), 
                                 args = list(mean = 3.5 , sd = 15 , lower.tail = F)) 
# p(X>=x) for N(3.5 , 15)
g1 = g + stat_function(fun = pnorm , position = "identity",
                                 mapping = aes(color = "sd = 5"), 
                                 args = list(mean = 3.5 , sd = 5, lower.tail = F))
# p(X>=x) for N(3.5 , 5)
g2 = g1 + stat_function(fun = pnorm , position = "identity",
                                 mapping = aes(color = "sd = 30"), 
                                 args = list(mean = 3.5 , sd = 30, lower.tail = F))
# p(X>=x) for N(3.5 , 30)
g3 = g2  + stat_function(fun = f1, mapping = aes(color = "mean/x")) + 
  ylim(0,1)+ xlim(0,30) + xlab("x") + ylab("P(X>=x)")
g3
```

<div id="sec_intro" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
    آن‌ها دیدند که با بزرگ‌تر شدن واریانس از عددی، نامساوی نقض می‌شود.
    </font>
    <br/>
    <font face="XB Zar" size=5 color=#FF00FF>
        کنجکاوی: 
    </font>
    <font face="XB Zar" size=5>
        واریانس از چه عددی برزگتر شود، نامساوی دیگر برقرار نخواهد بود؟
    </font>
    <br/>
</div>

<div style="direction:rtl;line-height:300%;">
    <font face="XB Zar" size = 6 color=#FF7500>
        نابرابری چبیشف
    </font>
    <hr/>
    <font face="XB Zar" size = 5>
    آن‌ها نتایج خود را به چبیشف نشان دادند. او که لبخندی بر لب داشت، آن‌ها را تشویق کرد و حکیمانه گفت:«نکته اصلی در نامساوی‌ای که اثباتش کردید این است که بدون دانستن pdf یک توزیع می‌توان کرانی بر روی CDF‌ آن به دست آورد.»
    از طرفی مارکوف از نامساوی‎ای که اثبات شده بود، راضی نبود. او کرانی دقیق‎تر برای احتمال عمر 300 ساله‎اش! می‎خواست. این جا بود که از استاد چبیشف پرسید: "آیا نمی‎توان این نابرابری را دقیق‎تر کرد؟"
    <br/>
        استاد چبیشف نابرابری خودش را به آن‎ها نشان داد و گفت:«
    <br/>
        اگر به جای متغیر تصادفی 
        $X$،
        $(E[X]-X)^2$
        و به جای $a$،
        $a^2$
        را قراردهیم آنگاه به دست می‎آید: 
    <br/>
    <center>
        <font size=6>
        $$
        P(|X-E[X]| \geq a) \leq \frac{Var(X)}{a^2}
        $$
        <p></p>
        </font>
    </center>
این نامساوی به نام <b>چبیشف</b>، یعنی نام من مشهور است.»
<br/>
    در ادامه استاد چبیشف از آن‎ها خواست تمرین‎های زیر را حل کنند:
    </font>
</div>

<div style="direction:rtl; line-height:300%" align=right>
    <font face="XB Zar" size=6 color=#0000FF>
        تمرین
    </font>
    <br/>
    <font face="XB Zar" size=5>
        <ol>
            <li>
                بدون استفاده از نابرابری مارکوف، نابرابری چبیشف را ثابت کنید.
            </li>
            <li>
                به مارکوف کمک کنید تا با استفاده از نابرابری چبیشف، کران بالای دقیق‎تری برای عمر 300 ساله‎اش پیدا کند.
            </li>
        </ol>
    </font>
</div>

<div id="sec_other_ine" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
        <font color=#FF7500 size=6>
دیگر نابرابری‎ها
        </font>
    <hr>
        استاد چبیشف برخی اوقات در شرایطی قرار می‎گیرد که می‎خواهد احتمالاتی از نوعِ
        $$
        P\{ X - \mu \ge a \}
        $$
        را تخمین بزند. در این نوع از احتمالات a یک مقدار مثبت است و ما احتمالا فقط میانگین و یا واریانس X را در اختیار داریم. راه اولی که برای حل این مسئله پیشنهاد شد، استفاده از نابرابری چبیشف بود.
        $$
        \begin{align*}
        P\{ X - \mu \ge a \} \le P\{ |X - \mu| \ge a \} \le \frac{\sigma^2}{a^2} && (a > 0)
        \end{align*}
        $$
    <p></p>
        اما استاد چبیشف ناراضی به نظر می‎رسید. بنابراین تلاش کرد تا نامساوی قویتری را ارائه دهد.
        <br/>
        او در ابتدا نابرابری یک طرفه خود را معرفی کرد. (این نابرابری به نابرابری یک طرفه چبیشف مشهور است.):
        <br/>
        <b>
        اگر 
        $ X $
        یک متغیر تصادفی با میانگین صفر و واریانس متناهی باشد، در آن صورت برای هر 
        $ a \gt 0 $
        داریم:
        </b>
        <font size=6>
        $$
        P\{X \ge a\} \le \frac{\sigma^2}{\sigma^2 + a^2}
        $$
        </font>
        <br/>
        سپس، اثبات ساده‎ای را برای این نامساوی ارائه کرد:
        <br/>
        اگر 
        $
        b
        $
        را مثبت در نظر بگیریم، می‎توانیم بنویسیم:
        $$
        \begin{align*}
        X \ge a && \equiv && X + b \ge a + b
        \end{align*}
        $$
        <p></p>
        در نتیجه،
        $$
        \begin{align*}
        P\{ X \ge a \} &= P\{X + b \ge a + b\} \\
        &\le P\{(X + b)^2 \ge (a+b)^2 \}
        \end{align*}
        $$
        <p></p>
        حال اگر نابرابری مارکوف را اعمال کنیم،
        $$
        P\{X \ge a\} \le \frac{E[(X+b)^2]}{(a+b)^2} = \frac{\sigma^2 + b^2}{(a+b)^2}
        $$
        <p></p>
        اگر به جای
        $
        b
        $
        بگذاریم
        $
        \frac{\sigma^2}{a}
        $
        ،نابرابری یک طرفه چبیشف ثابت می‎شود. نکته قابل توجه این که 
        $
        \frac{\sigma^2}{a}
        $
        مقداری برای 
        $
        b
        $
        است که به ازای آن 
        $
        \frac{\sigma^2 + b^2}{(a+b)^2}
        $
        کمترین مقدار خود را می‎گیرد.
        <br/>
        و به این ترتیب استاد چبیشف نابرابری یک طرفه خود را ثابت کرد.
        <br/>
        بعد از این اثبات، استاد به این موضوع اشاره کرد که اگر میانگین
        $ X $
        صفر نباشد، در آن صورت می‎توان در نابرابری به جای 
        $ X $
        از 
        $ X - \mu $
        استفاده کرد. یعنی:
        $$
        \begin{align*}
        P\{X - \mu \ge a\} \le \frac{\sigma^2}{\sigma^2 + a^2} \\
        \Rightarrow P\{X \ge a + \mu\} \le \frac{\sigma^2}{\sigma^2 + a^2}
        \end{align*}
        $$
</font>
</div>

<div id="conclusion" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
خلاصه و نتیجه‌گیری
        </font>
		<p></p>
		<hr>
        در این بخش با موارد زیر آشنا شدیم:
        <ul>
            <li>
            ابتدا نامساوی مارکوف را بررسی کردیم و بعد متوجه شدیم که کران بالایی که این نابرابری برای CDF یک توزیع می‎دهد خیلی دقیق نیست.
            </li>
            <li>
            در ادامه این نابرابری را به صورت شهودی و عملی مشاهده کردیم .
            </li>
            <li>
            پس از آن، این نابرابری را ثابت کردیم و سعی کردیم آن را به یک نابرابری دقیق‎تر تعمیم دهیم.
            </li>
            <li>
                نابرابری چبیشف را معرفی و از یک روش ساده برای نشان دادن آن استفاده کردیم. بعد چند تمرین در ارتباط با این دو نابرابری حل کردیم.
            </li>
            <li>
                در ادامه، نابرابری دیگری را معرفی و ثابت کردیم که به نابرابری یک‎طرفه چبیشف مشهور است. این نابرابری برای محاسبه احتمال اینکه یک داده چقدر می‎تواند از میانگین داده‎ها فاصله داشته باشد، کاربرد دارد. 
            </li>
        </ul>
