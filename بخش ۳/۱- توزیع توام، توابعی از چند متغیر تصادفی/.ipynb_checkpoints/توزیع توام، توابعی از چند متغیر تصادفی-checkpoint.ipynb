		<font color=#FF7500 size=6>
توزیع توام چیست و چرا مهم است؟
        </font>
		<p></p>
		<hr>
        یک متغیر تصادفی را خوب یاد گرفتیم. یاد گرفتیم درباره توزیع آن صحبت کنیم، تحلیلش کنیم و توابع پیچیده‌تری از روی آن بسازیم.
        <br>
        اما دنیا بسیار پیچیده تر از این است! اغلب موارد میخواهیم تاثیر چندین متغیر تصادفی را به صورت همزمان بررسی کنیم. مثال؟
        <ul>
            <li>
                احتمال سرطان داشتن را می‌خواهیم بررسی کنیم. اما با در نظر گرفتن سن، وزن و خوراکی‌های مورد علاقه شخص.
            </li>
            <li>
                می‌خواهیم احتمال تاثیرگذار بودن یک تبلیغ را بدانیم، اما علاقه‌مندی‌های شخص بسیار مهم است.
            </li>
        </ul>
        <br>
        <span style="color:red">
        توزیع هر متغیر رو جداگانه داریم دیگه! آیا این کافی نیست؟؟
        </span> <br>
            بیابید ببینیم کافی هست یا نه.
        <br>
        احتمال اینکه یک نفر برای نهار حلیم بخورد را در نظر بگیرید. با کمی اغماض فرض کنید برای همه ایرانیان(!!) برابر
        $ 0.4 $
        باشد.
        همینطور احتمال اینکه یک نفر به عنوان نهار قیمه بخورد را برابر
        $0.5$
        در نظر بگیرید.
        <br>
        اما حال بیاید به ترکیب این اتفاقات با هم فکر کنیم! ببینیم چقدر احتمال دارد که یک نفر این دو کار را با هم انجام دهد!
        <br>
        خب برای یک ایرانی عادی خوردن هم دو با هم تقریبا غیرممکن است!
        پس اگر جدولی مانند زیر بکشیم، که در تقاطع هر سطر و ستون احتمال انجام هر دو اتفاق سطر و ستون را نوشته‌ایم، در محل تلاقی
        "حلیم خوردن"
        و
        "قیمه خوردن"
        عدد صفر را مینویسم.
        <br>
        <figure>
            <img src="resource/other.png" alt="hello" style="float:center">
            <center> <caption> احتمال خوردن حلیم و قیمه برای یک ایرانی تیپیکال!</caption> </center>            
        </figure>
        <br>
        اما در مقابل اگر شخص مورد نظر مشهدی باشد، احتمال اینکه او حلیم و قیمه را با هم بخورد به شدت بالاست!
        تا جایی که تقریبا غیر ممکن است که یک مشهدی حلیم را بدون قیمه بخورد (گرچه ممکن است قیمه را بدون حلیم بخورد)
        <br>
        پس شاید جدول زیر نشان‌دهنده احتمال وقوع اتفاقات مختلف باشد:
        <br>
        <figure>
            <img src="resource/mashhadi.png" alt="hello" style="float:center">
            <center> <caption> احتمال خوردن حلیم و قیمه برای یک مشهدی </caption> </center>            
        </figure>
        <br>
        پس دیدیم چطور توزیع توام دو متغیر اطلاعات بسیار بیشتری از یک تک تک توزیع‌ها به ما می‌دهد
        و در این مورد خاص به ما کمک کرد اطلاعاتی درباره محل زندگی یک شخص بدست بیاوریم!
    </font>
</div>

<div id="sec_2" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
تعریف توزیع توام
        </font>
		<p></p>
		<hr>
        توزیع توام گسسته به نوعی احتمال رخ دادن چند اتفاق به صورت همزمان است. به عبارتی میتوان نوشت:
        <br>
        $$
        P_{X,Y}(x,y) = Pr(X=x \ , \ Y=y)
        $$
        جدول بالا درباره حلیم و قیمه نمونه بسیار خوبی هست.
        <br>
        اما حالت پیوسته کمی متفاوت است. باید تعاریف قبلی را دوباره تکرار کنیم. تابع توزیع تجمعی یا همان
        CDF
        برای یک تابع توام اینگونه تعریف می‌شود:
        $$
        F_{X,Y}(x,y) = Pr(X<x \ ,\ Y<y)
        $$و مشابه قبل تابع توزیع چگالی یا همان
        PDF
        با مشتق‌گیری از تابع توزیع تجمعی به دست می‌آید. البته با این تفاوت که این‌بار دو متغیر داریم. به عبارتی:
        $$
        f_{X,Y} (x,y)= \frac{\partial ^ 2}{\partial x \partial y} F_{X,Y}(x,y) 
        $$تعاریف بالا برای تعداد متغیرهای بیشتر از این هم صادق است. به این شکل:
        $$
        F_{X_1,X_2,...,X_n}(x_1, x_2, ..., x_n) = Pr(X_1<x_1 \ , \ X_2<x_2 \ , ... ,\ X_n<x_n) \\
        f_{X_1,X_2,...,X_n}(x_1, x_2, ..., x_n) = \frac{\partial ^ n}{\partial x_1 \partial x_2 ... \partial x_n} F_{X_1,X_2,...,X_n}(x_1, x_2, ..., x_n)
        $$
        در شکل زیر توزیع تجمعی و چگالی را برای دو متغیر می‌بینید. این تابع روی بازه
        $ [-3, 3] $
        تعریف شده است.
        <figure>
            <img src="resource/cdf.png" alt="hello" style="float:center">
            <center> <caption> تابع توزیع تجمعی (F) روی بازه‌ی [3, 3-] </caption> </center>            
        </figure>
        <figure>
            <img src="resource/pdf.png">
            <center> <caption>  تابع چگالی احتمال  (f) روی بازه‌ی [3, 3-] </caption> </center>
        </figure>
        <br>
        می‌توانید برای دیدن این شکل‌ها با دقت بیشتر به منبع دوم مراجعه کنید. 
    با این تعریف و در نظر داشتن مشابه‌های تک متغیره، چند ویژگی مهم برای توزیع توام وجود داره:
    $$
        F(-\infty, -\infty) = 0 , F(\infty, \infty) = 1 \\
        \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} f_{X,Y}(x,y) \ dx dy = 1
    $$
    البته در مواردی که دامنه توابع کل اعداد حقیقی نیست، به جای 
    $-\infty$
    و
    $\infty$
    کران بالا و پایین دامنه را قرار می‌دهیم.
    <br>
    مثلا در شکل بالا دامنه
    $(-3,3)$
    است و معادلاً داریم:
    $$
        F(-3,-3) = 0, F(3,3) = 1 \\
        \int_{-3}^{3} \int_{-3}^{3} f_{X,Y}(x,y) \ dx dy = 1
    $$
    </font>
</div>

	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
چگونه از توزیع توام استفاده کنیم؟
        </font>
		<p></p>
		<hr>
        یک مثال ساده را در نظر بگیرید:
        <figure>
            <img src="resource/discrete-sample.PNG">
            <center> <caption> برگرفته از MIT Courseware, Introduction to Probability and Statistics </caption> </center>
        </figure>
        <br>
        
        در این مثال
        $X$
        و
        $Y$
        متغیر تصادفی نتیجه دو تاس هستند. حال دنبال احتمال این هستیم که
        $Y-X \geq 2$
        باشد را حساب کنیم.
        از روی شکل می‌توان دید که این رویداد حتما در یکی از حالاتی که با رنگ نارنجی مشخص شده است رخ می‌دهد و با جمع این احتمال‌ها می‌توانیم جواب را حساب کنیم.
        <br>
        در حالت پیوسته با حالت مشابهی مواجه‌ایم. صرفا کافیست روی ناحیه‌ای که ویژگی مورد نظر ما را دارد از تابع چگالی انتگرال بگیریم.
        
        <br>
        در یک مثال این را بررسی میکنیم:
        
	</font>
</div>

<p></p>
<br/>
<div id="sec_conc_geometric" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
            مثال:
        </font>
        
        فرض کنید
        $X$ 
        و 
        $Y$ 
        دو متغیر تصادفی با توزیع توام به شکل زیر هستند:
        $$
            f_{X,Y}(x,y) = 4xy
        $$
        نشان دهید این یک توزیع است و ناحیه ای که نشاندهنده پیشامد
        $X>Y$
        است را نشان دهید. احتمال این اتفاق را حساب کنید.
        <br>
        جواب:
        <br>
        با استفاده از رابطه‌ی زیر نشان می‌دهیم یک توزیع است:
        $$
        \int_0^1\int_0^1 f_{X,Y}(x,y) dxdy = \int_0^1\int_0^1 4xy\ dxdy = \int_0^1 2x\ dx = 1
        $$
        برای پیدا کردن احتمال پیشامد
        $X>Y$
        کافیست ناحیه‌ای را پیدا کنیم که این اتفاق می‌افتد. شکل زیر این ناحیه را نشان می‌دهد.
        <figure>
            <img src="resource/occurence.PNG">
            <center> <caption> برگرفته از MIT Courseware, Introduction to Probability and Statistics </caption> </center>
        </figure>
        
        احتمال مورد نظر هم به این صورت حساب می‌شود:
        $$
            \int \int_{X>Y} f_{X,Y}(x,y) dxdy = \int_0^1 (\int_{0}^{x} 4xy \ dy)\ dx = \int_0^1 4x \frac{x^2}{2}\ dx = \frac{2}{3}
        $$
        
	</font>
</div>

<div id="sec_4" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
توابعی از چند متغیر تصادفی
        </font>
		<p></p>
		<hr>
        حال که بلدیم از توزیع توام استفاده کنیم، به راحتی میتوانیم توزیع توابعی از چند متغیر تصادفی را حساب کنیم.
        <br>
        		<font color=#FF7500 size=6>
مثال:
        </font>
         میخواهیم توزیع متغیر تصادفی
        $ Z = X + Y $
        حساب کنیم.
        $$
        F_Z (z) = Pr[Z \leq z] = Pr[ X + Y \leq z ] \\
                = \int \int_{X+Y\leq z} f_{X,Y}(x, y)
                = \int_{-\infty}^{\infty} \int_{-\infty}^{z-y} f_{X,Y}(x, y)\ dx dy
        $$
        و تابع چگالی احتمال هم با مشتق‌گیری از این تابع به دست می‌آید:
        $$
        f_Z(z) = \frac{d}{dz} F_Z(z) = \int_{-\infty}^{\infty}  f_{X,Y}(x, z-y)\ dx dy
        $$
        با همین روش میتوانیم توزیع هر تابعی از این دو متغیر را به دست آوریم.
	</font>
</div>

<div id="sec_conclusion" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
            توزیع توام بین چند تابع از چند متغیر تصادفی
        </font>
		<p></p>
		<hr>
        یاد گرفته‌ایم که از چند متغیر تصادفی توابعی بسازیم. اما ارتباط بین این متغیر‌ها چگونه است؟ 
        <br>
        حالت ساده دو متغیره را در نظر بگیرید:
        <br>
        $$
        U = g_1(X,Y)
        V = g_2(X,Y)
        $$
        فرض کنید این تبدیلات وارون‌پذیر باشند و داشته باشیم:
        $$
        X = h_1(U,V)
        Y = h_2(U,V)
        $$
        در این صورت می‌توان ثابت کرد:
        $$
        Jacoubian(u,v) = 
        \begin{bmatrix}
            \frac{\partial x}{\partial u} & \frac{\partial x}{\partial v} \\
            \frac{\partial y}{\partial u} & \frac{\partial y}{\partial v}
        \end{bmatrix}
        $$
        و خواهیم داشت:
        <br>
        $$
        x = h_1(u,v),
        y = h_2(u,v) \\
        \Rightarrow f_{U,V}(u,v) = |Jacoubian(u,v)|\ f_{X,Y}(x,y) 
        $$
	</font>
</div>

<div id="sec_conclusion" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
نتیجه‌گیری
        </font>
		<p></p>
		<hr>
        در این بخش با توزیع توام آشنا شدیم و تفاوت آن را با توزیع‌های تک تک متغیر‌های دیدیم. فهمیدیم که:
        <br>
        <ul>
            <li>
                توزیع توام اطلاعاتی بیشتر از توزیع تک متغیره در اختیار ما قرار می‌دهد.
            </li>
            <li>
                از روی توزیع‌های تک تک متغیرها نمی‌توان توزیع توام را بدست آورد.
            </li>
        </ul>
        همینطوری یاد گرفتیم چگونه از توزیع توام برای محاسبه پیشامد‌های دیگر و همینطور محاسبه توزیع متغیر تصادفی استفاده کنیم طوری که 
        <ul>
            <li>
                احتمال وقوع هر پیشامدی از جمع زدن تابع چگالی روی ناحیه مربوط به آن پیشامد به دست می‌آید.
            </li>
        </ul>
        و در آخر ارتباط متغیر‌های تصادفی را با توابعی که از روی آنها ساخته می‌شوند را دیدیم و فهمیدیم:
        <ul>
            <li>
                چگونه توزیع یک تابع از چند متغیر را محاسبه کنیم
            </li>
            <li>
                چگونه روابط بین توابع متغیر‌های تصادفی را پیدا کنیم. 
            </li>
        </ul>
	</font>
