<div id="sec_intro" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مقدمه
        </font>
		<p></p>
		<hr>
ما در زندگی با فرضیات مختلفی روبه‌رو می‌شویم و می‌خواهیم درست یا غلط بودن این فرضیات را بررسی کنیم. 
<br>
برای مثال: فرض کنید که شما مقداری پول دارید و دوست دارید بدانید که آیا خرید دلار به نفعتان هست یا نه. فرض اول این است که خرید دلار به نفعتان است که به آن $ H_{0}$ می‌گوییم و فرض دوم این است که خرید دلار به نفعتان نیست که به آن$ H_{1}$ می‌گوییم.
<br>        
مثال دیگر: فرض کنید که یک شرکت دارویی دارید، به تازگی داروی جدیدی تولید کرده‌اید. می‌خواهید ببینید که آیا این دارو برای بیماری سرطان مواثر است یا خیر. در اینجا نیز مواثر بودن را$ H_{0}$ و موثر نبودن را$ H_{1} $درنظر میگیریم.
<br>
        در آمار به فرض $H_{0}$، فرض صفر یا پیش فرض و به فرض $H_{1}$، فرض جایگزین می‌گویند.
<br>
برای رد یا تایید فرض باید مشاهدات را بررسی کنیم و با توجه به آن‌ها فرض را رد یا تایید کنیم . برای این‌کار نیاز به استفاده از آزمون فرض داریم.
<br>
برای درک بهتر به مثال زیر توجه کنید.
</div>

<br />
<div id="sec_problem" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
مثال
        </font>
		<p></p>
		<hr>
 می‌خواهیم بررسی کنیم که آیا یک سکه عادلانه‌است یا خیر.
   <br>
ابتدا مسٔله را به زبان آماری بررسی می‌کنیم:
        <br>
فرض میکنیم $\theta$ احتمال این است که سکه رو بیاید پس در این صورت فرض صفر و جایگزین به صورت زیر تعریف می‌شود$$H_{0}  : \theta = \frac{1}{2}$$
$$H_{1} : \theta \neq \frac{1}{2}$$
<br>
پاسخ:
<br>
ابتدا باید یک آزمون طراحی کنیم تا به وسیله آن فرض صفر را رد یا تایید کنیم.
برای این‌کار نیاز به مشاهده داریم، یک سکه را ۱۰۰ بار می‌اندازیم و X را تعداد دفعاتی که رو می‌آید در نظر می‌گیریم.
$$X  \sim Binomial\left(100, \theta \right)$$
<br>
در صورتی که سکه عادلانه باشد انتظار می‌رود که $\theta$ برابر $\frac{1}{2} $ باشد، در این صورت X باید عددی نزدیک به ۵۰ باشد. پس اگر X نزدیک به ۵۰ باشد فرض صفر را قبول می‌کنیم و در غیر این‌صورت فرض صفر را رد می‌کنیم.
$$\mid X - 50 \mid  \leq t  \Rightarrow accept\  H_{0}$$
$$\mid X - 50 \mid > t  \Rightarrow accept\  H_{1}$$
<br>
با توجه به فرمول بالا نیاز است که یک t در نظر بگیریم تا با استفاده از آن درباره فرض صفر تصمیم گیری کنیم.
<br>
در صورتی که t را کم و یا زیاد انتخاب کنیم می‌تواند بر روی احتمال خطا تاثیر بگذارد. یکی از خطا‌هایی که می‌تواند به وجود آید این است که با درست بودن فرض صفر ما فرض صفر را رد کنیم، این خطا را خطای نوع ۱ می‌نامیم.
$$P\left(type\ I\ error\right) = P\left(\mid X - 50\mid > t \mid H_{0}\right)$$
<br>
حال برای آنکه ما t را مشخص کنیم، نیاز است که برای خود مشخص کنیم که چه مقدار خطای نوع ۱ برای مدل ما قابل قبول است.
$$P\left(type\ I\ error\right)  \leq  \alpha = 0.05$$
<br>
در اینجا به $\alpha$ سطح معنی‌دار می‌گویند.
<br>
توزیع X را داریم پس فقط لازم است سطح معنی‌دار مورد نیاز را انتخاب کنیم. 
با توجه به اینکه n زیاد است می‌توانیم از CLT استفاده کنیم و توزیع چندجمله‌ای را با توزیع نرمال تخمین بزنیم.
$$Y = \frac{X - n\theta_{0}}{\sqrt{n\theta_{0}\left(1 - \theta_{0}\right)}} = \frac{X - 50}{5}$$
<br>
حال Y یک متغییر تصادفی از توزیع نرمال است.
$$P\left(type\ I\ error\right) = P\left(\mid X - 50\mid > t \mid H_{0}\right) = P\left(\mid \frac{X - 50}{5}\mid > \frac{t}{5}\mid H_{0}\right) = P\left(\mid Y\mid > \frac{t}{5}\mid H_{0}\right)$$
<br>
حال به‌جای پیدا کردن t باید به دنبال $\frac{t}{5}$ باشیم. برای راحتی کار به‌جای $\frac{t}{5}$ از c استفاده می‌کنیم
$$\alpha= P\left(\mid Y \mid> c\right) = 1 - P\left(-c \leq Y \leq c\right) \approx2 - 2\phi\left(c\right) 
\\
0.05 = 2 - 2\phi\left(c\right) \Rightarrow c = \phi^{-1}\left(0.975\right) = 1.96$$
<br>
حال اگر c در بازه $A = \left[-1.96, 1.96\right]$ باشد، ما فرض صفر را قبول می‌کنیم. به بازه‌ی A ، بازه قبول می‌گوییم. 
<img src="figs/acceptance_region.png">
<br>
بازه بالا برای Y است، حال بازه مناسب برای X را بدست ‌می‌آوریم.
$$Y = \frac{X - 50}{5} \Rightarrow\begin{cases}\mid X - 50\mid \leq 9.8 & accept\ H_{0}\\\mid X - 50\mid > 9.8& accept\ H_{1}\end{cases}$$
<br>
پس X ها بین ۴۱ تا ۵۹ را قبول می‌کنیم.
<br>
       در مورد مثال بالا به نکات زیر توجه کنید:
        <br/>
        ۱) در مورد رد کردن فرض صفر باید محتاطانه عمل کنیم. یعنی تا اثبات قطعی‌ای برای رد کردن آن نداریم نمی‌توانیم سکه را غیرمنصفانه اعلام کنیم.
        <br/>
        ۲) ناحیه‌ی رد از داده‌هایی تشکیل شده است که تحت فرض صفر به طور افراطی عمل می‌کنند. در این مثال دو طرف تابع احتمال که مقادیر خیلی کمی داشتند را در نظر گرفتیم. اینکه چقدر از مرکز تابع احتمال فاصله بگیریم به سطح اهمیت (significance level یا $\alpha$) بستگی دارد که به زودی به آن می‌پردازیم.
        <br/>
        ۳) در صورتی که سکه ۳۰ بار شیر بیاید می‌گوییم که آماره‌ی تست در ناحیه‌ی عدم رد قرار دارد. و حتی در صورتی که سکه ۵۰ بار شیر بیاید هم نمی‌توانیم فرض صفر را درست اعلام کنیم، و تنها می‌توانیم بگوییم که رد نشده است.
</div>

<div id="definitions" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
		<font color=#FF7500 size=6>
تعاریف
</font>
		<p></p>
		<hr>
در مثال بالا تعاریفی را به صورت پراکنده بیان کردیم که در اینجا تمام آن‌ها را به صورت منسجم می‌نویسیم.
<br>
فرض اولیه یا فرض صفر $H_{0}$ : فرضی است که در مورد پذیرش یا عدم پذیرش آن بحث می‌کنیم.
<br>
فرض جایگزین یا فرض مقابل $H_{1}$ : که در صورت عدم پذیرش $H_{0}$ درباره‌ی پذیرش یا عدم پذیرش آن بحث می‌کنیم.
<br>
X : داده‌ای است که از داده اصلی مشاهده کرده‌ایم.
<br>
ناحیه‌ی رد : اگر X در این ناحیه باشد $H_{0}$ را به ازای $H_{0}$ رد می‌کنیم.
<br>
ناحیه‌ی قبول: اگر X در این ناحیه باشد $H_{0}$ را قبول می‌کنیم.
<br>
آماره: هر اطلاعاتی که از داده‌ها به دست می‌آید آماره نام دارد. برای مثال با داشتن n داده از یک متغیر تصادفی می‌توانیم میانگین آن داده‌ها را به دست آوریم که به آن یک آماره می‌گویند. 
</div>

انواع خطا

        <br/> 
        ۱) خطای نوع اول: که در آن $H_{0}$ را به طور اشتباه رد کرده‌ایم.
        <br/>
        ۲) خطای نوع دوم: که در آن $H_{0}$ را اشتباهی رد نکرده‌ایم (یا پذیرفتیم). برای مثال سکه منصفانه نبوده و ما آن را منصفانه تلقی کردیم.
